## Основные теоретические положения
Случайный вектор $\vec{\xi} = (\xi_1, ..., \xi_n)^\intercal : \Omega \rightarrow \mathbb{R}^n$ называют нормальным (гауссовым), если существует вектор независимых стандартных нормальных случайных величин $\vec{\eta} = (\eta_1, ..., \eta_n)^\intercal$, матрица $A$ размерности $n \times m$ и вещественный вектор $\vec{b} = (b_1, ..., b_n)^\intercal$, такие что

$\vec{\xi} = A\vec{\eta} + \vec{b}$

Плотность распределения нормального случайного вектора:

$\rho_{\vec{\xi}}(x_1, ..., x_n) = \frac1{(2\pi)^{\frac{n}2} \sqrt{\det{\Sigma}}} \exp\left(-\frac12 (\vec{x} - \mu)^\intercal \Sigma^{-1} (\vec{x} - \mu) \right)$,

где $\mu = \mathrm{E}_{\vec{\xi}} = (\mathrm{E}_{\xi_1}, ..., \mathrm{E}_{\xi_n})^\intercal$ - вектор мат. ожидания,\
$\Sigma = \{ \mathrm{cov}(\xi_i, \xi_j) \}_{i,j=1,1}^n$ - ковариационная матрица.

Свойства многомерного нормального распределения:
1. Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределённой.
2. Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределёнными.
3. Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму.
4. Многомерное нормальное распределение устойчиво относительно линейных преобразований: если $\vec{\xi} \sim \mathcal{N}\left(\vec{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, $\vec{b}$ - произвольный вектор размерности $n$, то $A\vec{\xi} + \vec{b} \sim \mathcal{N}\left( A\vec{\mu}+\vec{b}, A \Sigma A^\intercal \right)$.

Условное распределение компоненты нормального случайного вектора:\
$\rho_{\xi|\eta=y}(x) = \frac{\rho_{\xi,\eta}(x,y)}{\rho_{\eta}(y)}$,
где $\rho_\eta(y)$ - одномерное нормальное распределение.

### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $\rho_{\xi,\eta}$.

$q(x, y) = 3x^2 - 3xy + 7y^2 - 9x - 8y + 13 =\\
= (3x^2 - 3xy - 9x) +7y^2 - 8y + 13 =\\
= 3(x^2 - x(y+3)) + 7y^2 - 8y + 13 =\\
= 3(x^2 - 2x\frac{y+3}2 + (\frac{y+3}2)^2 - (\frac{y+3}2)^2) + 7y^2 - 8y + 13 =\\
= 3(x - \frac{y+3}2)^2 - 3\left(\frac{y+3}2\right)^2 + 7y^2 - 8y + 13 = \\
= 3(x - \frac{y}2 -\frac32)^2 + \frac{25y^2 - 50y + 25}4 =\\
= 3((x-2) - \frac12(y-1))^2 + \frac{25}4(y-1)^2 = \\
= 3(x-2)^2 - 3(x-2)(y-1) + 7(y-1)^2 =\\
= 3(x-2)^2 - 2 \cdot 1.5(x-2)(y-1) + 7(y-1)^2$

$\mu = \begin{pmatrix} 2 \\ 1 \end{pmatrix}$,
$\Sigma^{-1} = \begin{pmatrix} 3 & -1.5 \\ -1.5 & 7 \end{pmatrix}$

$\Sigma = \frac1{\begin{vmatrix} 3 & -1.5 \\ -1.5 & 7 \end{vmatrix}} \begin{pmatrix} 7 & 1.5 \\ 1.5 & 3 \end{pmatrix} = \frac2{75} \begin{pmatrix} 14 & 3 \\ 3 & 6 \end{pmatrix}$

$\mathrm{D}_\xi = \frac{28}{75}, \mathrm{D}_\eta = \frac{4}{25}, \mathrm{cov}(\xi, \eta) = -\frac2{25}$\
$\rho(\xi, \eta) = \frac{\mathrm{cov}(\xi, \eta)}{\sqrt{\mathrm{D}_\xi \mathrm{D}_\eta}} = \frac{-\frac2{25}}{\sqrt{\frac{28}{75} \cdot \frac4{25}}} = -\frac{\sqrt3}{2\sqrt7}$

$\mathrm{C} = \frac1{(2\pi)^\frac{n}2 \sqrt{\det{\Sigma}}} = \frac{5\sqrt3}{4\pi}$

$\rho_{\xi, \eta}(x, y) = \frac{5\sqrt3}{4\pi} \exp{\left( -\frac12 \left( \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 2 \\ 1 \end{pmatrix} \right)^\intercal \begin{pmatrix} 3 & -1.5 \\ -1.5 & 7 \end{pmatrix} \left( \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 2 \\ 1 \end{pmatrix} \right) \right)}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# distribution parameters
mean = [2, 1]
cov = [[28/75, 6/75], [6/75, 12/75]]

# plot parameters
precision = 50
interval = 3

# figure parameters
figsize = [12, 12]
fontsize = 4 * min(figsize) / 3

# distribution
distr = multivariate_normal(mean, cov)

x = np.linspace(mean[0] - interval/2, mean[0] + interval/2, num=precision)
y = np.linspace(mean[1] - interval/2, mean[1] + interval/2, num=precision)

x, y = np.meshgrid(x, y)

z = distr.pdf(np.dstack((x, y)))

plt.rcParams.update({'font.size': fontsize})
fig = plt.figure(figsize=figsize)
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(x, y, z, cmap = "plasma")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("$\\rho_{\\xi, \\eta}(x, y)$")
plt.show()
```

$\mathrm{E} \begin{pmatrix} \xi_4 \\ \eta_4 \end{pmatrix} = \begin{pmatrix} -2 & -5 \\ -3 & -5 \end{pmatrix} \begin{pmatrix} 2 \\ 1 \end{pmatrix} = \begin{pmatrix} -2 \cdot 2 - 5 \\ -3 \cdot 2 - 5 \end{pmatrix} = \begin{pmatrix} -9 \\ -11 \end{pmatrix}$

### 5. Найти условное распределение $\xi$ при условии $\eta$. Вычислить $E(\xi|\eta)$ и $D(\xi|\eta)$.

$\rho_{\xi|\eta=y}(x) = \frac{\rho_{\xi,\eta}(x,y)}{\rho_{\eta}(y)}$

$\rho_\eta(y) = \frac1{\sqrt{2\pi} \sqrt{\det{\mathrm{D}_\eta}}} \exp{\left( -\frac12 \frac1{\mathrm{D}_\eta} (y - \mathrm{E}_\eta)^2 \right)} = \frac5{2\sqrt{2\pi}} \exp{\left( -\frac{25}8 (y-1)^2 \right)}$

$\rho_{\xi|\eta=y}(x) = \frac{-\frac12 \left( 3x^2 - 3xy + 7y^2 - 9x - 8y + 13 \right) }{\frac5{2\sqrt{2\pi}} \exp{\left( -\frac{25}8 (y-1)^2 \right)}} = \frac{\sqrt2\sqrt3}{2\sqrt\pi} \exp{\left( -\frac12 \cdot \frac34 \left( 4x^2 -4xy +y^2 -12x + 6y + 9 \right) \right)}= \\
= \frac{\sqrt2\sqrt3}{2\sqrt\pi} \exp{\left( -\frac12 \cdot \frac34 \left( 4(x^2 - x(y+3)) + (y+3)^2 \right) \right)} =\\
= \frac{\sqrt2\sqrt3}{2\sqrt\pi} \exp{\left( -\frac12 \cdot \frac34 \left( (2x)^2 - 2\cdot2x\cdot(y+3) + (y+3)^2 \right) \right)} = \\
=\frac{\sqrt2\sqrt3}{2\sqrt\pi} \exp{\left( -\frac32 \left( 4(x - \frac{y+3}2)^2 \right) \right)}$

$\rho_{\xi,\eta=y}(x) = \mathrm{C}_\xi \exp{\left( -\frac12 \cdot \frac1{\mathrm{D}_{\xi,\eta=y}} (x - \mathrm{E}_{\xi,\eta=y})^2 \right)}$

$\Rightarrow \mathrm{E}_{\xi,\eta=y} = \frac{y+3}2, \mathrm{D}_{\xi,\eta=y} = \frac13$
