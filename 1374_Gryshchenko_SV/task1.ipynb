## Основные теоретические положения

Случайный вектор $\overrightarrow{\xi}=(\xi_{1},\xi_{2},...,\xi_{n})^{T}: \Omega\to R^{n}$ называют нормальным, если существует вектор независимых стандартных нормальных СВ $\overrightarrow{\eta}=(\eta_{1},\eta_{2},...,\eta_{m})^{T}$, матрица A размерности $m\times n$ и вещественный вектор $B=(b_{1},b_{2},...,b_{m})^{T}$, такие, что $\overrightarrow{\xi}= A\overrightarrow{\eta}+B$.

Плотность распределения нормального вектора с вектором матожиданий $\overrightarrow{\mu}$ и матрицей ковариации $\Sigma$ имеет вид: $p_{\overrightarrow{\xi}}(x_{1},x_{2},...,x_{n})=\frac{1}{(2\pi)^{\frac{n}{2}}\sqrt {det \Sigma}}\exp\left (- \frac{1}{2} \left(\overrightarrow{x}-\overrightarrow{\mu} \right)^{T} \Sigma^{-1}\left(\overrightarrow{x}-\overrightarrow{\mu} \right)\right)$

Ковариационная матрица представима в виде $\Sigma=\begin{pmatrix} \mathbb{D}\xi & \mathop{\mathrm{cov}}\nolimits(\xi, \eta) \\ \mathop{\mathrm{cov}}\nolimits(\eta, \xi)& \mathbb{D}\eta\end{pmatrix}$
т.е. матрица ковариаций симметрична относительно диагонали, на которой находятся дисперсии СВ.

Условное распределение $\xi$ при условии $\eta$ вычисляется по формуле: $p_{\xi|\eta=y} (x)=\frac{p_{\xi,\eta} (x,y)}{p_{\eta} (y)}$

Дана плотность двумерного нормального распределения: $p_{\xi,\eta} (x,y)=C\cdot exp\left (-\frac{1}{2}(3x^{2}-3xy+7y^{2}-9x-8y+13 ) \right )$

### 1.Вычислим вектор математических ожиданий и ковариационные характеристики случайного вектора. Построим график плотности распределения $p_{\xi,\eta}\left (x,y\right )$et

$q(x,y)=3x^{2}-3xy+7y^{2}-9x-8y+13=(3x^{2}-3xy-9x)+7y^{2}-8y+13=(3x^{2}-3x\cdot(y+3))+7y^{2}-8y+13=$
$=3\cdot( x^{2}-x( y+3))+7y^{2}-8y+13=3\cdot\left( x^{2}-x( y+3)+\left( \frac{y+3}{2} \right)^{2}\right)-3\cdot\left( \frac{y+3}{2} \right)^{2}+7y^{2}-8y+13=$
$=3\cdot\left( x-\left( \frac{y}{2}+\frac{3}{2} \right)\right)^{2}-3\cdot\left( \frac{y+3}{2} \right)^{2}+7y^{2}-8y+13=3\cdot\left( x- \frac{y}{2}-\frac{3}{2} \right)^{2}-3\cdot\left( \frac{y^{2}+6y+9}{4} \right)+\frac{28y^{2}}{4}-\frac{32y}{4}+\frac{52}{4}=3\cdot\left( x- \frac{y}{2}-\frac{3}{2} \right)^{2}+\left( \frac{25y^{2}-50y+25}{4} \right)=$
$=3\cdot\left( x- \frac{y}{2}-\frac{3}{2} \right)^{2}+\frac{25}{4}\cdot( y-1 )^{2}=3\cdot\left( (x-2)- \frac{y}{2}+\frac{1}{2} \right)^{2}+\frac{25}{4}( y-1 )^{2}=$
$=3\cdot\left( (x-2)- \frac{1}{2}\cdot(y-1) \right)^{2}+\frac{25}{4}\cdot( y-1 )^{2}=3\cdot\left( (x-2)^{2}- (x-2)\cdot(y-1)+\frac{1}{4}\cdot(y-1)^{2} \right)+\frac{25}{4}\cdot( y-1 )^{2}=$
$=3\cdot(x-2)^{2}-3\cdot (x-2)\cdot(y-1)+\frac{3}{4}\cdot(y-1)^{2}+\frac{25}{4}\cdot( y-1 )^{2}=3\cdot(x-2)^{2}-3\cdot (x-2)\cdot(y-1)+7\cdot(y-1)^{2}$

$\mathbb{E}\xi\eta=\begin{pmatrix}2 \\ 1\end{pmatrix}$

$\Sigma^{-1}=\begin{pmatrix}3&-1.5 \\ -1.5&7\end{pmatrix}$

$\Sigma=\frac{4}{75}\cdot\begin{pmatrix}7&1.5 \\ 1.5&3\end{pmatrix}$

$\mathbb{D}\xi=\frac{28}{75}$

$\mathbb{D}\eta=\frac{4}{25}$

$\mathop{\mathrm{cov}}\nolimits(\xi, \eta)=\frac{2}{25}$

$\rho(\xi, \eta)=\frac{\frac{2}{25}}{\sqrt{\frac{4}{75}}\cdot\sqrt{\frac{2}{25}}}=\frac{\sqrt{6}}{2}$

$C=\frac{1}{2\pi\cdot\sqrt{\frac{4}{75}}}=\frac{5\sqrt{3}}{4\pi}$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-2, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
mu = np.array([2, 1]) # вектор математических ожиданий
cov = np.array([[28/75, 2/25], [2/25, 4/25]]) # ковариационная матрица
inv_cov = np.linalg.inv(cov) # обратная ковариационная матрица
det_cov = np.linalg.det(cov) # определитель ковариационной матрицы

def pdf(x, y):
    xy = np.array([x, y]) - mu
    exponent = -0.5 * np.dot(np.dot(xy, inv_cov), xy)
    return (1 / (2 * np.pi * np.sqrt(det_cov))) * np.exp(exponent)
    
Z = np.vectorize(pdf)(X, Y)
# построение графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```
