## Основные теоретические положения
**Многомерное нормальное распределение** — это обобщение одномерного нормального распределения. Случайный вектор, имеющий многомерное нормальное распределение, называется гауссовским вектором.  
**Случайный вектор $\vec\xi=(\xi_1, \xi_2, ..., \xi_n)^\intercal: \Omega \rightarrow \mathbb{R}^n$** - нормальный (гауссовский), если существует $\vec\xi=А\vec\eta+B$, где вектор независимой стандартной нормальнной случайной величины $\vec{\eta} = (\eta_1, ... , \eta_n)^\intercal $, матрица $А$ размерности $n \times m$ и вещественный вектор $B = (b_1, ... , b_n)^\intercal$.  
**Плотность нормального вектора имеет вид:**  
$p_{\vec{\xi}}(x_1, ... ,x_{n}) = \cfrac{1}{2\pi\sqrt{det\Sigma}}\cdot\exp\Bigl\{-\cfrac{1}{2}\biggl[\vec{x}-\mu\biggr]^{\intercal}\Sigma^{-1}\biggl[\vec{x}-\mu\biggr]\Bigr\}$  
**Свойства многомерного нормального распределения:**  
1) Линейные комбинации случайных переменных, принадлежащих многомерному нормальному распределению, также являются нормально распределеннными.
2) Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормальными распределениеми.
3) Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму)
4) Многомерное нормальное распределение устойчиво относительно линейных преобразований.

## Выполнение работы
Задача 1: Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора.  
Решение:  
Для вычисления математических ожиданий преобразуем выражение, стоящее под экспонентой, методом Лагранжа  
$3x^2-6xy+11y^2-12x-4y+20$  
$3(x^2-2x(y+2))+11y^2-4y+20$  
$3(x^2-2x(y+2)+(y+2)^2)-3(y+2)^2+11y^2-4y+20$  
$3(x-y-2)^2+8(y^2-2y+1)$  
$3(x-y-2)^2+8(y-1)^2$  
$3(x-3-(y-1))^2+8(y-1)^2$  
$3(x-3)^2-6(x-3)(y-1)+11(y-1)^2$ 

$\mathbb{E}\begin{pmatrix} \xi \\ \eta\end{pmatrix}=\begin{pmatrix} 3 \\ 1\end{pmatrix}$  
$\Sigma^{-1}=\begin{pmatrix} 3 & -3\\ -3 & 11 \end{pmatrix}$  
$\Sigma=\begin{pmatrix} \mathbb{D}_\xi & \mathrm{cov}(\xi{,}\eta)\\ \mathrm{cov}(\eta{,}\xi) & \mathbb{D}_\eta \end{pmatrix}$    $\Sigma=\cfrac{1}{24}\begin{pmatrix} 11 & 3\\ 3 & 3 \end{pmatrix}$  
$\mathop{\mathrm{cov}}\nolimits(\xi, \eta)=\cfrac{1}{8}$  
$\mathbb{D}_\xi=\cfrac{11}{24}$    $\mathbb{D}_\eta=\cfrac{1}{8}$  
$\rho(\xi,\eta)=\cfrac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta)}{\sqrt{\mathbb{D}_\xi}\cdot\sqrt{\mathbb{D}_\eta}}=\cfrac{\cfrac{1}{8}}{\sqrt{\cfrac{11}{24}}\cdot\sqrt{\cfrac{1}{8}}}=\cfrac{\sqrt{3}}{\sqrt{11}}$  
$C=\cfrac{1}{2\pi\sqrt{det\Sigma}}=\cfrac{1}{2\pi\sqrt{\cfrac1{24}}}=\cfrac{\sqrt{6}}{\pi}$  
$p_{\xi,\eta}(x,y)=\cfrac{\sqrt{6}}{\pi}\cdot\exp\Bigl\{-\cfrac{1}{2}\biggl[\begin{pmatrix} x \\ y\end{pmatrix}-\begin{pmatrix} 3 \\ 1\end{pmatrix}\biggr]^{\intercal}\begin{pmatrix} 3 & -3\\ -3 & 11 \end{pmatrix}\biggl[\begin{pmatrix} x \\ y\end{pmatrix}-\begin{pmatrix} 3 \\ 1\end{pmatrix}\biggr]\Bigr\}$

```python
import numpy as np
import matplotlib.pyplot as plt
fig2 = plt.figure()
ax = fig2.add_subplot(111, projection='3d')
x = range(0,5)
y = range(0,10)
xgrid, ygrid = np.meshgrid(x, y)
zgrid = np.sqrt(6)/np.pi*np.exp(-0.5*(3*np.square(xgrid-3)-6*(xgrid-3)*(ygrid-1)+11*np.square(ygrid-1)))
ax.plot_surface(xgrid, ygrid, zgrid)
ax.set_title("График плотности распределения")
plt.show()
```

Задание 2: Найти аффинное преобразование, переводящее исходный случайный вектор в стандартный нормальный.  
Решение:
$A\vec\xi+b\sim\mathcal{N}(0,E)$  
$q(x,y)=3(x-y-2)^2+8(y-1)^2$  
$q(x,y)=(\sqrt{3}x-\sqrt{3}y-2\sqrt{3})^2+(\sqrt{8}y-\sqrt{8})^2$  
$\begin{pmatrix} \xi_{st} \\ \eta_{st}\end{pmatrix}=\begin{pmatrix} \sqrt{3}& -\sqrt{3} \\ 0&\sqrt{8}\end{pmatrix}\begin{pmatrix} \xi \\ \eta\end{pmatrix}+\begin{pmatrix} -2\sqrt{3} \\ -\sqrt{8}\end{pmatrix}$  
Проверка:   
Так как мы ищем аффинное преобразование, переводящее случайный вектор в стандартный нормальный, то необходимо выполнение условий:  
$\mathbb{E}\begin{pmatrix} \xi_{st} \\ \eta_{st}\end{pmatrix}=\begin{pmatrix} 0 \\ 0\end{pmatrix}$  
$\Sigma_{st}=A\cdot\Sigma\cdot A^{\intercal}=E$   
Подставим $\begin{pmatrix} \xi \\ \eta\end{pmatrix}=\begin{pmatrix} 3 \\ 1\end{pmatrix}$ $\Sigma=\begin{pmatrix} 11&3 \\ 3&3\end{pmatrix}$  
$\mathbb{E}\begin{pmatrix} \xi_{st} \\ \eta_{st}\end{pmatrix}=\begin{pmatrix} \sqrt{3} &-\sqrt{3} \\ 0& \sqrt{8}\end{pmatrix}\begin{pmatrix} 3 \\ 1\end{pmatrix}+\begin{pmatrix} -2\sqrt{3} \\ -\sqrt{8}\end{pmatrix}=\begin{pmatrix} -2\sqrt{3} \\ -\sqrt{8}\end{pmatrix}-\begin{pmatrix} -2\sqrt{3} \\ -\sqrt{8}\end{pmatrix}=\begin{pmatrix} 0 \\ 0\end{pmatrix}$  
$\Sigma_{st}=\cfrac{1}{24}\begin{pmatrix} \sqrt{3}  &-\sqrt{3} \\ 0 & \sqrt{8}\end{pmatrix} \begin{pmatrix}11&3 \\ 3 &3\end{pmatrix}\begin{pmatrix} \sqrt{3}& 0 \\ -\sqrt{3} &\sqrt{8}\end{pmatrix}=\cfrac{1}{24}\begin{pmatrix} 8 \sqrt{3} &0 \\ 3 \sqrt{8} &3\sqrt{8}\end{pmatrix}\begin{pmatrix} \sqrt{3} &0 \\ -\sqrt{3} &\sqrt{8}\end{pmatrix}=\cfrac{1}{24}\begin{pmatrix} 24& 0 \\ 0 &24\end{pmatrix}=\begin{pmatrix} 1& 0 \\ 0 &1\end{pmatrix}$

Задание 4: Вычислить характеристики совместного распределения случайного вектора $(2\xi-2\eta,4\xi+3\eta)^{\intercal}$  
$\mathbb{E}\begin{pmatrix} \xi_{4} \\ \eta_{4}\end{pmatrix}=\begin{pmatrix} 2&-2 \\ 4&3\end{pmatrix}\begin{pmatrix} \xi \\ \eta\end{pmatrix}=\begin{pmatrix} 2&-2 \\ 4&3\end{pmatrix}\begin{pmatrix} 3 \\ 1\end{pmatrix}=\begin{pmatrix} 4 \\15\end{pmatrix}$  
$\Sigma_{4}=A\cdot\Sigma\cdot A^{\intercal}=\cfrac{1}{24}\cdot\begin{pmatrix} 2&-2 \\ 4&3\end{pmatrix}\begin{pmatrix}11&3 \\ 3&3\end{pmatrix}\begin{pmatrix} 2&4 \\ -2&3\end{pmatrix}=\begin{pmatrix} \cfrac{4}{3}&\cfrac{8}{3} \\ \cfrac{8}{3}&\cfrac{275}{24}\end{pmatrix}$   
$\mathrm{cov}(\xi,\eta)=\cfrac{8}{3}$  
$\mathbb{D}_{\xi_4}=\cfrac{4}{3}$    $\mathbb{D}_{\eta_4}=\cfrac{275}{24}$  
$\rho(\xi_4,\eta_4)=\frac{\cfrac{8}{3}}{\sqrt{\cfrac{4}{3}}\cdot\sqrt{\cfrac{275}{24}}}=\cfrac{8\sqrt{22}}{55}$  
$C=\cfrac{1}{2\pi\sqrt{det\Sigma_4}}=\cfrac{1}{2\pi\sqrt{\cfrac{49}{6}}}=\cfrac{\sqrt{6}}{14\pi}$  
$p_{\xi_4,\eta_4}(x,y)=\cfrac{\sqrt{6}}{14\pi}\cdot\exp\Bigl\{-\cfrac{1}{2}\biggl[\begin{pmatrix} x \\ y\end{pmatrix}-\begin{pmatrix} 4 \\ 15\end{pmatrix}\biggr]^{\intercal}\begin{pmatrix} 1\cfrac{79}{196} & -\cfrac{16}{49}\\ -\cfrac{16}{49} & \cfrac{8}{49} \end{pmatrix}\biggl[\begin{pmatrix} x \\ y\end{pmatrix}-\begin{pmatrix} 4 \\ 15\end{pmatrix}\biggr]\Bigr\}$

```python
fig4 = plt.figure()
ax = fig4.add_subplot(111, projection='3d')
x = range(0,10)
y = range(0,30)
xgrid, ygrid = np.meshgrid(x, y)
zgrid = np.sqrt(6)/(14*np.pi)*np.exp(-0.5*(275/196*np.square(xgrid-4)-32/49*(xgrid-4)*(ygrid-15)+8/49*np.square(ygrid-15)))
ax.plot_surface(xgrid, ygrid, zgrid)
ax.set_title("График плотности распределения")
plt.show()
```

Задание 5: Найти условное распределение $\xi$ при условии $\eta$. Вычислить $\mathbb{E}(\xi\textbar\eta)$ и $\mathbb{D}(\xi\textbar\eta)$
