## **Основные теоретические положения**


Случайный вектор $\overrightarrow{\xi} = \left(\xi_1, \dots, \xi_n\right)^\intercal:\Omega\to \mathbb{R}^n$ имеет многомерное нормальное распределение, если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1, \dots, \eta_m\right)^\intercal$, матрица $\mathbf{A}$ размерности $n \times m$ и вещественный вектор $\overrightarrow{b} = (b_1,\dots,b_n)^\intercal$, такие что: $\overrightarrow{\xi} = \mathbf{A}\overrightarrow{\eta} + \overrightarrow{b}$.

Плотность многомерного невырожденного нормального распределения имеет вид: $\mathcal{p}_{\overrightarrow{\xi}}\left(x_1, \dots, x_n\right) = \frac{1}{(2\pi)^{n/2}\sqrt{\det \Sigma}} \exp\left(-\frac{1}{2}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)^\intercal \Sigma^{-1}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)\right)$, где $\overrightarrow{\mu} = \mathbb{E}\overrightarrow{\xi} = \left(\mathbb{E}\xi_1, \dots, \mathbb{E}\xi_n\right)^\intercal$ - вектор математических ожиданий компонент, $\Sigma = \left\{\sigma_{i,j} = \mathop{\mathrm{cov}}\left(\xi_i,\xi_j\right)\right\}_{i,j=1}^n$ - ковариационная матрица.

Свойства многомерного нормального распределения:
- Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенной
- Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределенными. $\mathcal{p}_{\xi|\eta=y}(x) = \frac{\mathcal{p}_{\xi,\eta}(x,y)}{\mathcal{p}_{\eta}(y)}$
- Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму
- Многомерное нормальное распределение устойчиво относительно линейных преобразований. Если $\overrightarrow{\xi} \sim \mathcal{N}\left(\overrightarrow{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, $\overrightarrow{b}$ - произвольный вектор размерности $n \times 1$, то $A\overrightarrow{\xi}+\overrightarrow{b} \sim \mathcal{N}\left(A\overrightarrow{\mu}+\overrightarrow{b}, A \Sigma A^\intercal\right)$.

### 1. Вычисление характеристик данного случайного вектора. Построение графика плотности распределения $\mathcal{p}_{\xi,\eta}$.

${p}_{\xi,\eta}(x,y) = C \cdot \exp\left( -\frac{1}{2}\left(3x^2-4xy+6y^2-14x+21\right)\right)$

Применим метод Лагранжа для приведения исходного уравнения к каноническому виду.

$\begin{split}
  q\left(x,y\right) &= 3x^2 - 4xy + 6y^2 - 14x + 21 = \\
  &= \left(3x^2 - 4xy + 14x\right) + 6y^2 + 21 = \\
  &= \left(3x^2 - 2x\left(2y + 7\right)\right) + 6y^2 + 21 = \\
  &= 3\left(x^2 + 2x\frac{2y + 7}{3} + \left(\frac{\left(2y + 7\right)}{3}\right)^2\right) - 3\left(\frac{\left(2y + 7\right)}{3}\right)^2 + 6y^2 + 21 = \\
  &= 3\left(x - \frac{2}{3} y - \frac{7}{3}\right)^2 + \frac{4y^2 + 28y + 49}{3} + 6y^2 + 21 = \\
  &= 3\left(x - \frac{2}{3} y - \frac{7}{3}\right)^2 + \frac{14y^2 - 28y + 14}{3} = \\  
  &= 3\left(x - \frac{2}{3} y - \frac{7}{3}\right)^2 + \frac{14}{3}\left(y-1\right)^2 \\
\end{split}$

Последнее выражение понадобится при нахождении аффинного преобразования

$\begin{split}
  q\left(x,y\right) &=  3\left(\left(x - 3\right) - \frac{2}{3} \left(y - 1\right)\right)^2 + \frac{14}{3}\left(y-1\right)^2 = \\  
  &= 3\left(\left(x - 3\right)^2 -  2\cdot\frac{2}{3} \left(x - 3\right)  \left(y - 1\right) + \frac{4}{9} \left(y - 1\right)^2 \right) + \frac{14}{3}\left(y-1\right)^2 = \\  
  &= 3\left(x - 3\right)^2 - 2\cdot2\left(x - 3\right)\left(y - 1\right) + 6 \left(y - 1\right)^2 \\  
\end{split}$


Определим вектор математических ожиданий и обратную ковариационную матрицу.

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 3 \\ 1 \end{pmatrix}$

$\Sigma^{-1} = \begin{pmatrix} 3 & -2 \\ -2 & 6\end{pmatrix}$

Матрица ковариаций:

$\Sigma = \frac{1}{14} \begin{pmatrix} 6 & 2 \\ 2 & 3\end{pmatrix}$

Ковариационные характеристики:

$\mathbb{D}_\xi = \frac{6}{14} = \frac{3}{7},\ \mathbb{D}_\eta = \frac{3}{14},\ \mathop{\mathrm{cov}}(\xi, \eta) = \frac{2}{14} = \frac{1}{7}$

Коэффициент корреляции:

$\rho(\xi, \eta) = \frac{\mathop{\mathrm{cov}}(\xi, \eta)}{\sqrt{\mathbb{D}_\xi} \cdot \sqrt{\mathbb{D}_\eta}} = \frac{\frac{1}{7}}{\frac{\sqrt{3}}{\sqrt{7}} \cdot \frac{\sqrt{3}}{\sqrt{14}}} = \frac{7\sqrt{2}}{7 \cdot 3} = \frac{\sqrt{2}}{3} $

Неизвестная $C$:

$C = \frac{1}{2\pi\cdot \sqrt{\frac{1}{14}}} = \frac{\sqrt{14}}{2\pi} $

Функция плотности распределения:

$\mathcal{p}_{\xi,\eta}(x,y) = \frac{\sqrt{14}}{2\pi} \exp\left\{-\frac{1}{2}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 3 \\ 1 \end{pmatrix} \right]^\intercal \begin{pmatrix} 3 & -2 \\ -2 & 6\end{pmatrix}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 3 \\ 1 \end{pmatrix} \right]\right\}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
random_seed = 42

#Иницализация матрицы ковариаций и вектора математических ожиданий
cov = np.array([[6/14, 1/7], [1/7, 3/14]])
mean = np.array([3, 1])

#Генерация распределения
distr = multivariate_normal(cov = cov, mean = mean, seed = random_seed)

x = np.linspace(mean[0] - 3*np.sqrt(cov[0,0]), mean[0] + 3*np.sqrt(cov[0,0]), num=100)
y = np.linspace(mean[1] - 3*np.sqrt(cov[1,1]), mean[1] + 3*np.sqrt(cov[1,1]), num=100)
X, Y = np.meshgrid(x,y)

#Создание значений функции
Z = np.zeros(X.shape)
for i in range(X.shape[0]):
  for j in range(X.shape[1]):
    Z[i,j] = distr.pdf([X[i,j], Y[i,j]])

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, cmap = "plasma")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel(r'$\mathcal{p}_{\xi,\eta}(x,y)$', labelpad=5)
plt.title("График плотности вероятности случайного вектора")
plt.show()
```

Проверка вектора математических ожиданий и ковариационной матрицы полученного распределения:

$\mathbb{E}\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = \begin{pmatrix} \frac{3}{\sqrt{3}} & -\frac{2}{\sqrt{3}} \\ 0 & \sqrt{\frac{14}{3}}\end{pmatrix} \begin{pmatrix} 3 \\ 1 \end{pmatrix} + \begin{pmatrix} -\frac{7}{\sqrt{3}} \\ -\sqrt{\frac{14}{3}} \end{pmatrix} = \begin{pmatrix} \frac{9}{\sqrt{3}} - \frac{2}{\sqrt{3}} - \frac{7}{\sqrt{3}} \\ 0 + \sqrt{\frac{14}{3}} - \sqrt{\frac{14}{3}}\end{pmatrix}=\begin{pmatrix} 0 \\ 0 \end{pmatrix}$

### 3. Нахождение ортогонального преобразования, переводящего соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построение графика плотности полученного распределения.

### 4. Вычисление характеристик совместного распределения случайного вектора $(-4\xi + 2\eta, 2\xi + \eta)^\intercal$, нахождение его плотности. Построение графика плотности полученного распределения.

Данная матрица — невырожденная, т.к. $\det A = -15$. Характеристики распределения:

$\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} 1 & -2 \\ -5 & -5\end{pmatrix}\begin{pmatrix} \xi \\ \eta \end{pmatrix}$

$\mathbb{E}\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} 1 & -2 \\ -5 & -5\end{pmatrix} \begin{pmatrix} 3 \\ 1 \end{pmatrix} = \begin{pmatrix} 1 \\ -20 \end{pmatrix}$

$\Sigma_{4} = \frac{1}{14}\begin{pmatrix} 1 & -2 \\ -5 & -5\end{pmatrix}\begin{pmatrix} 6 & 2 \\ 2 & 3\end{pmatrix}\begin{pmatrix} 1 & -5 \\ -2 & -5\end{pmatrix} = \frac{1}{14}\begin{pmatrix} 2 & -4 \\ -40 & -25\end{pmatrix}\begin{pmatrix} -4 & 2 \\ 2 & 1\end{pmatrix} = \frac{2}{75}\begin{pmatrix} 296 & -100 \\ -100 & 50\end{pmatrix} = \frac{1}{14}\begin{pmatrix} 10 & 10 \\ 10 & 325\end{pmatrix} = \begin{pmatrix} \frac{5}{7} & \frac{5}{7} \\ \frac{5}{7} & \frac{325}{14}\end{pmatrix}$

\Sigma_{4}^{-1} = \frac{14}{225} \begin{pmatrix} \frac{325}{14} & -\frac{5}{7} \\ -\frac{5}{7} & \frac{5}{7}\end{pmatrix} = \begin{pmatrix} \frac{13}{9} & -\frac{2}{45} \\ -\frac{2}{45} & \frac{2}{45}\end{pmatrix}$

$C = \frac{1}{2\pi \cdot \sqrt{\frac{225}{14}}} = \frac{\sqrt{14}}{30\pi}$

Плотность полученного распределения:

$\mathcal{p}_{\xi_4,\eta_4}(x,y) = \frac{\sqrt{14}}{30\pi} \exp\left\{-\frac{1}{2}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 1 \\ -20 \end{pmatrix} \right]^\intercal \begin{pmatrix} \frac{13}{9} & -\frac{2}{45} \\ -\frac{2}{45} & \frac{2}{45}\end{pmatrix}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 1 \\ -20 \end{pmatrix} \right]\right\}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
random_seed = 42


#Иницализация матрицы ковариаций и вектора математических ожиданий
cov = np.array([[5/7, 5/7], [5/7, 325/14]])
mean = np.array([1, -20])

#Генерация распределения
distr = multivariate_normal(cov = cov, mean = mean, seed = random_seed)

#Создание сетки с интервалом 3 сигмы в каждую сторону
x = np.linspace(mean[0] - 3*np.sqrt(cov[0,0]), mean[0] + 3*np.sqrt(cov[0,0]), num=100)
y = np.linspace(mean[1] - 3*np.sqrt(cov[1,1]), mean[1] + 3*np.sqrt(cov[1,1]), num=100)
X, Y = np.meshgrid(x,y)

#Создание значений функции
Z = np.zeros(X.shape)
for i in range(X.shape[0]):
  for j in range(X.shape[1]):
    Z[i,j] = distr.pdf([X[i,j], Y[i,j]])

#Отрисовка графика
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, cmap = "plasma")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel(r'$\mathcal{p}_{\xi_{4},\eta_{4}}(x,y)$', labelpad=5)
plt.title(r"График плотности совместного распределения случайного вектора $(-4\xi + 2\eta, 2\xi + \eta)^\intercal$")
plt.show()
```

### 5. Нахождение условного распределения $\xi$ при условии $\eta$. Вычисление $\mathbb{E}(\xi|\eta)$ и $\mathbb{D}(\xi|\eta)$.

Условные распределения полученные из многомерного нормального, также являются многомерными нормальными.

$\mathcal{p}_{\xi|\eta=y}(x) = \frac{\mathcal{p}_{\xi,\eta}(x,y)}{\mathcal{p}_{\eta}(y)}$

Компоненты многомерного нормального распределения имеют одномерное нормальное распределение. Его плотность:

$\mathbb{E}\eta = 1,\ \mathbb{D}\eta = \frac{3}{14}$

$\mathcal{p}_{\eta}(y) = \frac{1}{\sqrt{2\pi} \cdot \sqrt{\frac{3}{14}}} \exp\left(-\frac{1}{2}(y - 1)\frac{14}{3}(y - 1)\right) = \frac{\sqrt{7}}{\sqrt{3\pi}} \exp\left(-\frac{1}{2}\left(\frac{14}{3}y^2 - \frac{28}{3}y + \frac{14}{3}\right)\right)$

Плотность условного распределения:

$\begin{split}
