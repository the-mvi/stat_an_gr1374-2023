## Основные теоретические положения

Случайный вектор $\overrightarrow{\xi} = \left(\xi_1, \dots, \xi_n\right)^\intercal:\Omega\to \mathbb{R}^n$ имеет многомерное нормальное распределение, если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1, \dots, \eta_m\right)^\intercal$, матрица $\mathbf{A}$ размерности $n \times m$ и вещественный вектор $\overrightarrow{b} = (b_1,\dots,b_n)^\intercal$, такие что: $\overrightarrow{\xi} = \mathbf{A}\overrightarrow{\eta} + \overrightarrow{b}$.

Плотность многомерного невырожденного нормального распределения имеет вид: $\mathcal{p}_{\overrightarrow{\xi}}\left(x_1, \dots, x_n\right) = \frac{1}{(2\pi)^{n/2}\sqrt{\det \Sigma}} \exp\left(-\frac{1}{2}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)^\intercal \Sigma^{-1}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)\right)$, где $\overrightarrow{\mu} = \mathbb{E}\overrightarrow{\xi} = \left(\mathbb{E}\xi_1, \dots, \mathbb{E}\xi_n\right)^\intercal$ - вектор математических ожиданий компонент, $\Sigma = \left\{\sigma_{i,j} = \mathop{\mathrm{cov}}\left(\xi_i,\xi_j\right)\right\}_{i,j=1}^n$ - ковариационная матрица.

Формула для плотности условного распределения:  
$p_{\xi|\eta=y}(x) = \frac{p_{\xi, \eta}(x, y)}{p_{\eta}(y)}$

Свойства многомерного нормального распределения:
- Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределенными. $\mathcal{p}_{\xi|\eta=y}(x) = \frac{\mathcal{p}_{\xi,\eta}(x,y)}{\mathcal{p}_{\eta}(y)}$
- Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму
- - Многомерное нормальное распределение устойчиво относительно линейных преобразований. Если $\overrightarrow{\xi} \sim \mathcal{N}\left(\overrightarrow{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, $\overrightarrow{b}$ - произвольный вектор размерности $n \times 1$, то $A\overrightarrow{\xi}+\overrightarrow{b} \sim \mathcal{N}\left(A\overrightarrow{\mu}+\overrightarrow{b}, A \Sigma A^\intercal\right)$.
-  Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенной

### 1.Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора.

Произведем преобразование Лагранжа
$q(x,y) = 3x^{2}+3xy+7y^{2}+12x-19y+37 = 3(x^{2} + xy + 4x) + 7y^{2} -19y +37 = 3(x^{2}+ \frac{2}{2}x(y+4)) + 7y^{2} - 19y + 37 = 3\left( x^2 + 2x\frac{y+4}{2}+ (\frac{y+4}{2})^{2} \right)- 3\left( \frac{y+4}{2} \right)^{2} +7y^{2} - 19y + 37 = 3(x + \frac{y}{2}+2)^{2} + \frac{25}{4}(y^2 - 4y + 4) = 3 \left((x+3) + \frac{1}{2}(y-2) \right)^{2} + \frac{25}{4}(y-2)^2 = 3(x+3) + 3(x+3)(y-2) + 7(y-2)$

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} -3 \\ 2 \end{pmatrix}$

$\Sigma^{-1} = \begin{pmatrix} 3 & \frac{3}{2} \\ \frac{3}{2} & 7 \end{pmatrix} \implies \det{\Sigma^{-1}} = \frac{75}{4}$

$\Sigma = \begin{pmatrix} 7 & -\frac{3}{2} \\ -\frac{3}{2} & 3 \end{pmatrix} \cdot \frac{1}{\det{\Sigma^{-1}}} = \begin{pmatrix} \frac{28}{75} & -\frac{2}{25} \\ -\frac{2}{25} & \frac{12}{75}\end{pmatrix} \implies \mathbb{D}\xi = \frac{28}{75}, \mathbb{D}\eta = \frac{4}{25}, \mathop{\mathrm{cov}}\nolimits(\xi, \eta) = -\frac{2}{25}$

$\sigma_{\xi} = \sqrt{\frac{28}{25}} = \frac{2\sqrt{7}}{5}$

$\sigma_{\eta} = \sqrt{\frac{4}{25}} = \frac{2}{5}$

$\rho_{\xi,\eta} = \frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta) }{\sigma_{\xi}\cdot\sigma_{\eta}} = \frac{-\frac{2}{25}}{\frac{2\sqrt{7}}{5}\cdot\frac{2}{5}} = -\frac{\sqrt{7}}{14}$

$C = \frac{1}{2\pi\cdot\sqrt{\det(\Sigma)}} = \frac{5\sqrt{3}}{4\pi}$

Плотность распределения: $P_{\xi\eta}(x,y) = \frac{5\sqrt{3}}{4\pi}exp\left\{ -\frac{1}{2}\cdot\left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -3 \\ 2 \end{pmatrix}\right]^{\intercal} \cdot \begin{pmatrix} 3 & \frac{3}{2} \\ \frac{3}{2} & 7 \end{pmatrix}  \cdot \left[ \begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} -3 \\ 2 \end{pmatrix} \right] \right\}$

```python
## Продолжение выполнения №1. Построим график плотности распределения pξ,η.
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 5, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

def normDist(x, y):

    mean = np.array([-3, 2])
    const = (5 * np.sqrt(3)) / (4 * np.pi)
    cov = np.array([[3, 3/2], [3/2, 7]])
    diff = np.stack((x, y), axis=-1) - mean
    power = -1/2 * np.einsum('...k,kl,...l->...', diff, cov, diff)
    return const * np.exp(power)

Z = normDist(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

### 2.Найти аффинное преобразование, переводящее исходный случайный вектор в стандартный нормальный.

Произведем нормирование q(x,y):

$q(x,y) = 3\left( x + \frac{y}{2} + 2 \right)^{2} + \frac{25}{4}(y-2)^{2} = \left( \frac{3}{\sqrt{3}}x + \frac{3}{2\sqrt(3)}y + \frac{6}{\sqrt(3)} \right)^{2} + \left( \frac{5}{2}y - 5 \right)^{2}$

A = $\begin{pmatrix} \frac{3}{\sqrt{3}} & \frac{3}{2\sqrt{3}} \\ 0 & \frac{5}{2} \end{pmatrix}$

$\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = \begin{pmatrix} \frac{3}{\sqrt{3}} & \frac{3}{2\sqrt{3}} \\ 0 & \frac{5}{2} \end{pmatrix}\cdot\begin{pmatrix} \xi \\ \eta \end{pmatrix} + \begin{pmatrix} \frac{6}{\sqrt{3}} \\ -5 \end{pmatrix} = \begin{pmatrix} \frac{3}{\sqrt{3}}\xi + \frac{3}{2\sqrt{3}}\eta + \frac{6}{\sqrt{3}} \\ \frac{5}{2}\eta - 5 \end{pmatrix}$

Произведем проверку:

$\mathbb{E}\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = \begin{pmatrix} \frac{3}{\sqrt{3}} & \frac{3}{2\sqrt{3}} \\ 0 & \frac{5}{2} \end{pmatrix}\cdot\begin{pmatrix} -3 \\ 2 \end{pmatrix} + \begin{pmatrix} \frac{6}{\sqrt{3}} \\ -5 \end{pmatrix} = \begin{pmatrix} -2\sqrt{3} \\ 5 \end{pmatrix} + \begin{pmatrix} \frac{6}{\sqrt{3}} \\ -5 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$

$\Sigma_{st} = \frac{4}{75}\cdot\begin{pmatrix} \frac{3}{\sqrt{3}} & \frac{3}{2\sqrt{3}} \\ 0 & \frac{5}{2} \end{pmatrix}\cdot\begin{pmatrix} 7 & -\frac{3}{2} \\ -\frac{3}{2} & 3 \end{pmatrix}\cdot\begin{pmatrix} \frac{3}{\sqrt{3}} & 0 \\ \frac{3}{2\sqrt{3}} & \frac{5}{2} \end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$

### 3.Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами.

$\Sigma^{-1} = \begin{pmatrix} 3 & \frac{3}{2} \\ \frac{3}{2} & 7 \end{pmatrix}$

$\begin{vmatrix} 3-\lambda & \frac{3}{2}\\ \frac{3}{2} & 7-\lambda \end{vmatrix} = (3-\lambda)\cdot(7-\lambda) - \frac{9}{4} = \lambda^{2} - 10\lambda + \frac{75}{4} = \left( \lambda-\frac{5}{2} \right)\cdot\left( \lambda-\frac{15}{2} \right)$

$\lambda_{1} = \frac{5}{2}$

$\left[\begin{array}{@{} rr|r @{\,}} \frac{1}{2} & \frac{3}{2} & 0 \\ \frac{3}{2} &  \frac{9}{2} &  0 \end{array}\right]\sim \left[\begin{array}{@{} rr|r @{\,}} \frac{3}{2} &  \frac{9}{2} & 0 \\ 0 &  0 &  0 \end{array}\right]$

$\frac{3}{2}x + \frac{9}{2}y = 0 \to 3x = -9y \to x = -3y \to \overrightarrow{\mathrm{V}_{1}} = \begin{pmatrix} -3 \\ 1 \end{pmatrix}$

$\lambda_{2} = \frac{15}{2}$

$Q^{\intercal} = \begin{pmatrix} -\frac{3}{\sqrt{10}} & \frac{1}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} & \frac{3}{\sqrt{10}} \end{pmatrix}$

$\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} = \begin{pmatrix} -\frac{3}{\sqrt{10}} & \frac{1}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} & \frac{3}{\sqrt{10}} \end{pmatrix}\cdot\begin{pmatrix} \xi \\ \eta \end{pmatrix}$

Найдем математические ожидания и другие числовые характеристики и запишем плотность полученнного распределния

$\mathbb{E}\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} = \begin{pmatrix} -\frac{3}{\sqrt{10}} & \frac{1}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} & \frac{3}{\sqrt{10}} \end{pmatrix}\cdot\begin{pmatrix} -3 \\ 2 \end{pmatrix} = \begin{pmatrix} \frac{11\sqrt{10}}{10} \\ \frac{3\sqrt{10}}{10} \end{pmatrix}$

$\Sigma_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}} = Q*\Sigma_{\begin{pmatrix} \xi \\ \eta \end{pmatrix}}\cdot Q^{\intercal} = \frac{4}{75}\cdot\begin{pmatrix} -\frac{3}{\sqrt{10}} & \frac{1}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} & \frac{3}{\sqrt{10}} \end{pmatrix}\cdot\begin{pmatrix} 7 & -\frac{3}{2} \\ -\frac{3}{2} & 3 \end{pmatrix}\cdot\begin{pmatrix} -\frac{3}{\sqrt{10}} & \frac{1}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} & \frac{3}{\sqrt{10}} \end{pmatrix} = \begin{pmatrix} -\frac{3\sqrt{10}}{25} & \frac{\sqrt{10}}{25} \\ \frac{\sqrt{10}}{75} & \frac{\sqrt{10}}{25} \end{pmatrix}\cdot\begin{pmatrix}-\frac{3}{\sqrt{10}} & \frac{1}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} & \frac{3}{\sqrt{10}} \end{pmatrix} = \begin{pmatrix} \frac{2}{5} & 0 \\ 0 & \frac{2}{15} \end{pmatrix} = \begin{pmatrix} \frac{1}{\lambda_{1}} & 0 \\ 0 & \frac{1}{\lambda_{2}} \end{pmatrix}$

$\Sigma_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}}^{-1} = \frac{1}{{\det\Sigma_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}}}}\cdot\Sigma_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}} = \frac{1}{\frac{4}{75}}\cdot\begin{pmatrix} \frac{2}{5} & 0 \\ 0 & \frac{2}{15} \end{pmatrix} = \begin{pmatrix} \frac{5}{2} & 0 \\ 0 & \frac{15}{2} \end{pmatrix}$

$\mathbb{D}_{\xi} = \frac{2}{5}; \mathbb{D}_{\eta} = \frac{2}{15}; \mathop{\mathrm{cov}}\nolimits(\xi, \eta) = 0 \Rightarrow \rho_{\xi,\eta} = 0$

$C = \frac{1}{2\pi\cdot\frac{2}{5\sqrt{3}}} = \frac{5\sqrt{3}}{4\pi}$

Плотность полученного распределения: $P_{\xi,\eta}(x,y)= \frac{5\sqrt{3}}{4\pi}\cdot exp\left\{ -\frac{1}{2}\cdot\left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} \frac{11\sqrt(10)}{10} \\ \frac{3\sqrt{10}}{10} \end{pmatrix}\right]^{\intercal}\cdot\begin{pmatrix} \frac{5}{2} & 0 \\ 0 & \frac{15}{2} \end{pmatrix} \cdot\left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} \frac{11\sqrt(10)}{10} \\ \frac{3\sqrt{10}}{10} \end{pmatrix} \right] \right\}$

```python
## №3. Построим график плотности полученного распределения.
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-10, 5, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

def normDist(x, y):

    mean = np.array([11*np.sqrt(10) / 10, 3*np.sqrt(10) / 10])
    const = (5 * np.sqrt(3)) / (4 * np.pi)
    cov =  np.array([[5/2, 0], [0, 15/2]])
    diff = np.stack((x, y), axis=-1) - mean
    power = -1/2 * np.einsum('...k,kl,...l->...', diff, cov, diff)
    return const * np.exp(power)

Z = normDist(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

### 4. Вычислить характеристики совместного распределения случайного вектора $(-5ξ + 2η, 4ξ + 3η)^{\intercal}$ и записать его плотность.

$A_{4} = \begin{pmatrix} -5 & 2 \\ 4 & 3 \end{pmatrix}\to \det(A) = -23 \neq 0$

$A_{4}^{-1} = -\frac{1}{23}\cdot\begin{pmatrix} 3 & -2 \\ -4 & -5 \end{pmatrix}$

$\mathbb{E}\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} -5 & 2 \\ 4 & 3 \end{pmatrix}\cdot\begin{pmatrix} -3 \\ 2 \end{pmatrix} = \begin{pmatrix} 19 \\ -6 \end{pmatrix}$

$\Sigma_{\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix}} = \frac{4}{75}\cdot\begin{pmatrix} -5 & 2 \\ 4 & 3 \end{pmatrix}\cdot\begin{pmatrix} 7 & -\frac{3}{2} \\ -\frac{3}{2} & 3 \end{pmatrix}\cdot\begin{pmatrix} -5 & 4 \\ 2 & 3 \end{pmatrix} = \begin{pmatrix} -\frac{152}{75} & \frac{18}{25} \\ \frac{94}{75} & \frac{4}{25} \end{pmatrix}\cdot\begin{pmatrix} -5 & 4 \\ 2 & 3 \end{pmatrix} = \begin{pmatrix} \frac{868}{75} & -\frac{446}{75} \\ -\frac{-446}{75} & \frac{412}{75} \end{pmatrix}$

Следовательно, из этой ковариационной матрицы мы получаем следующее:

$\mathbb{D}_{\xi} = \frac{868}{75}; \mathbb{D}_{\eta} = \frac{412}{75}; \mathop{\mathrm{cov}}\nolimits(\xi, \eta) = -\frac{446}{75}$

$\sigma_{\xi} = \frac{2\sqrt{651}}{15}; \sigma_{\eta} = \frac{2\sqrt{309}}{15}; \rho = \frac{\frac{-446}{75}}{\frac{2\sqrt{651}}{15}\cdot\frac{2\sqrt{309}}{15}} = -\frac{223}{2\sqrt{22351}}$

$\Sigma_{\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix}}^{-1} = \frac{1}{\frac{2116}{75}}\cdot\begin{pmatrix} \frac{868}{75} & -\frac{446}{75} \\ -\frac{-446}{75} & \frac{412}{75} \end{pmatrix} = \begin{pmatrix} \frac{217}{529} & -\frac{223}{1058} \\ -\frac{223}{1058} & \frac{103}{529} \end{pmatrix}$

$C = \frac{1}{2\pi\cdot\sqrt{\frac{2216}{75}}} = \frac{5\sqrt{1662}}{2216\pi}$

Плотность полученного распределения: $P_{\xi_{4},\eta_{4}}(x,y) = \frac{5\sqrt{1662}}{2216\pi}\cdot exp\left\{ -\frac{1}{2}\cdot\left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 19 \\ -6 \end{pmatrix}
 \right]^{\intercal}\cdot\begin{pmatrix} \frac{217}{529} & -\frac{223}{1058} \\ -\frac{223}{1058} & \frac{103}{529} \end{pmatrix} \cdot \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 19 \\ -6 \end{pmatrix} \right]\right\}$

```python
## №4. Построим график плотности полученного распределения.

x = np.linspace(-5, 30, 200)
y = np.linspace(-20, 4, 200)
X, Y = np.meshgrid(x, y)

def normDist(x, y):

    mean = np.array([19, -6])
    const = (5*np.sqrt(1662)) / (2216 * np.pi)
    cov =  np.array([[217/529, -223/1058], [-223/1058, 103/529]])
    diff = np.stack((x, y), axis=-1) - mean
    power = -1/2 * np.einsum('...k,kl,...l->...', diff, cov, diff)
    return const * np.exp(power)

Z = normDist(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

### 5.Найти условное распределение $\xi$  при условии $\eta$.Вычислить $\mathbb{E}(\xi|\eta)$  и $\mathbb{D}(\xi|\eta)$.

Возьмем данные из пункта 1. Поскольку для одномерного случая матрица $\Sigma_{\eta}$ состоит из $\mathbb{D}_{\eta}$, то получим $\Sigma_{\eta}=\mathbb{D}_{\eta}$

$P_{\xi,\eta}(x,y)=C*exp\left( -\frac{1}{2}(3x^{2}+3xy+7y^{2}+12x-19y+37) \right)$

$\mathbb{E}_{\eta} = 2;$

$\mathbb{D}_{\eta} = \frac{4}{25};$

$\Sigma_{\eta}=\frac{4}{25}\to \Sigma_{\eta}^{-1} = \frac{25}{4}$;Подставив в формулу  $P_{\eta}(y) = \frac{1}{\sqrt{2\pi}\cdot\sqrt{\det(\Sigma)}}\cdot exp^{-\frac{1}{2}\cdot\left( (y-\mathbb{E}_{\eta})\cdot\Sigma_{\eta}^{-1}\cdot(y-\mathbb{E}_{\eta}) \right)}$ все наши параметры, мы получим плотность распределения $P_{\eta}$:  $P_{\eta}(y) = \frac{5\sqrt{3}}{2\sqrt{2\pi}}\cdot exp^{-\frac{1}{2}\cdot(y-2)\cdot\frac{25}{4}\cdot(y-2)}$

Теперь мы можем найти условное распределение $\xi$ при условии $\eta$ воспользуемся формулой условного распределения $P_{(\xi|\eta=y)} = \frac{P_{\xi,\eta}(x,y)}{P_{\eta}(y)}$: $P(\xi|\eta=y) = \frac{\frac{5\sqrt{3}}{4\pi}\cdot exp^{ -\frac{1}{2}(3x^{2}+3xy+7y^{2}+12x-19y+37) }}{\frac{5\sqrt{3}}{2\sqrt{2\pi}}\cdot exp^{-\frac{1}{2}(\frac{25}{4}y^2-25y+25)}} = \frac{1}{\sqrt{2\pi}}\cdot exp^{-\frac{1}{2}\left( 3x^{2}+3xy+7y^{2}+12x-19y+37 -  \frac{25}{4}y^{2} - 25y - 25 \right)} = \frac{1}{\sqrt{2\pi}}\cdot exp^{-\frac{1}{2}\left( 3x^{2} + 3xy +\frac{3}{4}y^{2} + 12x + 6y +12 \right)} = \frac{1}{\sqrt{2\pi}}\cdot exp^{-\frac{1}{2}\Biggl( 3\left( x^{2} + xy + \frac{1}{4}y^{2} + 4x + 2y + 4\right) \Biggl)} = \frac{1}{\sqrt{2\pi}}\cdot exp^{\frac{1}{2}\Biggl(3\left( x + \frac{1}{2}y + 2 \right)^2\Biggl)}$

Мы получили условное распределение из многомерного нормального распределения, и оно также является многомерным распределением. Зная, что $P(\xi|\eta=y) = C\cdot exp^{-\frac{1}{2}\cdot\left( \frac{1}{\mathbb{D}_{(\xi|\eta)}}\cdot(x-\mathbb{E}_{(\xi|\eta)})^{2} \right)}$, найдем $\mathbb{E}(\xi|\eta=y)$ и $\mathbb{D}(\xi|\eta)$

$P(\xi|\eta=y) = \frac{1}{\sqrt{2\pi}}\cdot exp^{\frac{1}{2}\Biggl(3\left( x + \frac{1}{2}y + 2 \right)^2\Biggl)}
\Rightarrow \mathbb{E}(\xi|\eta=y) = \frac{1}{2}y + 2; \mathbb{D}(\xi|\eta) = \frac{1}{3}$
