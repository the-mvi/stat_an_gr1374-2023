1. Чтобы вычислить вектор математических ожиданий воспользуемся преобразовением Лагранжа, в результате которого мы получим запись следующего вида: $q(x,y) = a\cdot(x - \alpha) +2c\cdot(x - \alpha)\cdot(y - \beta) + b\cdot(y - \beta)$

Для нахождения дисперсий ковариации нам поможет Ковариационная матрица $\Sigma$: $\Sigma = \begin{pmatrix} \mathbb{D}\xi & \mathop{\mathrm{cov}}\nolimits(\xi, \eta) \\ \mathop{\mathrm{cov}}\nolimits(\eta, \xi) & \mathbb{D}\eta \end{pmatrix}$

Кроме таких числовых характеристик,как $\mathbb{E}\xi, \mathbb{E}\eta, \mathbb{D}\xi, \mathbb{D}\eta$ и $ \mathop{\mathrm{cov}}\nolimits(\xi, \eta)$ , мы можем найти коэффициент корреляции $\rho$ и СКО $\sigma_{\xi}, \sigma_{\eta}$.

$\sigma_{\xi} = \sqrt{\mathbb{D_{\xi}}}$   

$\rho_{\xi\eta} = \frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta) }{\sigma_{\xi}\cdot\sigma_{\eta}}$

$C = \frac{1}{2\pi\cdot\sqrt{\det(\Sigma)}}$

2. Новый случайный вектор  $\mathbf{Y}$ можно определить, применяя аффинное преобразование g:  $\mathbb{R}^{n} \to \mathbb{R}^{n}$ к случайному вектору  $\mathbf{X}$:

$\textbf{Y} = A\cdot\textbf{X} + B$, где A является матрицей коэффициентов при $\xi, \eta$ размера $n\cdot n$, матрица B - матрица свободных членов размера $n\cdot 1$.

Или можно представить в таком виде: $A\cdot\overrightarrow{\xi} + B\sim \mathcal{N(o,\mathbb{E})}$. Для достижения цели нормируем q(x,y) и проведем проверки: $\mathbb{E}\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = A\cdot\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} + B = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$ и $\Sigma_{st} = A\cdot\Sigma\cdot A^{\intercal} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$

3. Для ортогонального преобразования, переводящий соответствующий центрированный случайны вектор в вектор с независимыми компонентами нам нужно решить следующее уравнение $\lambda^{2} - tr(A)\cdot\lambda + \det(A) = 0$

Находим ковариационные числа $\left\{ \mathrm{\lambda}_{1},...,\mathrm{\lambda}_{k} \right\}$ и соответствующие им собственные вектора $\left\{ \mathrm{V}_{1},...,\mathrm{V}_{k} \right\}$ ковариционной матрицы $\Sigma$ и нормируем вектора, получая ортонормированный базис

Собственные векторы, отвечающие различным собственным значениям, ортогональны между собой. Сформировав матрицу Q из нормированных собственных векторов, остается перемножить матрицу на центрированный вектор, чтобы получить вектор с независимыми компонентами.

Поскольку мы получаем вектор с независимыми переменнами, то по свойству ковариации: Если $\xi,\eta$ - независимые переменные, то $\mathop{\mathrm{cov}}\nolimits(\xi, \eta) = 0$

4. При нахождении условного распределения $\xi$ при условии $\eta$ воспользуемся формулой условного распределения $P_{(\xi|\eta=y)} = \frac{P_{\xi,\eta}(x,y)}{P_{\eta}(y)}$.



Для подсчета $\mathbb{E}_{(\xi|\eta)}$ и $\mathbb{D}_{(\xi|\eta)}$ воспользуемся следующей формулой $P_{\eta}(y) = \frac{1}{\sqrt{2\pi}\cdot \sqrt{\det(\Sigma)}}\cdot exp^{-\frac{1}{2}\cdot\left( (y-\mathbb{E}_{\eta})\cdot\Sigma^{-1}\cdot(y-\mathbb{E}_{\eta}) \right)} = C\cdot exp^{-\frac{1}{2}\cdot\left( \frac{1}{\mathbb{D}_{(\xi|\eta)}}\cdot(x-\mathbb{E}_{(\xi|\eta)})^{2} \right)}$

### 1.Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $P_{\xi,\eta}$.

Произведем преобразование Лагранжа
$q(x,y) = 3x^{2} + 4xy + 6y^{2} - 10x - 16y + 13 = (3x^{2} + 4xy - 10x) + 6y^{2} - 16y + 13 = (3x^{2} + 2x(2y - 5)) + 6y^{2} - 16y + 13 = 3\left( x^2 + \frac{2x(2y-5)}{3}\right) + 6y^{2} -16y + 13 = 3\left( x^{2} + \frac{2x(2y-5)}{3} + \left( \frac{2y-5}{3} \right)^{2} \right) - 3\left( \frac{2y-5}{3}\right)^{2} + 6y^2 - 16y + 13 = 3\left( x + \frac{2}{3}y - \frac{5}{3}\right)^{2} - \frac{4y^{2} - 20y + 25}{3} + 6y^{2} - 16y +13 = 3\left(x + \frac{2}{3}y - \frac{5}{3} \right)^{2} + \frac{14y^{2} - 28y + 14}{3} = 3 \left( x + \frac{2}{3}y - \frac{5}{3}\right)^{2} + \frac{14}{3}\left( y - 1 \right)^{2} =3\left( (x-1) + \frac{2}{3}(y-1)\right)^{2} + \frac{14}{3}(y-1)^{2} = 3\left( (x-1)^{2} + 2\cdot\frac{2}{3}(x-1)(y-1) + \frac{4}{9}(y-1)^{2}\right) + \frac{14}{3}(y-1)^{2} = 3(x-1)^{2} + 2\cdot2\cdot(x-1)(y-1) +6\cdot(y-1)^{2}   $

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$ #нашли вектор математических ожиданий

$\Sigma^{-1} = \begin{pmatrix} 3 & 2 \\ 2 & 6 \end{pmatrix} \implies {\Sigma} = \begin{pmatrix} 6 & -2 \\ -2 & 3 \end{pmatrix}$



$\mathbb{D}\xi = \frac{6}{14} = \frac{3}{7}, \mathbb{D}\eta = \frac{3}{14}, \mathop{\mathrm{cov}}\nolimits(\xi, \eta) = \frac{-2}{14} = -\frac{1}{7}$



$\rho_{\xi,\eta} = \frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta) }{\sqrt{{D}\xi}\cdot\sqrt{{D}\eta}} = \frac{\frac{1}{7}}{\frac{\sqrt3}{\sqrt7}\cdot\frac{\sqrt3}{\sqrt14}} = -\frac{\sqrt{2}}{3}$

$C = \frac{1}{2\pi\cdot\sqrt{\det(\Sigma)}} = \frac{\sqrt{14}}{2\pi}$

$P_{\xi\eta}(x,y) = \frac{14}{2\pi}\cdot exp ⋅\left\{ -\frac{1}{2}\cdot\left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 1 \\ 1 \end{pmatrix}\right]^{\intercal} \cdot \begin{pmatrix} 3 & -2 \\ -2 & 6 \end{pmatrix}  \cdot \left[ \begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} 1 \\ 1 \end{pmatrix} \right] \right\}$

```python

## Продолжение выполнения №1. Построим график плотности распределения pξ,η.
import numpy as np
import matplotlib.pyplot as plt

def normal_distribution(x, y):
    constant = np.sqrt(14) / (2 * np.pi)
    mean = np.array([1, 1])
    covariance = np.array([[3, -2],
                           [-2, 6]])
    diff = np.stack((x, y), axis=-1) - mean
    power = -0.5 * np.einsum('...k,kl,...l->...', diff, covariance, diff)
    return constant * np.exp(power)

x = np.linspace(-10, 10, 200)
y = np.linspace(-10, 10, 200)
X, Y = np.meshgrid(x, y)

Z = normal_distribution(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Плотность распределения')

plt.show()
```

### 3.Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построить график плотности полученного распределения.

$\Sigma^{-1} = \begin{pmatrix} 3 & 2 \\ 2 & 6 \end{pmatrix}$

$\begin{vmatrix} 3-\lambda & 2\\ 2 & 6-\lambda \end{vmatrix} = 0;$
$\lambda^{2} - 9\lambda + 14 = 0$ ⇒ $\lambda_{1} = 2, \lambda_{7} =7$

Для $\lambda_{1}$

$\left[\begin{array}{@{} rr|r @{\,}} 1 & 2 & 0 \\ 2 &  4 &  0 \end{array}\right]\sim \left[\begin{array}{@{} rr|r @{\,}} 1 &  2 & 0 \\ 0 &  0 &  0 \end{array}\right] ⇒\overrightarrow{\mathrm{V}_{1}^{'}} = \frac{1}{\sqrt5}\begin{pmatrix} -2 \\ 1 \end{pmatrix}$


Для $\lambda_{2}$

$\left[\begin{array}{@{} rr|r @{\,}} -4 & 2 & 0 \\ 2 &  -1 &  0 \end{array}\right]\sim \left[\begin{array}{@{} rr|r @{\,}} 2 &  -1 & 0 \\ 0 &  0 &  0 \end{array}\right] ⇒\overrightarrow{\mathrm{V}_{2}^{'}} = \frac{1}{\sqrt5}\begin{pmatrix} 1 \\ 2 \end{pmatrix}$


$Q^{\intercal} = \begin{pmatrix} -\frac{2}{\sqrt{5}} & \frac{1}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} & \frac{2}{\sqrt{5}} \end{pmatrix} ⇒ Q = \begin{pmatrix} -\frac{2}{\sqrt{5}} & \frac{1}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} & \frac{2}{\sqrt{5}} \end{pmatrix} $   


$\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} = \frac{1}{\sqrt5} \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix}\cdot\begin{pmatrix} \xi \\ \eta \end{pmatrix}$

                                    B = 0

$\mathbb{E}\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} =A\cdot E\cdot\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} + B =  \frac{1}{\sqrt5}\cdot \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix}\cdot\begin{pmatrix} 1 \\ 1 \end{pmatrix} = \frac{1}{\sqrt 5} \begin{pmatrix} -1 \\ 3 \end{pmatrix}$

$\Sigma_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}} =  \frac{1}{14}\cdot \begin{pmatrix} 6 & -2 \\ -2 & 3 \end{pmatrix};$

$\Sigma_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}} = Q*\Sigma_{\begin{pmatrix} \xi \\ \eta \end{pmatrix}}\cdot Q^{\intercal} = \frac{1}{14} ⋅\frac{1}{5}\cdot\begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix}\cdot\begin{pmatrix} 6 & -2 \\ -2 & 3 \end{pmatrix}\cdot\begin{pmatrix} -2 & 1 \\ 1 & -2 \end{pmatrix} = \frac{1}{14}⋅\frac{1}{5}\cdot\begin{pmatrix} -14 & 7 \\ 2 & 4 \end{pmatrix}\cdot\begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix} = \begin{pmatrix} \frac{1}{2} & 0 \\ 0 & \frac{1}{7} \end{pmatrix} = \begin{pmatrix} \frac{1}{\lambda_{1}} & 0 \\ 0 & \frac{1}{\lambda_{2}} \end{pmatrix}$

$\Sigma_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}}^{-1} = \begin{pmatrix} 2 & 0 \\ 0 & 7 \end{pmatrix}$


$C = \frac{1}{2\pi\cdot\sqrt\frac{1}{14}} = \frac{\sqrt{14}}{2\pi}$

Получили плотность: $P_{\xi,\eta}(x,y)= \frac{\sqrt{14}}{2\pi}\cdot exp\left\{ -\frac{1}{2}\cdot\left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -\frac{1}{\sqrt{5}} \\ \frac{3}{\sqrt{5}} \end{pmatrix}\right]^{\intercal}\cdot\begin{pmatrix} 2 & 0 \\ 0 & 7 \end{pmatrix} \cdot\left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -\frac{1}{\sqrt{5}} \\ \frac{3}{\sqrt{5}} \end{pmatrix} \right] \right\}$

```python
## №3. Построим график плотности полученного распределения.
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def normal_distribution(x, y):
    constant = np.sqrt(14) / (2 * np.pi)
    mean = 1 / np.sqrt(5) * np.array([-1, 3])
    covariance = np.array([[2, 0],
                           [0, 7]])
    inv_covariance = np.linalg.inv(covariance)
    diff = np.stack((x, y), axis=-1) - mean
    power = -0.5 * np.einsum('...k,kl,...l->...', diff, inv_covariance, diff)
    return constant * np.exp(power)

x = np.linspace(-10, 10, 200)
y = np.linspace(-10, 10, 200)
X, Y = np.meshgrid(x, y)

Z = normal_distribution(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Плотность распределения')

plt.show()
```

### 4. Вычислить характеристики совместного распределения случайного вектора $(-5ξ - η, -2ξ - 5η)^{\intercal}$ и записать его плотность. Построить график плотности полученного распределения.

$A = \begin{pmatrix} -5 & -1 \\ -2 & -5 \end{pmatrix} ⇒ \det(A) = 27 \neq 0$


$\mathbb{E}\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} -5 & 1 \\ -2 & -5 \end{pmatrix}\cdot\begin{pmatrix} 1 \\ 1 \end{pmatrix} = \begin{pmatrix} -4 \\ -7 \end{pmatrix}$

$\Sigma_{\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix}} = \frac{1}{14}\cdot\begin{pmatrix} -5 & 1 \\ -2 & -5 \end{pmatrix}\cdot\begin{pmatrix} 6 & -2 \\ -2 & 3 \end{pmatrix}\cdot\begin{pmatrix} -5 & -2 \\ 1 & -5 \end{pmatrix} = \frac{1}{14}\cdot\begin{pmatrix} -32 & 13 \\ -2 & -11 \end{pmatrix}\cdot\begin{pmatrix} -5 & -2 \\ 1 & -5 \end{pmatrix} = \frac{1}{14}\cdot\begin{pmatrix} 173 & -1 \\ -1 & 59 \end{pmatrix}$

Отсюда следует:

$\mathbb{D}_{\xi} = \frac{173}{14}; \mathbb{D}_{\eta} = \frac{59}{14}; \mathop{\mathrm{cov}}\nolimits(\xi, \eta) = -\frac{1}{14}$

$\rho = \frac{-\frac{1}{14}}{{\sqrt\frac{173}{14} ⋅ \sqrt{\frac{59}{14}}}} = -\frac{1}{\sqrt{10207}}$

$detΣ^{-1}_{4} = \frac{173}{14} ⋅ \frac{59}{14} - \frac{1}{14}⋅\frac{1}{14} = \frac{729}{14}$

$\Sigma_{\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix}}^{-1} = \frac{14}{729}\cdot\frac{1}{14}\cdot\begin{pmatrix} 59 & 1 \\ 1 & 173 \end{pmatrix} = \frac{1}{729}\cdot\begin{pmatrix} 59 & 1 \\ 1 & 173 \end{pmatrix} = \begin{pmatrix} \frac{59}{729} & \frac{1}{729} \\ \frac{1}{729} & \frac{173}{729} \end{pmatrix}$

$C = \frac{1}{2\pi\cdot\sqrt{\frac{729}{14}}} = \frac{\sqrt{14}}{54\pi}$

Плотность полученного распределения: $P_{\xi_{4},\eta_{4}}(x,y) = \frac{\sqrt{14}}{54\pi}\cdot exp\left\{ -\frac{1}{2}\cdot\left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -4 \\ -7 \end{pmatrix}
 \right]^{\intercal}\cdot\begin{pmatrix} \frac{59}{729} & \frac{1}{729} \\ \frac{1}{729} & \frac{173}{729} \end{pmatrix} \cdot \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -4 \\ -7 \end{pmatrix} \right]\right\}$

```python
## №4. Построим график плотности полученного распределения.
import numpy as np
import matplotlib.pyplot as plt

def normal_distribution(x, y):
    constant = np.sqrt(14) / (54 * np.pi)
    mean = np.array([-4, -7])
    covariance = np.array([[59/729, 1/729],
                           [1/729, 173/729]])
    inv_covariance = np.linalg.inv(covariance)
    diff = np.stack((x, y), axis=-1) - mean
    power = -0.5 * np.einsum('...k,kl,...l->...', diff, inv_covariance, diff)
    return constant * np.exp(power)

x = np.linspace(-15, 5, 200)
y = np.linspace(-20, 4, 200)
X, Y = np.meshgrid(x, y)

Z = normal_distribution(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Плотность распределения')

plt.show()
```
