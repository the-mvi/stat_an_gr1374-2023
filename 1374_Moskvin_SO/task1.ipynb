## Основные теоретические положения
Случайный вектор $\overrightarrow{\xi} = (\xi_{1},...,\xi_{n})^{\intercal}:\Omega 	\rightarrow \mathbb{R}^n$ имеет многомерное нормальное распределение, если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = (\eta_{1},...,\eta_{m})^{\intercal}$, матрица A размерности $n \times m$ и вещественный вектор $\overrightarrow{b} = (b_{1},...,b_{n})^{\intercal}$ такие что $\overrightarrow{\xi} = A\overrightarrow{\eta}+\overrightarrow{b}.$  
Плотность многомерного невырожденного нормального распределения имеет вид: $p_{\overrightarrow{\xi}}(x_{1},...,x_{n}) = \frac{1}{(2\pi)^{\frac{n}{2}}\sqrt{det\Sigma}}\exp\left(-\frac{1}{2}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)^{\intercal}\Sigma^{-1}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)\right)$, где $\overrightarrow{\mu}=\mathbb{E}\overrightarrow{\xi} = (\mathbb{E}\xi_{1},...,\mathbb{E}\xi_{n})^{\intercal} $ - вектор математических ожиданий компонент,$\Sigma = {\{\sigma_{i,j}=\mathop{\mathrm{cov}}\nolimits(\xi_{i}, \xi_{j})\}}^n_{i,j=1}$ - ковариационная матрица.
Свойства:
1. Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенной.
2. Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределенными. $p_{\xi|\eta = y}(x) = \frac{p_{\xi,\eta}(x,y)}{p_{\eta}(y)}$
3. Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму.
4. Многомерное нормальное распределение устойчиво относительно линейных преобразований. Если $\overrightarrow{\xi} \sim \mathcal{N}(\overrightarrow{\mu}, \Sigma)$, А произвольная матрица размерности $m \times n, \overrightarrow{b}$ произвольный вектор размерности $n \times 1$, то $ A\overrightarrow{\xi}+\overrightarrow{b} \sim \mathcal{N}(A\overrightarrow{\mu}, A\Sigma A^{\intercal})$

### 1.Вычислить вектор математических ожиданий и ковариационные характеристики случайного вектора. Построить график плотности распределения.  
$p_{\xi,\eta}(x,y) = C \cdot \exp\left(-\frac{1}{2}(4x^2-9xy+16y^2-15x-5y+25)\right)$
$q(x,y) = 4x^2-9xy+16y^2-15x-5y+25 = (4x^2-9xy-15x)+16y^2-5y+25 = 
4\left(x^2+x\frac{-9y-15}{4}\right)+16y^2-5y+25 = 4\left(x^2+2x\frac{-9y-15}{8}+\frac{(-9x-15)^2}{64}\right)+16y^2-5y+25-4\frac{(-9x-15)^2}{64} = 4\left(x+\frac{-9y-15}{8}\right)^2+\frac{1}{16}(256y^2-80y-400-81y^2-270y-225) = 4\left(x+\frac{-9y-15}{8}\right)^2+\frac{1}{16}(175y^2-350y+175) = 4\left(x-\frac{9y}{8}-\frac{15}{8}\right)^2+\frac{175}{16}(y-1)^2 = 4(x-3-\frac{9}{8}(y-1))^2+\frac{175}{16}(y-1)^2 = 4((x-3)^2-2\frac{9}{8}(x-3)(y-1)+\frac{81}{64}(y-1)^2)+\frac{175}{16}(y-1)^2 = 4(x-3)^2-9(x-3)(y-1)+16(y-1)^2$
$\mathbb{E}\begin{pmatrix}\xi \\ \eta\end{pmatrix} = \begin{pmatrix}3 \\ 1 \end{pmatrix}$  
$\Sigma^{-1} = \begin{pmatrix} 4 & -\frac{9}{2} \\ -\frac{9}{2} & 16\end{pmatrix}$  
$det\Sigma^{-1} = \frac{175}{4} $  
$\Sigma = \frac{2}{175}\begin{pmatrix} 32 & 9 \\ 9 & 8\end{pmatrix}$  
$\mathbb{D}\xi = \frac{64}{175}, \mathbb{D}\eta = \frac{16}{175}, \mathop{\mathrm{cov}}\nolimits(\xi, \eta) = \frac{18}{175}, r(\xi,\eta) = \frac{9}{16}$  
$det\Sigma = \frac{4}{175}, C = \frac{5\sqrt{7}}{4\pi}$  
$p_{\xi,\eta}(x,y) = \frac{5\sqrt{7}}{4\pi}\exp\left\{-\frac{1}{2}\begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 3 \\ 1 \end{pmatrix} \end{bmatrix}^{\intercal}\begin{pmatrix} 4 & -\frac{9}{2} \\ -\frac{9}{2} & 16\end{pmatrix}\begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 3 \\ 1 \end{pmatrix} \end{bmatrix}\right\}$

```python
import numpy as np
import matplotlib.pyplot as plt

def distribution_density(x, y, const):
    z = np.zeros(x.shape)
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            z[i,j] = const*np.exp((-1/2)*(4*x[i,j]**2-15*x[i,j]+25-9*x[i,j]*y[i,j]-5*y[i,j]+16*y[i,j]**2))
    return z

const = 5 * np.sqrt(7) / (4 * np.pi)
x = np.linspace(3-3,3+3, 100)
y = np.linspace(1-3,1+3, 100)
X, Y = np.meshgrid(x, y)
Z = distribution_density(X, Y, const)
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, cmap = "inferno")
plt.show()
```
