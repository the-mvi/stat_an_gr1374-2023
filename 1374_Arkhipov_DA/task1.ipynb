## Основные теоретические положения

Случайный вектор $\overrightarrow{\xi} = \left(\xi_1, \dots, \xi_n\right)^\intercal:\Omega\to \mathbb{R}^n$ имеет многомерное нормальное распределение, если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1, \dots, \eta_m\right)^\intercal$, матрица A размерности $n \times m$ и вещественный вектор $\overrightarrow{b} = (b_1,\dots,b_n)^\intercal$, такие что: $\overrightarrow{\xi} = \mathbf{A}\overrightarrow{\eta} + \overrightarrow{b}$.

Плотность многомерного невырожденного нормального распределения имеет вид: $\mathcal{p}_{\overrightarrow{\xi}}\left(x_1, \dots, x_n\right) = \frac{1}{(2\pi)^{n/2}\sqrt{\det \Sigma}} \exp\left(-\frac{1}{2}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)^\intercal \Sigma^{-1}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)\right)$  
где $\overrightarrow{\mu} = \mathbb{E}\overrightarrow{\xi} = \left(\mathbb{E}\xi_1, \dots, \mathbb{E}\xi_n\right)^\intercal$ - вектор математических ожиданий компонент, $\Sigma = \left\{\sigma_{i,j} = \mathop{\mathrm{cov}}\left(\xi_i,\xi_j\right)\right\}_{i,j=1}^n$ - ковариационная матрица.

Другая формула двумерного нормального распределения - $f(x,y) = \frac{1}{2\pi\cdot \sigma_{x}\cdot \sigma_{y}\cdot \sqrt{1 - \rho^{2}}}\cdot exp^{-\frac{1}{2\cdot(1-\rho^{2})}\cdot \left[ \frac{(x - \mathbb{E}_{x})^{2}}{\sigma_{x}^{2}} - \frac{2\rho\cdot(x-\mathbb{E}_{x})\cdot(y-\mathbb{E}_{y})}{\sigma_{x}\cdot \sigma_{y}} + \frac{(y-\mathbb{E}_{y})^{2}}{\sigma_{y}^{2}} \right]}$

Ковариационная матрица: $\Sigma = \begin{pmatrix} \mathbb{D}\xi & \mathop{\mathrm{cov}}\nolimits(\xi, \eta) \\ \mathop{\mathrm{cov}}\nolimits(\eta, \xi) & \mathbb{D}\eta \end{pmatrix}$

Формула плотности условного распределения $\xi$ при условии $\eta$: $p_{\xi|\eta=y}(x) = \frac{p_{\xi, \eta}(x, y)}{p_{\eta}(y)}$

## Выполнение работы


### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $p_{\xi, \eta}$

$q_{\xi, \eta} = 3x^2-3xy-12x+7y^2-19y+37 = 3 \left(x^2-2x \left(\frac y2 + 2 \right) + \left (\frac y2 + 2 \right)^2 \right) - \frac {3y^2}4 - 6y - 12 + 7y^2 - 19y + 37 = 3\left(x - \left(\frac y2 + 2 \right) \right)^2 + \frac{25}4 \left(y^2 - 4y + 4 \right) = 3 \left(x - \left(\frac y2 + 2 \right)\right)^2 + \frac{25}4 \left(y - 2 \right)^2 = 3\left((x-1) - \frac12 \left(y-2 \right)\right)^2 + \frac{25}4 \left(y - 2 \right)^2 = 3(x-3)^2 - 2\cdot\frac32 \left((x+1)(y-2) \right) + 7(y-2)^2 $  

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 3 \\ 2 \end{pmatrix}$  
$\Sigma = \frac4{75}\begin{pmatrix} 7 & \frac32 \\ \frac32 & 3 \end{pmatrix}$  

$\mathbb{D}\xi = \frac{28}{75}$  
$\mathbb{D}\eta = \frac4{25}$  
$\sigma_{\xi} = \frac{2\sqrt{7}}{5\sqrt{3}}$  
$\sigma_{\eta} = \frac25$  
$cov(\xi, \eta) = \frac32$  
$\rho(\xi, \eta) = \frac{cov(\xi, \eta)}{\sqrt{\mathbb{D}\xi\mathbb{D}\eta}} = \frac{75\sqrt{21}}{56}$  

$C = \frac1{2\pi\sqrt{\frac4{75}}} = \frac{5\sqrt3}{4\pi}$

$$ f_{\xi,\eta}(x, y) = \frac{5\sqrt3}{4\pi} exp\biggl[-\frac12 \begin{bmatrix} \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 3 \\ 2 \end{pmatrix} \end{bmatrix} ^{\intercal} \begin{pmatrix} 3 & -\frac32 \\ -\frac32 & 7 \end{pmatrix} \begin{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 3 \\ 2 \end{pmatrix} \end{pmatrix}\biggr]$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

mean = [3, 2]
cov = [[28/75, 6/75], [6/75, 12/75]]

x = np.linspace(1, 5, 100)
y = np.linspace(-1, 4, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

Z = multivariate_normal.pdf(pos, mean, cov)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Вероятность')

plt.show()
```

### 4. Вычислить характеристики совместного распределения случайного вектора $(4ξ - 4η, ξ + η)^{\intercal}$ и записать его плотность. Построить график полученной плотности.

$ A = \begin{pmatrix} 4 & -4 \\ 1 & 1 \end{pmatrix}$; $det(A) = 8 \neq 0 $  
$\mathbb{E}\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} 4 & -4 \\ 1 & 1 \end{pmatrix} \begin{pmatrix} 3 \\ 2 \end{pmatrix}  = \begin{pmatrix} 4 \\ 5 \end{pmatrix} $  
$\Sigma_{\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix}} = A \cdot \Sigma_{\begin{pmatrix} \xi \\ \eta \end{pmatrix}}\cdot A^{\intercal} = \frac{4}{75}\begin{pmatrix} 4 & -4 \\ 1 & 1 \end{pmatrix}\begin{pmatrix} 7 & \frac32 \\ \frac32 & 3\end{pmatrix} \begin{pmatrix} 4 & 1 \\ -4 & 1 \end{pmatrix} = \frac{4}{75}\begin{pmatrix} 22 & -6 \\ \frac{17}{2} & \frac92 \end{pmatrix} \begin{pmatrix} 4 & 1 \\ -4 & 1 \end{pmatrix} = \frac{4}{75} \begin{pmatrix} 112 & 16 \\ 16 & 13 \end{pmatrix}$

$\mathbb{D}_{\xi_4} = \frac{448}{75}$  
$\mathbb{D}_{\eta_4} = \frac{52}{75}$  
$\mathop{\mathrm{cov}}\nolimits(\xi_4, \eta_4) = \frac{64}{75}$  
$\rho(\xi_4, \eta_4) = \frac{cov(\xi_4, \eta_4)}{\sqrt{\mathbb{D}\xi_4\mathbb{D}\eta_4}} = \frac{4}{\sqrt{91}}$  
$\sigma_{\xi_4} = \frac{8\sqrt{7}}{5\sqrt{3}}$  
$\sigma_{\eta_4} = \frac{2\sqrt{13}}{5\sqrt{3}}$  

$ det\left(\Sigma_{\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix}}\right) = \frac{256}{75}$  
$C = \frac{1}{2\pi\sqrt{\frac{256}{75}}} = \frac{5\sqrt3}{32}$  
$ \Sigma_{\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix}}^{-1} = \frac{75}{256} \cdot \frac{4}{75} \begin{pmatrix} 112 & 16 \\ 16 & 13 \end{pmatrix} = \begin{pmatrix} \frac74 & \frac14 \\ \frac14 & \frac{13}{64} \end{pmatrix}$

$$ P_{\xi_{4},\eta_{4}}(x,y) = \frac{5\sqrt{3}}{32\pi} \cdot exp\left\{ -\frac{1}{2}\cdot\left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 4 \\ 5 \end{pmatrix}
 \right]^{\intercal}\cdot\begin{pmatrix} \frac{7}{4} & \frac{1}{4} \\ \frac{1}{4} & \frac{13}{64} \end{pmatrix} \cdot \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 4 \\ 5 \end{pmatrix} \right]\right\} $$ 

```python
mean = [4, 5]
cov = [[7/4, 1/4], [1/4, 13/64]]

x = np.linspace(-1, 10, 100)
y = np.linspace(0, 8, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

Z = multivariate_normal.pdf(pos, mean, cov)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Вероятность')

plt.show()
```
