## –û—Å–Ω–æ–≤–Ω—ã–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–ª–æ–∂–µ–Ω–∏—è

### –ú–µ—Ç–æ–¥ –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤

$RSS\left(\beta\right)=\left(Y-X^T\beta\right)^T\left(Y-X^T\beta\right)\underset\beta{\rightarrow\min}$
–°–∏—Å—Ç–µ–º–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π –≤ –º–∞—Ç—Ä–∏—á–Ω–æ–º –≤–∏–¥–µ: $XX^{\intercal}\beta=XY$
–†–µ—à–µ–Ω–∏–µ —Å–∏—Å—Ç–µ–º—ã –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π: $\widehat\beta=\left(XX^{\intercal}\right)^{-1}XY$

### –õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–∞—è –º–æ–¥–µ–ª—å

$–õ–∏–Ω–µ–π–Ω–∞—è\;—Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–∞—è\;–º–æ–¥–µ–ª—å:$<br>
${\mathbb{E}}_\theta\left(Y\vert z\right)=x\left(z\right)^T\beta$<br>
$–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è\;–º–æ–¥–µ–ª—å$<br>
${\mathbb{E}}_\theta\left(Y\vert z\right)=x^T\beta$<br>
$–ú–µ—Ç–æ–¥\;–Ω–∞–∏–º–µ–Ω—å—à–∏—Ö\;–∫–≤–∞–¥—Ä–∞—Ç–æ–≤:$<br>
$RSS\left(\beta\right)=\left|\left|Y-X^T\beta\right|\right|=\left(Y-X^T\beta\right)^T\left(Y-X^T\beta\right)\rightarrow min$<br>
$XX^T\beta=XY$<br>
$–†–µ—à–µ–Ω–∏–µ\;—Å–∏—Å—Ç–µ–º—ã:$<br>
$\widehat B=\left(XX^T\right)^{-1}XY$<br>
### –ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏:

$s^2=\frac{ss_e}{n-r}=\frac{RSS\left(\widehat\beta\right)}{n-r}=\frac{\left(Y-X^T\widehat\beta\right)^T\left(Y-X^T\widehat\beta\right)}{n-r}$ - –Ω–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ä–∞ $\sigma^2$
### –ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑—ã —Å–æ–≥–ª–∞—Å–∏—è:

$\mathbb{F} - $ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø—Ä–∏ –æ—Å–Ω–æ–≤–Ω–æ–π –≥–∏–ø–æ—Ç–µ–∑–µ –∏–º–µ–µ—Ç —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –§–∏—à–µ—Ä–∞-–°–Ω–µ–¥–µ–∫–æ—Ä–∞ $\mathcal{F}_{q, n - r}$<br>
–ö—Ä–∏—Ç–∏—á–µ—Å–∫–∞—è –æ–±–ª–∞—Å—Ç—å: $x_\alpha:\mathcal{F}_{q, n - r}(x_\alpha) = 1 - \alpha $<br>

### –î–ù–û —Ñ—É–Ω–∫—Ü–∏—è

$\psi\left(\beta\right)=C^{\intercal}\beta$ - –î–ù–û —Ñ—É–Ω–∫—Ü–∏—è, –¥–æ–ø—É—Å–∫–∞–µ—Ç –Ω–µ—Å–º–µ—â–µ–Ω–Ω–æ–µ –æ—Ü–µ–Ω–∏–≤–∞–Ω–∏–µ –µ—Å–ª–∏ $\exists L\left(Y\right)=Y\beta\;\forall\beta:\;\mathbb{E}_\beta L\left(Y\right)=C^{\intercal}\beta$<br>
$\widehat\psi\left(\beta\right)=C^\intercal\widehat\beta$ - –ú–ù–ö –æ—Ü–µ–Ω–∫–∞<br>
$\widehat\psi\sim N\left(\psi,\sigma^2B\right)$ –≥–¥–µ $B=C^{\intercal}\left(XX^{\intercal}\right)^{-1}C$<br>

### –î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π —ç–ª–ª–∏–ø—Å–æ–∏–¥ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $\psi$

$\frac{(\hat\psi - \psi)^{\intercal} B^{-1}(\hat\psi - \psi)}{qs^2} \sim \mathcal{F}_{q,n-r}$ <br>
–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π —ç–ª–ª–∏–ø—Å–æ–∏–¥ —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1-\alpha$ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $\psi$ <br>
$A_{\alpha}=\{\vec{x}: (\vec{x}-\hat\psi)^{\intercal} B^{-1} (\vec{x}-\hat\psi) \leq qs^2x_{\alpha} \}$, –≥–¥–µ <br>
$x_{\alpha}: \mathcal{F}_{q, n-r}(x_{\alpha}) = 1 - \alpha$ <br>

### –ö—Ä–∏—Ç–µ—Ä–∏–π –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞:
$–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞\;–ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞:$<br>
$D_n=sup_{x\in\mathbb{R}}\left|F_n\left(x\right)-F_0\left(x\right)\right|$<br>

$P_{F_0}\left(\sqrt nD_n<x\right)\rightarrow K\left(x\right),\mathbb{R}\left(x\right)$

–ö—Ä–∏—Ç–µ—Ä–∏–π:
$\widehat{\varphi\left(\overrightarrow x\right)}=\left\{\begin{array}{l}0,\widehat{D_n}<x_\alpha\\1,\widehat{D_n}\geq x_\alpha\end{array}\right.$<br>

### –¢–µ–æ—Ä–µ–º–∞ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞
$\sqrt nD_n\rightarrow K$
$–î–ù–û\;—Ñ—É–Ω–∫—Ü–∏—è\;\psi=C^T\beta,\;–≥–¥–µ\;C=\left(C_1,C_2..C_m\right)^T$

–§–æ—Ä–º—É–ª–∞ F —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏
$\;F=\frac{\widehat\psi^T\left(\left(C^T\left(XX^T\right)^{-1}C\right)^{-1}\right)\widehat\psi}{qS^2}\sim F_{q,n-r}$
$C^T=\left(0‚Äá1\right)$
$\psi=\begin{pmatrix}0&1\end{pmatrix}\times\begin{pmatrix}\beta_0\\\beta_1\end{pmatrix}=\beta_1$

–ö–≤–∞–Ω—Ç–∏–ª—å
$\operatorname{ùë•}_\alpha:F_{q,n-r}\left(\operatorname{ùë•}_\alpha\right)=1-\alpha$
$–û—Ü–µ–Ω–∫–∞\;–ø–∞—Ä–∞–º–µ—Ç—Ä–∞\;S^2:$
$S^2=\frac{RSS\left(\widehat\beta\right)}{\left(n-r\right)}$

$S^2=\frac1{\left(n-r\right)}\left(Y-X^T\beta\right)^T\left(Y-X^T\beta\right)$

 –æ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X. –ü–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ–¥–µ—Ä–∂–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å.

## –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã

### 1.–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞. –°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –ª–∏–Ω–µ–π–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω—É—é –º–æ–¥–µ–ª—å –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –ø–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –º–µ—Ç–æ–¥ –Ω–∞–∏–º–µ–Ω—å—à–µ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–∞ –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —Å–¥–≤–∏–≥–∞ $\beta_0$ –∏ –º–∞—Å—à—Ç–∞–±–∞ $\beta_1$.  –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—É—é –ª–∏–Ω–∏—é —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –û—Ü–µ–Ω–∏—Ç—å –≤–∏–∑—É–∞–ª—å–Ω–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–∏


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


x = np.array([6, 4, 2, 1, 4, 2, 6, 2, 5, 0, 3, 5, 3, 1, 0, 4, 5, 6, 2, 4, 6, 5, 3, 3, 4, 5, 0, 5, 2, 1, 5, 6, 6, 3, 3, 1, 2, 1, 5, 6, 0, 5, 4, 3, 2, 4, 6, 2, 3, 5])
y = np.array([16.91, 14.48, 11.30, 17.22, 12.27, 17.99, 4.67, 2.92, 9.78, 5.44, 18.80, 11.81, 8.36, 10.58, 11.42, 15.32, 5.94, 9.03, 13.87, 16.20, 9.13, 19.68, 4.73, 19.47, 24.51, 10.03, 0.89, 14.45, 11.16, 14.04, 16.82, 2.63, 12.59, 11.68, 17.44, 15.01,13.31, 16.14, 8.32, 16.45, 15.36, 13.92, 7.63, 7.68, 16.49, 8.39, 6.08, 8.69, 14.85, 11.73])
alpha, h = 0.20, 2.20


plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('–ì—Ä–∞—Ñ–∏—á–µ—Å–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ Y –æ—Ç –•')
plt.show()
```


–†–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–∞—è –º–æ–¥–µ–ª—å: $Y=X^{\intercal}\beta+\varepsilon$;
$\varepsilon$ - –≤–µ–∫—Ç–æ—Ä –æ—à–∏–±–æ–∫, $\mathbb{E}_\theta\varepsilon=0\ ;\mathbb{D}_\theta\varepsilon=\sigma^2E$<br>
–õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–∞—è –º–æ–¥–µ–ª—å –ø—Ä–∏ $\beta=\begin{pmatrix}\beta_0\\\beta_1\end{pmatrix}$: $y_i=x_i\beta_1+\beta_0+\varepsilon_i$ i=1...n<br>

$Y = \begin{pmatrix}1&x_1\\1&x_2\\...&...\\1&x_{49}\\1&x_{50}\end{pmatrix}\begin{pmatrix}\beta_0\\\beta_1\end{pmatrix}$

<br>

$X = \begin{pmatrix}1&1&...&1&1\\x_1&x_2&...&x_{49}&x_{50}\end{pmatrix}$

–†–µ—à–µ–Ω–∏–µ —Å–∏—Å—Ç–µ–º—ã –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π: $\widehat\beta=\left(XX^{\intercal}\right)^{-1}XY$
–ú–ù–ö –æ—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $\widehat\beta=\begin{pmatrix}{\widehat\beta}_0\\{\widehat\beta}_1\end{pmatrix}=\begin{pmatrix}12.432\\-0.105\end{pmatrix}$

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
beta_1 = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean) ** 2)
beta_0 = y_mean - beta_1 * x_mean

print('–°–¥–≤–∏–≥–∞ beta_0 = ',round(beta_0,3),', –º–∞—Å—à—Ç–∞–±–∞ beta_1 = ',round(beta_1,3))
```

```python
y_pred = beta_0 + beta_1 * x
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('–õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è Y –æ—Ç X')
plt.legend()
plt.grid(True)
plt.show()
```

### 2.–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –∏ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å –Ω–µ—Å–º–µ—â—ë–Ω–Ω—É—é –æ—Ü–µ–Ω–∫—É –¥–∏—Å–ø–µ—Ä—Å–∏–∏. –ù–∞ –±–∞–∑–µ –æ—à–∏–±–æ–∫ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —Å —à–∞–≥–æ–º h. –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –Ω–∞ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –ø–æ $\chi^{2}$. –û—Ü–µ–Ω–∏—Ç—å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–∏ –¥–æ –∫–ª–∞—Å—Å–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –ø–æ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤—É. –í–∏–∑—É–∞–ª—å–Ω–æ –æ—Ü–µ–Ω–∏—Ç—å –¥–∞–Ω–Ω—ã–π —Ñ–∞–∫—Ç

```python
from scipy.stats import chisquare, kstest, norm, chi2, stats, t

n = len(y)

variance_estimation= np.sum((y - y_mean)**2) / (n - 1)
print("–ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏:", variance_estimation)

epsilon = y - y_pred #–æ—à–∏–±–∫–∏
# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –æ—à–∏–±–æ–∫ —Å –∑–∞–¥–∞–Ω–Ω—ã–º —à–∞–≥–æ–º h
plt.hist(epsilon, bins=np.arange(min(epsilon), max(epsilon) + h, h), edgecolor='black')
plt.xlabel('…õ')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—à–∏–±–æ–∫')
plt.grid(True)
plt.show()
```

```python
s_2 = np.var(epsilon, ddof=1)
n=len(x)
se_beta_0 = np.sqrt(s_2 * ((1/n) + (x_mean**2 / ((n - 1) * np.sum((x - x_mean)**2)))))
se_beta_1 = np.sqrt(s_2 / (np.sum((x - x_mean)**2)))
t_critical = -np.abs(t.ppf(alpha/2, n - 2))
# –í—ã—á–∏—Å–ª—è–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞
CI_beta_0 = (beta_0 - t_critical * se_beta_0, beta_0 + t_critical * se_beta_0)
CI_beta_1 = (beta_1 - t_critical * se_beta_1, beta_1 + t_critical * se_beta_1)

print("–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è beta_0:", CI_beta_0)
print("–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è beta_1:", CI_beta_1)

# –í—ã—á–∏—Å–ª—è–µ–º –ø–æ–ª—É–æ—Å–∏ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–∞
semiaxis_1 = t_critical * se_beta_0
semiaxis_2 = t_critical * se_beta_1

print("–ü–æ–ª—É–æ—Å—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–∞:", semiaxis_1)
print("–ü–æ–ª—É–æ—Å—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–∞:",semiaxis_2)
```

### 5.–°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –º–æ–¥–µ–ª—å –≤–∫–ª—é—á–∞—é—â—É—é –¥–æ–ø–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π —á–ª–µ–Ω —Å $X^2$.–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –ú–ù–ö –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\beta_0,\beta_1,\beta_2$ –≤ –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏. –ò–∑–æ–±—Ä–∞–∑–∏—Ç—å –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ –ø–æ–ª—É—á–µ–Ω–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω—É—é –º–æ–¥–µ–ª—å.
$Y = \beta_0 + \beta_1X + \beta_2X^2 + \varepsilon,$
$–ù–æ—Ä–º–∞–ª—å–Ω—ã–µ\;—É—Ä–∞–≤–Ω–µ–Ω–∏—è:$<br>

$XX^T\beta=XY$

$–†–µ—à–µ–Ω–∏–µ\;—Å–∏—Å—Ç–µ–º—ã:$

$\widehat B=\left(XX^T\right)^{-1}XY$

```python
x = np.array([6, 4, 2, 1, 4, 2, 6, 2, 5, 0, 3, 5, 3, 1, 0, 4, 5, 6, 2, 4, 6, 5, 3, 3, 4, 5, 0, 5, 2, 1, 5, 6, 6, 3, 3, 1, 2, 1, 5, 6, 0, 5, 4, 3, 2, 4, 6, 2, 3, 5])
y = np.array([16.91, 14.48, 11.30, 17.22, 12.27, 17.99, 4.67, 2.92, 9.78, 5.44, 18.80, 11.81, 8.36, 10.58, 11.42, 15.32, 5.94, 9.03, 13.87, 16.20, 9.13, 19.68, 4.73, 19.47, 24.51, 10.03, 0.89, 14.45, 11.16, 14.04, 16.82, 2.63, 12.59, 11.68, 17.44, 15.01,13.31, 16.14, 8.32, 16.45, 15.36, 13.92, 7.63, 7.68, 16.49, 8.39, 6.08, 8.69, 14.85, 11.73])

# –°–æ–∑–¥–∞–Ω–∏–µ –Ω–æ–≤–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ —Å X^2
x_squared = x**2

# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ X, X^2 –∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ —Å –µ–¥–∏–Ω–∏—Ü–∞–º–∏
x_extended = np.column_stack((x, x_squared))
x_extended = np.column_stack([np.ones(len(x)), x_extended])

# –ú–ù–ö –æ—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≤ —Ä–∞—Å—à–∏—Ä–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏
Beta_extended = np.linalg.inv(x_extended.T @ x_extended) @ x_extended.T @ y

# –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤
beta_0, beta_1, beta_2 = Beta_extended
print(f"–û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ beta_0: {beta_0}")
print(f"–û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ beta_1: {beta_1}")
print(f"–û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ beta_2: {beta_2}")

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏
x_plot = np.linspace(min(x), max(x), 100)
x_plot_extended = np.column_stack((np.ones(100), x_plot, x_plot**2))
y_plot_extended = x_plot_extended @ Beta_extended

# –ì—Ä–∞—Ñ–∏—á–µ—Å–∫–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ
plt.scatter(x, y, label='–î–∞–Ω–Ω—ã–µ')
plt.plot(x_plot, y_plot_extended, color='red', label='–†–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–∞—è –º–æ–¥–µ–ª—å (—Å $X^2$)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('–†–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–∞—è –º–æ–¥–µ–ª—å —Å $X^2$')
plt.grid(True)
plt.show()
```

### 7.–í –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\beta_0,\beta_1,\beta_2$ —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1-\alpha$. –ù–∞–ø–∏—Å–∞—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–æ–∏–¥–∞ —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1-\alpha$.

$–î–ù–û\;—Ñ—É–Ω–∫—Ü–∏—è:$
$\psi=–°^–¢\beta,C=\left(C_1,...C_n\right)^T$

$–û—Ü–µ–Ω–∫–∞:$
$\widehat\psi=C^T\widehat\beta$
$–≠–ª–µ–º–µ–Ω—Ç\;–¥–∏—Å–ø–µ—Ä—Å–∏–∏\;–æ—Ü–µ–Ω–∫–∏:$
$\;b_\psi=C^T\left(XX^T\right)^{-1}C$

$–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π\;–∏–Ω—Ç–µ—Ä–≤–∞–ª\;—É—Ä–æ–≤–Ω—è\;–¥–æ–≤–µ—Ä–∏—è\;1-\alpha:$
$\left[\widehat\psi-x_\alpha s\sqrt b;\widehat\psi+x_\alpha s\sqrt b\right]\ni\psi$
$\chi_\alpha:S_{n-r}\left(x_\alpha\right)=1-\frac\alpha2$
$–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π\;–∏–Ω—Ç–µ—Ä–≤–∞–ª\;—ç–ª–ª–∏–ø—Å–∞\;—É—Ä–æ–≤–Ω—è\;–¥–æ–≤–µ—Ä–∏—è\;1-\alpha$
$A_\alpha=\left\{\overrightarrow x:\left(\overrightarrow x-\widehat\psi\right)^–¢b^{-1}\left(\overrightarrow x-\widehat\psi\right)\leq qS^2x_\alpha\right\}$

```python
cov_matrix = s_2 * np.linalg.inv(x_extended.T.dot(x_extended))
print(cov_matrix)
se_beta_0 = np.sqrt(cov_matrix[0, 0])
se_beta_1 = np.sqrt(cov_matrix[1, 1])
se_beta_2 = np.sqrt(cov_matrix[2, 2])
critical_value = norm.ppf(1 - alpha / 2)
std_eps = np.sqrt(np.diag(cov_matrix))

confidence_intervals0 = np.column_stack((se_beta_0 - critical_value * std_eps, se_beta_0 + critical_value * std_eps))
confidence_intervals1 = np.column_stack((se_beta_1 - critical_value * std_eps, se_beta_1 + critical_value * std_eps))
confidence_intervals2 = np.column_stack((se_beta_2 - critical_value * std_eps, se_beta_2 + critical_value * std_eps))
print("–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã:")
print('Œ≤0:', confidence_intervals0)
print('Œ≤1:',confidence_intervals1 )
print('Œ≤2:',confidence_intervals2 )
```

### 8.–°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –æ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X –∏ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –µ—ë –∑–Ω–∞—á–∏–º–æ—Å—Ç—å –Ω–∞ —É—Ä–æ–≤–Ω–µ $\alpha$.


$–î–ù–û\;—Ñ—É–Ω–∫—Ü–∏—è\;\psi=C^T\beta,\;–≥–¥–µ\;C=\left(C_1,C_2..C_m\right)^T$<br>
$–§–æ—Ä–º—É–ª–∞\;F\;—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏$
$\;F=\frac{\widehat\psi^T\left(\left(C^T\left(XX^T\right)^{-1}C\right)^{-1}\right)\widehat\psi}{qS^2}\sim F_{q,n-r}$
$C^T=\begin{pmatrix}0&0&1\end{pmatrix}$
$\psi=\begin{pmatrix}0&0&1\end{pmatrix}\times\begin{pmatrix}\beta_0\\\beta_1\\\beta_2\end{pmatrix}=\beta_2$

–ö–≤–∞–Ω—Ç–∏–ª—å
$\operatorname{ùë•}_\alpha:F_{q,n-r}\left(\operatorname{ùë•}_\alpha\right)=1-\alpha$

```python
SSE = np.sum((y - (y_pred)) ** 2)
SSR = np.sum((y_pred - np.mean(y)) ** 2)
#SST = SSE + SSR
p=2
MSE = SSE / (n - p-1)  # –°—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤
MSR = SSR / p  # –°—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ–±—ä—è—Å–Ω–µ–Ω–Ω—ã—Ö –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π
n=len(y)
F_stat = MSR / MSE
F_critical = f.ppf(1 - alpha, p, n - p)
if F_stat > F_critical:
    print("–ü—Ä–∏–Ω–∏–º–∞–µ–º –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –≥–∏–ø–æ—Ç–µ–∑—É")


print("F-–∫—Ä–∏—Ç–µ—Ä–∏–π:", F_stat)
print("–ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ F:", F_critical)
```
