## Основные теоретические положения

Двумерное нормальное распределение вероятностей:

$f(x,y) = \frac{1}{2\pi\cdot \sigma_{x}\cdot \sigma_{y}\cdot \sqrt{1 - \rho^{2}}}\cdot exp^{-\frac{1}{2\cdot(1-\rho^{2})}\cdot \left[ \frac{(x - \mathbb{E}_{x})^{2}}{\sigma_{x}^{2}} - \frac{2\rho\cdot(x-\mathbb{E}_{x})\cdot(y-\mathbb{E}_{y})}{\sigma_{x}\cdot \sigma_{y}} + \frac{(y-\mathbb{E}_{y})^{2}}{\sigma_{y}^{2}} \right]}$

$\rho$ - коэффициент коррелции случайно величины X и Y

$\sigma_{x}$ - среднее квадратическое отклонение случайной величины X

$\sigma_{y}$ - среднее квадратическое отклонение случайной величины Y

$\mathbb{E}x$ - математическое ожидание случайной величины X

$\mathbb{E}y$ - математическое ожидание случайной величины Y

1. Чтобы вычислить вектор математических ожиданий воспользуемся преобразовением Лагранжа, в результате которого мы получим запись следующего вида: $q(x,y) = a\cdot(x - \alpha) +2c\cdot(x - \alpha)\cdot(y - \beta) + b\cdot(y - \beta)$

Для нахождения дисперсий ковариации нам поможет Ковариационная матрица $\Sigma$: $\Sigma = \begin{pmatrix} \mathbb{D}\xi & \mathop{\mathrm{cov}}\nolimits(\xi, \eta) \\ \mathop{\mathrm{cov}}\nolimits(\eta, \xi) & \mathbb{D}\eta \end{pmatrix}$

Кроме таких числовых характеристик,как $\mathbb{E}\xi, \mathbb{E}\eta, \mathbb{D}\xi, \mathbb{D}\eta$ и $ \mathop{\mathrm{cov}}\nolimits(\xi, \eta)$ , мы можем найти коэффициент корреляции $\rho$ и СКО $\sigma_{\xi}, \sigma_{\eta}$.

$\sigma_{\xi} = \sqrt{\mathbb{D_{\xi}}}$   

$\rho_{\xi\eta} = \frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta) }{\sigma_{\xi}\cdot\sigma_{\eta}}$

$C = \frac{1}{2\pi\cdot\sqrt{\det(\Sigma)}}$

2. Новый случайный вектор  $\mathbf{Y}$ можно определить, применяя аффинное преобразование g:  $\mathbb{R}^{n} \to \mathbb{R}^{n}$ к случайному вектору  $\mathbf{X}$:

$\textbf{Y} = A\cdot\textbf{X} + B$, где A является матрицей коэффициентов при $\xi, \eta$ размера $n\cdot n$, матрица B - матрица свободных членов размера $n\cdot 1$.

Или можно представить в таком виде: $A\cdot\overrightarrow{\xi} + B\sim \mathcal{N(o,\mathbb{E})}$.

3. Для ортогонального преобразования, переводящий соответствующий центрированный случайны вектор в вектор с независимыми компонентами нам нужно решить следующее уравнение $\lambda^{2} - tr(A)\cdot\lambda + \det(A) = 0$

Находим ковариационные числа $\left\{ \mathrm{\lambda}_{1},...,\mathrm{\lambda}_{k} \right\}$ и соответствующие им собственные вектора $\left\{ \mathrm{V}_{1},...,\mathrm{V}_{k} \right\}$ ковариционной матрицы $\Sigma$ и нормируем вектора, получая ортонормированный базис.

Собственные векторы, отвечающие различным собственным значениям, ортогональны между собой. Сформировав матрицу Q из нормированных собственных векторов, остается перемножить матрицу на центрированный вектор, чтобы получить вектор с независимыми компонентами.


4. При нахождении условного распределения $\xi$ при условии $\eta$ воспользуемся формулой условного распределения $P_{(\xi|\eta=y)} = \frac{P_{\xi,\eta}(x,y)}{P_{\eta}(y)}$.

Условное распределение полученные из многомерного нормального распределения, также является многомерным распределением. Для одномерного нормального распределения ковариционная матрица $\Sigma$ будет состоять только из дисперсии.

Для подсчета $\mathbb{E}_{(\xi|\eta)}$ и $\mathbb{D}_{(\xi|\eta)}$ воспользуемся следующей формулой $P_{\eta}(y) = \frac{1}{\sqrt{2\pi}\cdot \sqrt{\det(\Sigma)}}\cdot exp^{-\frac{1}{2}\cdot\left( (y-\mathbb{E}_{\eta})\cdot\Sigma^{-1}\cdot(y-\mathbb{E}_{\eta}) \right)} = C\cdot exp^{-\frac{1}{2}\cdot\left( \frac{1}{\mathbb{D}_{(\xi|\eta)}}\cdot(x-\mathbb{E}_{(\xi|\eta)})^{2} \right)}$

Свойства многомерного нормального распределения:
- Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенной
- Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределенными. $\mathcal{p}_{\xi|\eta=y}(x) = \frac{\mathcal{p}_{\xi,\eta}(x,y)}{\mathcal{p}_{\eta}(y)}$
- Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму
- Многомерное нормальное распределение устойчиво относительно линейных преобразований. Если $\overrightarrow{\xi} \sim \mathcal{N}\left(\overrightarrow{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, $\overrightarrow{b}$ - произвольный вектор размерности $n \times 1$, то $A\overrightarrow{\xi}+\overrightarrow{b} \sim \mathcal{N}\left(A\overrightarrow{\mu}+\overrightarrow{b}, A \Sigma A^\intercal\right)$.

### 1.Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора.

Произведем преобразование Лагранжа
$q(x,y) = 7x^{2}+8xy+13y^{2}-6x+18y+12 = (7x^{2} +8xy - 6x) + 13y^{2} +18y +12 = (7x^{2}+2x(4y-3)) + 13y^{2} +18y + 12 = 7\left( x^2 + \frac{2}{7}x(4y-3)+ (\frac{4y-3}{7})^{2} \right)-7\left( \frac{4y-3}{7} \right)^{2} +13y^{2} +18y + 12 = 7\left( x + \frac{4}{7}y-\frac{3}{7} \right)^{2} - \frac{4}{7}y^{2} + \frac{24}{7}y - \frac{21}{7} -\frac{4}{7}y^{2} +13y^{2}+18y + 12 = 7\left( x + \frac{4}{7}y -\frac{3}{7} \right)^{2} + \frac{75}{7}y^{2} + \frac{150}{7}y - \frac{75}{7} = 7\left( (x -1) +\frac{4}{7}(y+1) \right)^{2} + \frac{75}{7}\left( y+1 \right)^{2} = 7\left( (x-1)^{2} +2 \frac{4}{7}(y++1)(x-1)+\frac{16}{49}(y+1)^{2}) \right) + \frac{75}{7}\left( y + 1 \right)^{2} = 7\left( x - 1\right)^{2} + 8( y+ 1)(y-1)+ \frac{16}{7}\left( y + 1 \right)^{2} + \frac{75}{7}\left( y + 1 \right)^{2} = 7(x-1)^{2} +8(x - 1)(y + 1) + 13(y + 1)^{2}$

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 1 \\ -1 \end{pmatrix}$

$\Sigma^{-1} = \begin{pmatrix} 7 & 4 \\ 4 & 13 \end{pmatrix} \implies \det{\Sigma^{-1}} =75$

Остальные числовые характеристики данного случайного вектора.

$\Sigma = \begin{pmatrix} 13 & -4 \\ -4 & 7 \end{pmatrix} \cdot \frac{1}{\det{\Sigma^{-1}}} = \begin{pmatrix} \frac{13}{75} & -\frac{4}{25} \\ -\frac{4}{25} & \frac{7}{75}\end{pmatrix} \implies \mathbb{D}\xi = \frac{13}{75}, \mathbb{D}\eta = \frac{7}{75}, \mathop{\mathrm{cov}}\nolimits(\xi, \eta) = -\frac{4}{75}$



$\rho_{\xi,\eta} = \frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta) }{\sigma_{\xi}\cdot\sigma_{\eta}} = \frac{-\frac{4}{75}}{\frac{\sqrt{39}}{15}\cdot\frac{\sqrt{21}}{15}} = -\frac{4\sqrt{91}}{91}$

```python
##  №1. Построим график плотности распределения pξ,η.
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 5, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)


sigmavec = np.array([1, -1])
C = (5 * np.sqrt(3)) / (2 * np.pi)
sigmamat = np.array([[13, -4], [-4, 7]])
d = np.stack((X, Y), axis=-1) - sigmavec
power = -1/2 * np.einsum('...k,kl,...l->...', d, sigmamat, d)

Z = C * np.exp(power)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='magma')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```
