## Основные теоретические положения


Случайный вектор $\overrightarrow{\xi} = \left(\xi_1, \dots, \xi_n\right)^\intercal:\Omega\to \mathbb{R}^n$ имеет многомерное нормальное распределение, если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1, \dots, \eta_m\right)^\intercal$, матрица $\mathbf{A}$ размерности $n \times m$ и вещественный вектор $\overrightarrow{b} = (b_1,\dots,b_n)^\intercal$, такие что: $\overrightarrow{\xi} = \mathbf{A}\overrightarrow{\eta} + \overrightarrow{b}$.

Плотность многомерного невырожденного нормального распределения имеет вид: $\mathcal{p}_{\overrightarrow{\xi}}\left(x_1, \dots, x_n\right) = \frac{1}{(2\pi)^{n/2}\sqrt{\det \Sigma}} \exp\left(-\frac{1}{2}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)^\intercal \Sigma^{-1}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)\right)$, где $\overrightarrow{\mu} = \mathbb{E}\overrightarrow{\xi} = \left(\mathbb{E}\xi_1, \dots, \mathbb{E}\xi_n\right)^\intercal$ - вектор математических ожиданий компонент, $\Sigma = \left\{\sigma_{i,j} = \mathop{\mathrm{cov}}\left(\xi_i,\xi_j\right)\right\}_{i,j=1}^n$ - ковариационная матрица.

Свойства многомерного нормального распределения:
- Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенной
- Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределенными. $\mathcal{p}_{\xi|\eta=y}(x) = \frac{\mathcal{p}_{\xi,\eta}(x,y)}{\mathcal{p}_{\eta}(y)}$
- Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму
- Многомерное нормальное распределение устойчиво относительно линейных преобразований. Если $\overrightarrow{\xi} \sim \mathcal{N}\left(\overrightarrow{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, $\overrightarrow{b}$ - произвольный вектор размерности $n \times 1$, то $A\overrightarrow{\xi}+\overrightarrow{b} \sim \mathcal{N}\left(A\overrightarrow{\mu}+\overrightarrow{b}, A \Sigma A^\intercal\right)$.

### 1. Вычисление характеристик данного случайного вектора. Построение графика плотности распределения $\mathcal{p}_{\xi,\eta}$.

${}\mathcal{p}_{\xi,\eta}(x,y) = C \cdot \exp\left( -\frac{1}{2}\left(3x^2+3xy+7y^2+12x-19y+37\right)\right)$

Для нахождения характеристик было использовано преобразование квадратичной формы к каноническому виду методом Лагранжа.

$\begin{split}
  q(x,y) &= 3x^2+3xy+7y^2+12x-19y+37 = \\
  &= \left(3x^2 + 3xy+12x\right) + 7y^2-19y+37 = \\
  &= \left(3x^2+3x\left(y+4\right)\right)+7y^2-19y+37 = \\
  &= 3\left(x^2+2x\cdot\frac{y+4}{2}+\frac{\left(y^2+4\right)^2}{4}\right)+7y^2-19y+37-\frac{3(y+4)^2}{4} = \\
  &= 3\left(x+\frac{y+4}{2}\right)^2+7y^2-19y+37-\frac{3\left(y^2+8y+16\right)}{4} = \\
  &= 3\left(x+\frac{y}{2}+2\right)^2+\frac{1}{4}\left(28y^2-76y+148-3y^2-24y-48\right) = \\
  &= 3\left(x+\frac{y}{2}+2\right)^2+\frac{1}{4}\left(25y^2-100y+100\right) = \\
  &= 3\left(x+\frac{y}{2}+2\right)^2+\frac{25}{4}\left(y^2-4y+4\right) = \\
  &= 3\left(x+\frac{y}{2}+2\right)^2+\frac{25}{4}\left(y-2\right)^2\\
\end{split}$

Запомним последнее выражение, оно пригодится для нахождение аффинного преобразования, переводящего исходный случайный вектор в стандартный нормальный. Продолжим вычисления.

$\begin{split}
  q(x,y) &= 3\left(x+\frac{y}{2}+2\right)^2+\frac{25}{4}\left(y-2\right)^2 = \\
  &= 3\left(x+3+\frac{1}{2}\left(y-2\right)\right)^2+\frac{25}{4}\left(y-2\right)^2 = \\
  &= 3\left(\left(x+3\right)^2+(x+3)(y-2)+\frac{1}{4}\left(y-2\right)^2\right)+\frac{25}{4}\left(y-2\right)^2 = \\
  &= 3\left(x+3\right)^2+3(x+3)(y-2)+7\left(y-2\right)^2
\end{split}$

Выпишем вектор математических ожиданий и обратную ковариационную матрицу.

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} -3 \\ 2 \end{pmatrix}$

$\Sigma^{-1} = \begin{pmatrix} 3 & \frac{3}{2} \\ \frac{3}{2} & 7\end{pmatrix}$

Найдем матрицу ковариаций.

$\Sigma = \frac{4}{75} \begin{pmatrix} 7 & -\frac{3}{2} \\ -\frac{3}{2} & 3\end{pmatrix} = \frac{2}{75} \begin{pmatrix} 14 & -3 \\ -3 & 6\end{pmatrix}$

$\mathbb{D}\xi = \frac{28}{75},\ \mathbb{D}\eta = \frac{4}{25},\ \mathop{\mathrm{cov}}(\xi, \eta) = -\frac{2}{25}$

Найдем коэффициент корреляции.

$\rho(\xi, \eta) = -\frac{2}{25} \cdot \frac{5\sqrt{3}}{2\sqrt{7}}\cdot\frac{5}{2} = -\frac{\sqrt{3}}{2\sqrt{7}}$

Найдем неизвестную $C$.

$C = \frac{1}{2\pi\cdot\frac{2}{5\sqrt{3}}} = \frac{5\sqrt{3}}{4\pi}$

Выпишем функцию плотности распределения.

$\mathcal{p}_{\xi,\eta}(x,y) = \frac{5\sqrt{3}}{4\pi} \exp\left\{-\frac{1}{2}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -3 \\ 2 \end{pmatrix} \right]^\intercal \begin{pmatrix} 3 & \frac{3}{2} \\ \frac{3}{2} & 7\end{pmatrix}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -3 \\ 2 \end{pmatrix} \right]\right\}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
random_seed = 42


#Иницализация матрицы ковариаций и вектора математических ожиданий
cov = np.array([[28/75, -2/25], [-2/25, 4/25]])
mean = np.array([-3,2])

#Генерация распределения
distr = multivariate_normal(cov = cov, mean = mean, seed = random_seed)

#Создание сетки с интервалом 3 сигмы в каждую сторону
x = np.linspace(mean[0] - 3*np.sqrt(cov[0,0]), mean[0] + 3*np.sqrt(cov[0,0]), num=100)
y = np.linspace(mean[1] - 3*np.sqrt(cov[1,1]), mean[1] + 3*np.sqrt(cov[1,1]), num=100)
X, Y = np.meshgrid(x,y)

#Создание значений функции
Z = np.zeros(X.shape)
for i in range(X.shape[0]):
  for j in range(X.shape[1]):
    Z[i,j] = distr.pdf([X[i,j], Y[i,j]])

#Отрисовка графика
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, cmap = "plasma")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel(r'$\mathcal{p}_{\xi,\eta}(x,y)$', labelpad=5)
plt.title("График плотности вероятности случайного вектора")
plt.show()
```

### 4. Вычисление характеристик совместного распределения случайного вектора $(-4\xi + 2\eta, 2\xi + \eta)^\intercal$, нахождение его плотности. Построение графика плотности полученного распределения.

Выпишем матрицу линейного преобразования.

$A = \begin{pmatrix} -4 & 2 \\ 2 & 1\end{pmatrix}$

Данная матрица — невырожденная ($\det A = -8$). Вычислим характеристики распределения.

$\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} -4 & 2 \\ 2 & 1\end{pmatrix}\begin{pmatrix} \xi \\ \eta \end{pmatrix}$

$\mathbb{E}\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} -4 & 2 \\ 2 & 1\end{pmatrix} \begin{pmatrix} -3 \\ 2 \end{pmatrix} = \begin{pmatrix} 16 \\ -4 \end{pmatrix}$

$\Sigma_{4} = \frac{2}{75}\begin{pmatrix} -4 & 2 \\ 2 & 1\end{pmatrix}\begin{pmatrix} 14 & -3 \\ -3 & 6\end{pmatrix}\begin{pmatrix} -4 & 2 \\ 2 & 1\end{pmatrix} = \frac{2}{75}\begin{pmatrix} -62 & 24 \\ 25 & 0\end{pmatrix}\begin{pmatrix} -4 & 2 \\ 2 & 1\end{pmatrix} = \frac{2}{75}\begin{pmatrix} 296 & -100 \\ -100 & 50\end{pmatrix} = \frac{4}{75}\begin{pmatrix} 148 & -50 \\ -50 & 25\end{pmatrix}$

$\mathbb{D}\xi_{4} = \frac{592}{75},\ \mathbb{D}\eta_{4} = \frac{4}{3},\ \mathop{\mathrm{cov}}(\xi_{4}, \eta_{4}) = -\frac{8}{3},\ \rho(\xi_{4}, \eta_{4}) = -\frac{8}{3} \cdot \frac{5\sqrt{3}}{4\sqrt{37}} \cdot \frac{\sqrt{3}}{2} = -\frac{5}{\sqrt{37}}$

$\det \Sigma_4 = \frac{16}{75\cdot75}\cdot25(148 - 100) = \frac{256}{75}\\
\Sigma_{4}^{-1} = \frac{75}{256} \cdot \frac{4}{75} \begin{pmatrix} 25 & 50 \\ 50 & 148\end{pmatrix} = \begin{pmatrix} \frac{25}{64} & \frac{25}{32} \\ \frac{25}{32} & \frac{37}{16}\end{pmatrix}$

$C = \frac{1}{2\pi\cdot\frac{16}{5\sqrt{3}}} = \frac{5\sqrt{3}}{32\pi}$

Выпишем плотность полученного распределения.

$\mathcal{p}_{\xi_4,\eta_4}(x,y) = \frac{5\sqrt{3}}{32\pi} \exp\left\{-\frac{1}{2}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 16 \\ -4 \end{pmatrix} \right]^\intercal \begin{pmatrix} \frac{25}{64} & \frac{25}{32} \\ \frac{25}{32} & \frac{37}{16}\end{pmatrix}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 16 \\ -4 \end{pmatrix} \right]\right\}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
random_seed = 42


#Иницализация матрицы ковариаций и вектора математических ожиданий
cov = np.array([[592/75, -8/3], [-8/3, 4/3]])
mean = np.array([16, -4])

#Генерация распределения
distr = multivariate_normal(cov = cov, mean = mean, seed = random_seed)

#Создание сетки с интервалом 3 сигмы в каждую сторону
x = np.linspace(mean[0] - 3*np.sqrt(cov[0,0]), mean[0] + 3*np.sqrt(cov[0,0]), num=100)
y = np.linspace(mean[1] - 3*np.sqrt(cov[1,1]), mean[1] + 3*np.sqrt(cov[1,1]), num=100)
X, Y = np.meshgrid(x,y)

#Создание значений функции
Z = np.zeros(X.shape)
for i in range(X.shape[0]):
  for j in range(X.shape[1]):
    Z[i,j] = distr.pdf([X[i,j], Y[i,j]])

#Отрисовка графика
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, cmap = "plasma")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel(r'$\mathcal{p}_{\xi_{4},\eta_{4}}(x,y)$', labelpad=5)
plt.title(r"График плотности совместного распределения случайного вектора $(-4\xi + 2\eta, 2\xi + \eta)^\intercal$")
plt.show()
```

### 5. Нахождение условного распределения $\xi$ при условии $\eta$. Вычисление $\mathbb{E}(\xi|\eta)$ и $\mathbb{D}(\xi|\eta)$.

Условные распределения полученные из многомерного нормального, также являются многомерными нормальными.

$\mathcal{p}_{\xi|\eta=y}(x) = \frac{\mathcal{p}_{\xi,\eta}(x,y)}{\mathcal{p}_{\eta}(y)}$

Компоненты многомерного нормального распределения имеют одномерное нормальное распределение. Запишем его плотность.

$\mathbb{E}\eta = 2,\ \mathbb{D}\eta = \frac{4}{25}$

$\mathcal{p}_{\eta}(y) = \frac{1}{\sqrt{2\pi}\cdot\frac{2}{5}} \exp\left(-\frac{1}{2}(y-2)\frac{25}{4}(y-2)\right) = \frac{5}{2\sqrt{2\pi}} \exp\left(-\frac{1}{2}\left(\frac{25}{4}y^2-25y+25\right)\right)$

Найдем плотность условного распределения.

$\begin{split}
  \mathcal{p}_{\xi|\eta=y}(x) &= \frac{\frac{5\sqrt{3}}{4\pi}\exp\left(-\frac{1}{2}\left(3x^2+3xy+7y^2+12x-19y+37\right)\right)}{\frac{5}{2\sqrt{2\pi}} \exp\left(-\frac{1}{2}\left(\frac{25}{4}y^2-25y+25\right)\right)} = \\
  &= \frac{\sqrt{3}}{\sqrt{2\pi}}\exp\left(-\frac{1}{2}\left(3x^2+3xy+7y^2+12x-19y+37-\frac{25}{4}y^2+25y-25\right)\right) = \\
  &= \frac{\sqrt{3}}{\sqrt{2\pi}}\exp\left(-\frac{1}{2}\left(3x^2+3xy+\frac{3}{4}y^2+12x+6y+12\right)\right) = \\
  &= \frac{\sqrt{3}}{\sqrt{2\pi}}\exp\left(-\frac{1}{2} \cdot \frac{1}{3}\left(x^2+xy+\frac{1}{4}y^2+4x+2y+4\right)\right) = \\
  &= \frac{\sqrt{3}}{\sqrt{2\pi}}\exp\left(-\frac{1}{2} \cdot \frac{1}{3}\left(x^2+2x\left(\frac{1}{2}y + 2\right)+\frac{1}{4}y^2+2y+4\right)\right) = \\
  &= \frac{\sqrt{3}}{\sqrt{2\pi}}\exp\left(-\frac{1}{2} \cdot \frac{1}{3}\left(x + \left(\frac{1}{2}y + 2\right)\right)^2\right)
\end{split}$

Выпишем его математическое ожидание и дисперсию.

$\mathbb{E}(\xi|\eta=y) = -\frac{1}{2}y-2,\ \mathbb{D}(\xi|\eta=y) = \frac{1}{3}$
