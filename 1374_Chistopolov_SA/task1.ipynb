## Основные теоретические положения
Двумерное нормальное распределение - это вероятностное распределение двух случайных переменных, которые коррелированы и имеют нормальное распределение. Основные положения двумерного нормального распределения:

1. Математическое ожидание: Математическое ожидание двумерного нормального распределения задается вектором средних значений, обозначаемым как $\mathbb{E}$. Он представляет собой центр распределения и определяет среднее значение каждой случайной переменной.

2. Ковариационная матрица: Ковариационная матрица двумерного нормального распределения, обозначаемая как $\Sigma$, определяет степень корреляции и дисперсию каждой переменной.

3. Плотность распределения: Плотность вероятности двумерного нормального распределения задается формулой:
$f(x,y) = \frac{1}{2\pi \cdot \sigma_{x} \cdot \sigma_{y} \cdot \sqrt{1 - \rho^{2}}} \cdot exp^{-\frac{1}{2 \cdot(1-\rho^{2})} \cdot \left[ \frac{(x - \mathbb{E}_{x})^{2}}{\sigma_{x}^{2}} - \frac{2\rho \cdot(x-\mathbb{E}_{x})\cdot(y-\mathbb{E}_{y})}{\sigma_{x} \cdot \sigma_{y}} + \frac{(y-\mathbb{E}_{y})^{2}}{\sigma_{y}^{2}} \right]}$

4. Форма распределения: Двумерное нормальное распределение имеет форму эллипса в двумерном пространстве. Ориентация и размер эллипса определяются ковариационной матрицей, а среднее значение задает центр эллипса.

5. Корреляция: Коэффициент корреляции $\rho$ определяет степень линейной связи между переменными $x$ и $y$. Значение $\rho$ находится в диапазоне от -1 до 1, где отрицательное значение указывает на обратную связь, положительное - на прямую связь, а нулевое - на отсутствие линейной связи.

### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $P_{\xi,\eta}$
Для выполнения задания начнем с преобразования Лагранжа:
$q(x,y) = 2x^{2}-4xy+5y^{2}-12y+12 = 2(x^{2}-2xy)+5y^{2}-12y+12 = 2(x^{2}-2xy+y^{2})-2y^{2}+5y^{2}-12y+12 = 2(x-y)^{2}+3(y^{2}-4y+4) = 2(x-y)^{2}+3(y-2)^{2} = 2(x-(y-2)-2)^{2}+3(y-2)^{2} = 2((x-2)-(y-2))^{2}+3(y-2)^{2} = 2((x-2)^{2}-2(x-2)(y-2)+(y-2)^{2})+3(y-2)^{2} = 2(x-2)^{2}-4(x-2)(y-2)+5(y-2)^{2}$

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 2 \\ 2 \end{pmatrix}$

$\Sigma^{-1} = \begin{pmatrix} 2 & -2 \\ -2 & 5 \end{pmatrix} \implies \det{\Sigma^{-1}} = 10-4 = 6$

$\Sigma = \frac{1}{\det{\Sigma^{-1}}} \cdot \begin{pmatrix} 5 & 2 \\ 2 & 2 \end{pmatrix} = \frac{1}{6} \cdot \begin{pmatrix} 5 & 2 \\ 2 & 2 \end{pmatrix} = \begin{pmatrix} \frac{5}{6} & \frac{1}{3} \\ \frac{1}{3} & \frac{1}{3} \end{pmatrix}$

Из $\Sigma$ матрицы получаем следующие характеристики:

$\mathbb{D}\xi = \frac{5}{6}, \mathbb{D}\eta = \frac{1}{3}, \mathop{\mathrm{cov}}\nolimits(\xi, \eta) = \frac{1}{3}$

Найдем СКО, коэффициент корреляции и С:

$\sigma_{\xi} = \sqrt{\frac{5}{6}}$

$\sigma_{\eta} = \sqrt{\frac{1}{3}}$

$\rho_{\xi,\eta} = \frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta) }{\sigma_{\xi} \cdot \sigma_{\eta}} = \frac{\frac{1}{3}} {\sqrt{\frac{5}{6} \cdot \frac{1}{3}}} = \frac{\frac{1}{3}} {\sqrt{\frac{5}{9 \cdot 2}}} = \frac{\frac{1}{3}} {\frac{1}{3} \sqrt{\frac{5}{2}}} = \frac{1}{\frac{\sqrt{5}}{\sqrt{2}}} = \frac{\sqrt{2}}{\sqrt{5}}$

$C = \frac{1}{2\pi \cdot \sqrt{\frac{1}{\det(\Sigma)^{-1}}}} = \frac{1}{2\pi \cdot \sqrt{\frac{1}{6}}} = \frac{\sqrt{6}}{2\pi}$

Тогда плотность распределения будет равна:

$P_{\xi,\eta}(x,y) = \frac{\sqrt{6}}{2\pi} \cdot exp\left\{ -\frac{1}{2} \cdot \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 2 \\ 2 \end{pmatrix}\right]^{\intercal} \cdot \begin{pmatrix} 2 & -2 \\ -2 & 5 \end{pmatrix}  \cdot \left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 2 \\ 2 \end{pmatrix} \right] \right\}$

```python
## График плотности распределения к №1.
import numpy as np
import matplotlib.pyplot as plt

def density(x, y):
    mean = np.array([2, 2])
    covariance = np.array([[2, -2], [-2, 5]])
    diff = np.stack((x, y), axis=-1) - mean
    exponent = -0.5 * np.einsum('...k,kl,...l->...', diff, covariance, diff)
    return np.sqrt(6) / (2 * np.pi) * np.exp(exponent)

x = np.linspace(-3, 7.5, 100)
y = np.linspace(-2, 6, 100)
X, Y = np.meshgrid(x, y)

Z = density(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

### 2. Найти аффинное преобразование, переводящее исходный случайный вектор в стандартный нормальный.

Произведем нормирование $q(x,y)$:

$q(x,y) = 2(x-y)^{2}+3(y-2)^{2} = \left(\frac{x}{\sqrt{2}} - \frac{y}{\sqrt{2}}\right)^{2} + \left(\frac{y}{\sqrt{3}} - \frac{2}{\sqrt{3}}\right)^{2}$

Построим матрицу:

A = $\begin{pmatrix} \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \\ 0 & \frac{1}{\sqrt{3}} \end{pmatrix}$

$\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = \begin{pmatrix} \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \\ 0 & \frac{1}{\sqrt{3}} \end{pmatrix} \cdot \begin{pmatrix} \xi \\ \eta \end{pmatrix} + \begin{pmatrix} 0 \\ -\frac{2}{\sqrt{3}} \end{pmatrix} = \begin{pmatrix} \frac{1}{\sqrt{2}}\xi - \frac{1}{\sqrt{2}}\eta + 0 \\ 0\xi + \frac{1}{\sqrt{3}}\eta - \frac{2}{3} \end{pmatrix}$

Сделаем проверку:

$\mathbb{E}\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = \begin{pmatrix} \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \\ 0 & \frac{1}{\sqrt{3}} \end{pmatrix} \cdot \begin{pmatrix} 2 \\ 2 \end{pmatrix} + \begin{pmatrix} 0 \\ -\frac{2}{\sqrt{3}} \end{pmatrix} = \begin{pmatrix} \frac{2}{\sqrt{2}} - \frac{2}{\sqrt{2}} + 0 \\ 0 + \frac{2}{\sqrt{3}} - \frac{2}{\sqrt{3}} \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$

$\Sigma_{st} = A \cdot \Sigma \cdot A^{\intercal} = \frac{1}{6} \cdot \begin{pmatrix} \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \\ 0 & \frac{1}{\sqrt{3}} \end{pmatrix} \cdot \begin{pmatrix} 5 & 2 \\ 2 & 2 \end{pmatrix} \cdot \begin{pmatrix} \frac{1}{\sqrt{2}} & 0 \\ -\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{3}} \end{pmatrix} = \frac{1}{6} \cdot \begin{pmatrix} \frac{3}{\sqrt{2}} & 0 \\ \frac{2}{\sqrt{3}} & \frac{2}{\sqrt{3}} \end{pmatrix} \cdot \begin{pmatrix} \frac{1}{\sqrt{2}} & 0 \\ -\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{3}} \end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$

### 3. Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построить график плотности полученного распределения.

$\Sigma^{-1} = \begin{pmatrix} 2 & -2 \\ -2 & 5 \end{pmatrix}$

Посчитаем определитель и найдем $\lambda$:

$\begin{vmatrix} 2-\lambda & -2 \\ -2 & 5-\lambda \end{vmatrix} = (2-\lambda) \cdot (5-\lambda)-4 = 10-2\lambda-5\lambda+\lambda^{2}-4 = \lambda^{2}-7\lambda+6 = (\lambda-1)(\lambda-6)$

Рассмотрим случай, когда $\lambda_{1}=1$:

$\left(\begin{array}{@{} rr|r @{\\,}} 1 & -2 & 0 \\ -2 & 4 & 0 \end{array}\right) \sim \left(\begin{array}{@{} rr|r @{\\,}} 1 & -2 & 0 \\ 1 & -2 & 0 \end{array}\right) \sim \left(\begin{array}{@{} rr|r @{\\,}} 1 & -2 & 0 \\ 0 & 0 & 0 \end{array}\right)$

$x-2y = 0 \implies x = 2y \implies \overrightarrow{\mathrm{V}_{1}} = \begin{pmatrix} 2 \\ 1 \end{pmatrix}$

$\left\| \overrightarrow{\mathrm{V}_{1}} \right\| = \sqrt{2^{2}+1^{2}} = \sqrt{5} \implies \overrightarrow{\mathrm{V}_{1}^{'}} = \begin{pmatrix} \frac{2}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} \end{pmatrix}$

Рассмотрим случай, когда $\lambda_{2}=6$:

$\left(\begin{array}{@{} rr|r @{\\,}} -4 & -2 & 0 \\ -2 & 4 & 0 \end{array}\right) \sim \left(\begin{array}{@{} rr|r @{\\,}} -2 & -1 & 0 \\ -2 & -1 & 0 \end{array}\right) \sim \left(\begin{array}{@{} rr|r @{\\,}} 0 & 0 & 0 \\ -2 & -1 & 0 \end{array}\right)$

$-2x-y = 0 \implies -2x = y \implies \overrightarrow{\mathrm{V}_{2}} = \begin{pmatrix} 1 \\ -2 \end{pmatrix}$

$\left\| \overrightarrow{\mathrm{V}_{2}} \right\| = \sqrt{1^{2}+(-2)^{2}} = \sqrt{5} \implies \overrightarrow{\mathrm{V}_{2}^{'}} = \begin{pmatrix} \frac{1}{\sqrt{5}} \\ -\frac{2}{\sqrt{5}} \end{pmatrix}$

$Q = \begin{pmatrix} \frac{2}{\sqrt{5}} & \frac{1}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} & -\frac{2}{\sqrt{5}} \end{pmatrix}$

$Q^{\intercal} = \begin{pmatrix} \frac{2}{\sqrt{5}} & \frac{1}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} & -\frac{2}{\sqrt{5}} \end{pmatrix}$

$\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} = \begin{pmatrix} \frac{2}{\sqrt{5}} & \frac{1}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} & -\frac{2}{\sqrt{5}} \end{pmatrix} \cdot \begin{pmatrix} \xi \\ \eta \end{pmatrix}$



$\mathbb{E}\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} = \begin{pmatrix} \frac{2}{\sqrt{5}} & \frac{1}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} & -\frac{2}{\sqrt{5}} \end{pmatrix} \cdot \begin{pmatrix} 2 \\ 2 \end{pmatrix} = \begin{pmatrix} \frac{6}{\sqrt{5}} \\ -\frac{2}{\sqrt{5}} \end{pmatrix}$

$\Sigma_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}} = Q*\Sigma_{\begin{pmatrix} \xi \\ \eta \end{pmatrix}} \cdot Q^{\intercal} = \frac{1}{6} \cdot \begin{pmatrix} \frac{2}{\sqrt{5}} & \frac{1}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} & -\frac{2}{\sqrt{5}} \end{pmatrix} \cdot \begin{pmatrix} 5 & 2 \\ 2 & 2 \end{pmatrix} \cdot \begin{pmatrix} \frac{1}{\sqrt{5}} & \frac{1}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} & -\frac{2}{\sqrt{5}} \end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 0 & \frac{1}{6} \end{pmatrix} = \begin{pmatrix} \frac{1}{\lambda_{1}} & 0 \\ 0 & \frac{1}{\lambda_{2}} \end{pmatrix} \implies \Sigma^{-1}_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}} = \begin{pmatrix} \lambda_{1} & 0 \\ 0 & \lambda_{2} \end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 0 & 6 \end{pmatrix}$

Выпишем числовые характеристики, полученные в $\Sigma_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}}$:

$\mathbb{D}_{\xi} = 1; \mathbb{D}_{\eta} = \frac{1}{6}; \mathop{\mathrm{cov}}\nolimits(\xi,\eta) = 0 \implies \rho_{\xi,\eta} = 0$

$\sigma_{\xi} = \sqrt{1} = 1; \sigma_{\eta} = \frac{1}{\sqrt{6}}$

$C = \frac{1}{2\pi \cdot \frac{1}{\sqrt{6}}} = \frac{\sqrt{6}}{2\pi}$

Запишем плотность полученного распределения: 

$P_{\xi,\eta}(x,y)= \frac{\sqrt{6}}{2\pi} \cdot exp\left\{-\frac{1}{2} \cdot \left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} \frac{6}{\sqrt{5}} \\ -\frac{2}{\sqrt{5}} \end{pmatrix}\right]^{\intercal} \cdot \begin{pmatrix} 1 & 0 \\ 0 & 6 \end{pmatrix} \cdot \left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} \frac{6}{\sqrt{5}} \\ -\frac{2}{\sqrt{5}} \end{pmatrix} \right] \right\}$

```python
## График плотности распределения к №3.
import numpy as np
import matplotlib.pyplot as plt

def density(x, y):
    mean = np.array([6/np.sqrt(5), -2/np.sqrt(5)])
    covariance = np.array([[1, 0], [0, 6]])
    diff = np.stack((x, y), axis=-1) - mean
    exponent = -0.5 * np.einsum('...k,kl,...l->...', diff, covariance, diff)
    return np.sqrt(6) / (2 * np.pi) * np.exp(exponent)

x = np.linspace(-3, 7.5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = density(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

### 4. Вычислить характеристики совместного распределения случайного вектора $(3ξ - 2η, 3ξ + 4η)^{\intercal}$ и записать его плотность. Построить график плотности полученного распределения.

$A_{4} = \begin{pmatrix} 3 & -2 \\ 3 & 4 \end{pmatrix} \implies \det(A_{4}) = 12+6 = 18 \neq 0$

$A_{4}^{-1} = \frac{1}{18} \cdot \begin{pmatrix} 3 & -3 \\ 2 & 4 \end{pmatrix}$

$\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = A_{4} \cdot \begin{pmatrix} \xi \\ \eta \end{pmatrix}$

$\mathbb{E} \begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} 3 & -2 \\ 3 & 4 \end{pmatrix} \cdot \begin{pmatrix} 2 \\ 2 \end{pmatrix} = \begin{pmatrix} 2 \\ 14 \end{pmatrix}$

Найдем $\Sigma$ и числовые характеристики распределения:

$\Sigma_{4} = \frac{1}{6} \cdot \begin{pmatrix} 3 & -2 \\ 3 & 4 \end{pmatrix} \cdot \begin{pmatrix} 5 & 2 \\ 2 & 2 \end{pmatrix} \cdot \begin{pmatrix} 3 & 3 \\ -2 & 4 \end{pmatrix} = \frac{1}{6} \cdot \begin{pmatrix} 11 & 2 \\ 23 & 14 \end{pmatrix} \cdot \begin{pmatrix} 3 & 3 \\ -2 & 4 \end{pmatrix} = \frac{1}{6} \cdot \begin{pmatrix} 29 & 41 \\ 41 & 125 \end{pmatrix} = \begin{pmatrix} \frac{29}{6} & \frac{41}{6} \\ \frac{41}{6} & \frac{125}{6} \end{pmatrix}$

$\mathbb{D}_{\xi} = \frac{29}{6}; \mathbb{D}_{\eta} = \frac{125}{6}; \mathop{\mathrm{cov}}\nolimits(\xi,\eta) = \frac{41}{6}$

$\sigma_{\xi} = \sqrt{\frac{29}{6}}; \sigma_{\eta} = \frac{5\sqrt{5}}{6}$

$\rho_{\xi,\eta} = \frac{\mathop{\mathrm{cov}}\nolimits(\xi,\eta)}{\sigma_{\xi} \cdot \sigma_{\eta}} = \frac{\frac{41}{6}}{5\sqrt{\frac{29 \cdot 5}{6 \cdot 6}}} = \frac{\frac{41}{6}}{\frac{5}{6}\sqrt{145}} = \frac{41 \cdot 6}{6 \cdot 5 \cdot \sqrt{145}} = \frac{41}{5\sqrt{145}}$

$\Sigma_{4}^{-1} = \frac{1}{54} \cdot \frac{1}{6} \cdot \begin{pmatrix} 29 & 41 \\ 41 & 125 \end{pmatrix} = \begin{pmatrix} \frac{29}{324} & \frac{41}{324} \\ \frac{41}{324} & \frac{125}{324} \end{pmatrix}$

$C = \frac{1}{2\pi \cdot \sqrt{\det{\Sigma_{4}}}} = \frac{1}{2\pi \cdot \sqrt{54}} = \frac{1}{6\pi \sqrt{6}}$

Запишем плотность распределения: 

$P_{\xi,\eta}(x,y) = \frac{1}{6\pi \sqrt{6}} \cdot exp \left\{-\frac{1}{2} \cdot \left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 2 \\ 14 \end{pmatrix} \right]^{\intercal} \cdot \begin{pmatrix} \frac{29}{324} & \frac{41}{324} \\ \frac{41}{324} & \frac{125}{324} \end{pmatrix} \cdot \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 2 \\ 14 \end{pmatrix} \right]\right\}$

```python
## График плотности распределения к №4.
import numpy as np
import matplotlib.pyplot as plt

def density(x, y):
    mean = np.array([2, 14])
    covariance = np.array([[29/324, 41/324], [41/324, 125/324]])
    diff = np.stack((x, y), axis=-1) - mean
    exponent = -0.5 * np.einsum('...k,kl,...l->...', diff, covariance, diff)
    return 1 / (6 * np.pi * np.sqrt(6)) * np.exp(exponent)

x = np.linspace(-20, 20, 100)
y = np.linspace(0, 30, 100)
X, Y = np.meshgrid(x, y)

Z = density(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

### 5. Найти условное распределение $\xi$ при условии $\eta$. Вычислить $\mathbb{E}(\xi|\eta)$ и $\mathbb{D}(\xi|\eta)$.

$P_{(\xi|\eta=y)} = \frac{P_{\xi,\eta}(x,y)}{P_{\eta}(y)}$

Найдем для начала $P_{\eta}(y)$ по формуле:

$P_{\eta}(y) = \frac{1}{\sqrt{2\pi} \cdot \sqrt{\det{\Sigma}}} \cdot exp^{-\frac{1}{2}(y-\mathbb{E}_{y}) \cdot \Sigma^{-1} \cdot (y-\mathbb{E}_{y})}$

Возьмем некоторые данные из №1:

$\mathbb{E}_{\eta} = 2$

$\mathbb{D}_{\eta} = \frac{1}{3}$

Рассмотрим одномерный случай, тогда: 

$\Sigma_{\eta} = \mathbb{D}_{\eta} = \frac{1}{3} \implies \Sigma^{-1}_{\eta} = 3$

$P_{\eta}(y) = \frac{\sqrt{3}}{\sqrt{2\pi}} \cdot exp^{-\frac{1}{2}(y-2) \cdot 3(y-2)}$

Теперь найдем условное распределение:

$P_{(\xi|\eta=y)} = \frac{\frac{\sqrt{6}}{2\pi} \cdot exp^{-\frac{1}{2}(2x^{2}-4xy+5y^{2}-12y+12)}}{\frac{\sqrt{3}}{\sqrt{2\pi}} \cdot exp^{-\frac{1}{2}(3y^{2}-12y+12)}} = \frac{\sqrt{6} \cdot \sqrt{2\pi}}{2\pi \cdot \sqrt{3}} \cdot exp^{-\frac{1}{2}(2x^{2}-4xy+5y^{2}-12y+12-3y^{2}+12y-12)} = \frac{1}{\sqrt{\pi}} \cdot exp^{-\frac{1}{2}(2x^{2}-4xy+2y^{2})} = \frac{1}{\sqrt{\pi}} \cdot exp^{-\frac{1}{2}(\frac{1}{2}(x^{2}-2xy+y^{2}))} = \frac{1}{\sqrt{\pi}} \cdot exp^{-\frac{1}{2}(\frac{1}{2}(x-y)^{2})}$

Воспользуемся формулой:

$P_{(\xi|\eta)} = C \cdot exp^{-\frac{1}{2} \cdot \left(\frac{1}{\mathbb{D}_{(\xi|\eta)}} \cdot(x-\mathbb{E}_{(\xi|\eta)})^{2} \right)}$ 

И найдем $\mathbb{E}(\xi|\eta)$ и $\mathbb{D}(\xi|\eta)$:

$P_{(\xi|\eta)} = \frac{1}{\sqrt{\pi}} \cdot exp^{-\frac{1}{2}\left(\frac{1}{2}(x-y)^{2} \right)} \implies \mathbb{E}(\xi|\eta) = y; \mathbb{D}(\xi|\eta) = \frac{1}{2}$
