## Основные теоретические положения

Двумерное нормальное распределение вероятностей:

$f(x,y) = \frac{1}{2*\pi*\sigma_{x}*\sigma_{y}*\sqrt{1 - \rho^{2}}}*exp^{-\frac{1}{2*(1-\rho^{2})}\left[ \frac{(x - \mathbb{E}x)^{2}}{\sigma_{x}^{2}} - \frac{2*\rho*(x-\mathbb{E}x)*(y-\mathbb{E}y))}{\sigma_{x}*\sigma_{y}} + \frac{(y-\mathbb{E}y)^{2}}{\sigma_{y}^{2}} \right]}$

$\rho$ - коэффициент коррелции случайно величины X и Y

$\sigma_{x}$ - среднее квадратическое отклонение случайной величины X

$\sigma_{y}$ - среднее квадратическое отклонение случайной величины Y

$\mathbb{E}x$ - математическое ожидание случайной величины X

$\mathbb{E}y$ - математическое ожидание случайной величины Y

1. Чтобы вычислить вектор математических ожиданий воспользуемся преобразовением Лагранжа, в результате которого мы получим запись следующего вида: $q(x,y) = a*(x - \alpha) +2*c*(x - \alpha)*(y - \beta) + b*(y - \beta)$ 

Для нахождения дисперсий ковариации нам поможет Ковариационная матрица $\Sigma$: $\Sigma = \begin{pmatrix} \mathbb{D}\xi & \mathop{\mathrm{cov}}\nolimits(\xi, \eta) \\ \mathop{\mathrm{cov}}\nolimits(\eta, \xi) & \mathbb{D}\eta \end{pmatrix}$

Кроме таких числовых характеристик,как $\mathbb{E}\xi, \mathbb{E}\eta, \mathbb{D}\xi, \mathbb{D}\eta$ и $ \mathop{\mathrm{cov}}\nolimits(\xi, \eta)$ , мы можем найти коэффициент корреляции $\rho$ и СКО $\sigma_{\xi}, \sigma_{\eta}$.

$\sigma_{\xi} = \sqrt{\mathbb{D_{\xi}}}$   

$\rho_{\xi\eta} = \frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta) }{\sigma_{\xi}*\sigma_{\eta}}$

$C = \frac{1}{2*\pi*\sqrt{\det(\Sigma)}}$

2. Новый случайный вектор  $\mathbf{Y}$ можно определить, применяя аффинное преобразование g:  $\mathbb{R}^{n} \to \mathbb{R}^{n}$ к случайному вектору  $\mathbf{X}$:

$\textbf{Y} = A*\textbf{X} + B$, где A является матрицей коэффициентов при $\xi, \eta$ размера $n*n$, матрица B - матрица свободных членов размера $n*1$.

Или можно представить в таком виде: $A*\overrightarrow{\xi} + B\sim \mathcal{N(o,\mathbb{E})}$. Для достиженияцели нормируем q(x,y) и проведем проверки: $\mathbb{E}\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = A*\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} + B = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$ и $\Sigma_{st} = A*\Sigma*A^{T} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$

3. Для ортогонального преобразования, переводящий соответствующий центрированный случайны вектор в вектор с независимыми компонентами нам нужно решить следующее уравнение $\lambda^{2} - tr(A)*\lambda + \det(A) = 0$

Находим ковариационные числа $\left\{ \mathrm{\lambda}_{1},...,\mathrm{\lambda}_{k} \right\}$ и соответствующие им собственные вектора $\left\{ \mathrm{V}_{1},...,\mathrm{V}_{k} \right\}$ ковариционной матрицы $\Sigma$ и нормируем вектора, получая ортонормированный базис

Собственные векторы, отвечающие различным собственным значениям, ортогональны между собой. Сформировав матрицу Q из нормированных собственных векторов, остается перемножить матрицу на центрированный вектор, чтобы получить вектор с независимыми компонентами.

Поскольку мы получаем вектор с независимыми переменнами, то по свойству ковариации: Если $\xi,\eta$ - независимые переменные, то $\mathop{\mathrm{cov}}\nolimits(\xi, \eta) = 0$

4. При нахождении условного распределения $\xi$ при условии $\eta$ воспользуемся формулой условного распределения $P_{(\xi|\eta=y)} = \frac{P_{\xi,\eta}(x,y))}{P_{\eta}(y)}$.

Условное распределение полученные из многомерного нормального распределения, также является многомерным распределением. Для одномерного нормального распределения ковариционная матрица $\Sigma$ будет состоять только из дисперсии.

Для подсчета $\mathbb{E}_{(\xi|\eta)}$ и $\mathbb{D}_{(\xi|\eta)}$ воспользуемся следующей формулой $P_{\eta}(y) = \frac{1}{\sqrt{2*\pi}*\sqrt{\det(\Sigma))}}*exp^{-\frac{1}{2}*\left( (y-\mathbb{E}_{\eta})*\Sigma^{-1}*(y-\mathbb{E}_{\eta}) \right)} = C*exp^{-\frac{1}{2}*\left( \frac{1}{\mathbb{D}_{(\xi|\eta)}}*(x-\mathbb{E}_{(\xi|\eta)})^{2} \right)}$

$P_{\xi,\eta}(x,y)=C*exp\left( -\frac{1}{2}(3x^{2}-3xy+7y^{2}-9x-8y+13) \right)$

${\bf 1. \ Вычислить \ вектор \ математических \ ожиданий \ и \ ковариационные \ характеристики \ данного \ случайного \ вектора.}$

#Произведем преобразование Лагранжа
$q(x,y) = 3x^{2}-3xy+7y^{2}-9x-8y+13 = (3x^{2} - 3xy - 9x) + 7y^{2} -8y +13 = 3(x^{2}-x(y+3)) + 7y^{2} - 8y + 13 = 3\left( x^2 - \frac{2}{2}(y+3)+ (\frac{y+3}{2})^{2} \right)-\left( \frac{y+3}{2} \right)^{2} +7y^{2} - 8y + 13 = 3\left( x - \frac{1}{2}(y+3) \right)^{2} - \frac{3}{4}y^{2} - \frac{18}{4}y - \frac{27}{4} + 7y^{2} - 8y + 13 = 3\left( x - \frac{1}{2}(y+3) \right)^{2} + \frac{25}{4}y^{2} - \frac{50}{4}y + \frac{25}{4} = 3\left( x - \frac{1}{2}(y+3) \right)^{2} + \frac{25}{4}\left( y^{2} - 2y + 1 \right) = 3\left( x - \frac{1}{2}(y+3) \right)^{2} + \frac{25}{4}\left( y - 1 \right)^{2} = 3\left( x - \frac{1}{2}(y-1) - \frac{1}{2} - \frac{3}{2} \right)^{2} + \frac{25}{4}\left( y - 1 \right)^{2} = 3\left( x -\frac{1}{2}(y - 1) - 2 \right)^{2} + \frac{25}{4}\left( y - 1 \right)^{2} = 3\left( (x - 2) - \frac{1}{2}(y - 1) \right)^{2} + \frac{25}{4}\left( y - 1 \right)^{2} = 3(x-2)^{2} - 3(x - 2)(y - 1) + 7(y - 1)^{2}$

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 2 \\ 1 \end{pmatrix}$ #нашли вектор математических ожиданий

$\Sigma^{-1} = \begin{pmatrix} 3 & -\frac{3}{2} \\ -\frac{3}{2} & 7 \end{pmatrix} \implies \det{\Sigma^{-1}} = \frac{75}{4}$

#Найдем остальные числовые характеристики данного случайного вектора

$\Sigma = \begin{pmatrix} 7 & \frac{3}{2} \\ \frac{3}{2} & 3 \end{pmatrix}* \frac{1}{\det{\Sigma^{-1}}} = \begin{pmatrix} \frac{28}{75} & \frac{2}{25} \\ \frac{2}{25} & \frac{12}{75}\end{pmatrix} \implies \mathbb{D}\xi = \frac{28}{75}, \mathbb{D}\eta = \frac{4}{25}, \mathop{\mathrm{cov}}\nolimits(\xi, \eta) = \frac{2}{25}$

$\sigma_{\xi} = \sqrt{\frac{28}{25}} = \frac{2*\sqrt{7}}{5}$

$\sigma_{\eta} = \sqrt{\frac{4}{25}} = \frac{2}{5}$

$\rho_{\xi,\eta} = \frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta) }{\sigma_{\xi}*\sigma_{\eta}} = \frac{\frac{2}{25}}{\frac{2*\sqrt{7}}{5}*\frac{2}{5}} = \frac{\sqrt{7}}{14}$

$C = \frac{1}{2*\pi*\sqrt{\det(\Sigma)}} = \frac{5*\sqrt{3}}{4*\pi}$

Плотность распределения: $P_{\xi\eta}(x,y) = \frac{5*\sqrt{3}}{4*\pi}exp\left\{ -\frac{1}{2}*\left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 2 \\ 1 \end{pmatrix}\right]^{T} * \begin{pmatrix} 3 & -\frac{3}{2} \\ -\frac{3}{2} & 7 \end{pmatrix}  * \left[ \begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} 2 \\ 1 \end{pmatrix} \right] \right\}$

```python
#Продолжение выполнения №1. Построим график плотности распределения pξ,η.
import numpy as np
import matplotlib.pyplot as plt

def normal_distribution(x, y):
    constant = 5 * np.sqrt(3) / (4 * np.pi)
    mean = np.array([2, 1])
    covariance = np.array([[3, -3/2],
                           [-3/2, 7]])
    diff = np.stack((x, y), axis=-1) - mean
    power = -0.5 * np.einsum('...k,kl,...l->...', diff, covariance, diff)
    return constant * np.exp(power)

x = np.linspace(-10, 10, 200)
y = np.linspace(-10, 10, 200)
X, Y = np.meshgrid(x, y)

Z = normal_distribution(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Плотность распределения')

plt.show()
```

```python
#Продолжение выполнения №3. Построим график плотности полученного распределения.
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def normal_distribution(x, y):
    constant = 5 * np.sqrt(3) / (4 * np.pi)
    mean = np.array([7 / np.sqrt(10), -1 / np.sqrt(10)])
    covariance = np.array([[5/2, 0],
                           [0, 15/2]])
    inv_covariance = np.linalg.inv(covariance)
    diff = np.stack((x, y), axis=-1) - mean
    power = -0.5 * np.einsum('...k,kl,...l->...', diff, inv_covariance, diff)
    return constant * np.exp(power)

x = np.linspace(-10, 10, 200)
y = np.linspace(-10, 10, 200)
X, Y = np.meshgrid(x, y)

Z = normal_distribution(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Плотность распределения')

plt.show()
```

$C = \frac{1}{2*\pi*\sqrt{\frac{16}{3}}} = \frac{\sqrt{3}}{8*\pi}$

Плотность полученного распределения: $P_{\xi_{4},\eta_{4}}(x,y) = \frac{\sqrt{3}}{8*\pi}*exp\left\{ -\frac{1}{2}*\left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -11 \\ -8 \end{pmatrix}
 \right]^{T}*\begin{pmatrix} \frac{7}{4} & \frac{5}{4} \\ \frac{5}{4} & 1 \end{pmatrix} * \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -11 \\ -8 \end{pmatrix} \right]\right\}$

```python
#Продолжение выполнения №4. Построим график плотности полученного распределения.
import numpy as np
import matplotlib.pyplot as plt

def normal_distribution(x, y):
    constant = np.sqrt(3) / (8 * np.pi)
    mean = np.array([-11, -8])
    covariance = np.array([[7/4, 5/4],
                           [5/4, 1]])
    inv_covariance = np.linalg.inv(covariance)
    diff = np.stack((x, y), axis=-1) - mean
    power = -0.5 * np.einsum('...k,kl,...l->...', diff, inv_covariance, diff)
    return constant * np.exp(power)

x = np.linspace(-15, 5, 200)
y = np.linspace(-20, 4, 200)
X, Y = np.meshgrid(x, y)

Z = normal_distribution(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Плотность распределения')

plt.show()
```
