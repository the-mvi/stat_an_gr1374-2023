## Основные теоретические положения


Многомерное нормальное распределение - это вероятностное распределение, которое описывает случайные величины, имеющие многомерную нормальную зависимость. Оно характеризуется вектором средних значений и ковариационной матрицей, которая определяет степень зависимости между различными компонентами случайных величин.

Случайный вектор $\overrightarrow{\xi} = \left(\xi_1, \dots, \xi_n\right)^\intercal:\Omega\to \mathbb{R}^n$ имеет многомерное нормальное распределение(называется нормальным), если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1, \dots, \eta_m\right)^\intercal$, матрица $\mathbf{A}$ размерности $n \times m$ и вещественный вектор $\overrightarrow{b} = (b_1,\dots,b_n)^\intercal$, такие что: $\overrightarrow{\xi} = \mathbf{A}\overrightarrow{\eta} + \overrightarrow{b}$.

Плотность многомерного невырожденного нормального распределения имеет вид: $\mathcal{p}_{\overrightarrow{\xi}\left(x_1, \dots, x_n\right)} = \frac{1}{(2\pi)^{n/2}\sqrt{\det \Sigma}} \exp\left(-\frac{1}{2}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)^\intercal \Sigma^{-1}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)\right)$, где $\overrightarrow{\mu} = \mathbb{E}\overrightarrow{\xi} = \left(\mathbb{E}\xi_1, \dots, \mathbb{E}\xi_n\right)^\intercal$ - вектор математических ожиданий компонент, $\Sigma = \left\{\sigma_{i,j} = \mathop{\mathrm{cov}}\left(\xi_i,\xi_j\right)\right\}_{i,j=1}^n$ - ковариационная матрица.

### 1. Вычислить вектор математических ожиданий и ковариационные характеристики случайного вектора. Построить график плотности распределения $p_{\xi,\eta}$.

${}\mathcal{p}_{\xi,\eta}(x,y) = C \cdot exp(\frac{-1}{2}(3x^2 + 6xy + 11y^2 + 6x - 10y + 11))$

Для рассчетов необходимых характеристик было использовано преобразование квадратичной формы к каноническому виду методом Лагранжа.

$\begin{split}
  q(x,y) &= 3x^2 + 6xy + 11y^2 + 6x - 10y + 11 = \\
  &= \left(3x^2 + 6xy + 6x\right) + 11y^2 - 10y + 11 = \\
  &= \left(3x^2 + 6x\left(y + 1\right)\right) + 11y^2 - 10y + 11 = \\
  &= 3\left(x^2 + 2x(y + 1) \right) + 11y^2 - 10y + 11 = \\
  &= 3\left( x^2 + 2x\left(y+1\right)+ \left(y+1\right)^2\right) - 3\left(y+1\right)^2 +11y^2 - 10y + 11  = \\
  &= 3\left( x^2 + 2x + 1 \right)^2 - 3y^2 -6y - 3 + 11y^2 - 10y + 11  = \\
  &= 3\left(x + y + 1\right)^2 + \left(8y^2 - 16y + 8\right) = \\
  &= 3\left(x+y+1\right)^2 + 8\left(y-1\right)^2 = 3\left(\left(x+2\right) + \left(y-1\right)\right)^2 +8\left(y-1\right)^2 =\\
  &= 3\left(\left(x+2\right)^2 + 2 \left(y-1\right)\right)^2 +8\left(y-1\right)^2 = \\
  &= 3\left(\left(x+2\right)^2 + 2\left(x+2\right)\left(y-1\right)+\left(y-1\right)^2\right) + 8\left(y-1\right)^2 \\
  &= 3\left(x+2\right)^2 + 3  \cdot 2\left(x+2\right)\left(y-1\right) + 11\left(y-1\right)^2 \\
\end{split}$

Полученное выражение будет использоваться для нахождения аффинного преобразования, которое переводит исходный случайный вектор в стандартный нормальный.

Выпишем вектор мат. ожиданий и обратную ковариационную матрицу.

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} -2 \\ 1 \end{pmatrix}$

$\Sigma^{-1} = \begin{pmatrix} 3 & 3\\ 3 & 11\end{pmatrix}$

Найдем матрицу ковариаций.

$\Sigma = \frac{1}{24} \begin{pmatrix} 11 & -3 \\ -3 & 3\end{pmatrix}$

$\mathbb{D_\xi} = \frac{11}{24},\ \mathbb{D_\eta} = \frac{1}{8},\ \mathop{\mathrm{cov}}(\xi, \eta) = -\frac{1}{8}$

Коэффициент корреляции:

$\rho(\xi, \eta) = -\frac{\sqrt{8} \cdot \sqrt{8 \cdot 3}}{8 \cdot \sqrt{11}} = -\frac{\sqrt{3}}{\sqrt{11}}$

Найдем константу $C$.

$C = \frac{1}{2\pi\cdot\frac{1}{\sqrt{24}}} = \frac{\sqrt{6}}{\pi}$

Плотность двумерного распределения.

$\mathcal{p}_{\xi,\eta}(x,y) = \frac{\sqrt{6}}{\pi} \exp\left\{-\frac{1}{2}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -2 \\ 1 \end{pmatrix} \right]^\intercal \begin{pmatrix} 3 & 3 \\ 3 & 11\end{pmatrix}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -2 \\ 1 \end{pmatrix} \right]\right\}$

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import scipy.stats
import random
calc_seed = round(random.random()*100)

D_xi = 11/24
D_eta = 3/24
covariation = -1/8
E_xi = -2
E_eta = 1

x_data = np.linspace(E_xi - 3*np.sqrt(D_xi), E_xi + 3*np.sqrt(D_xi), num=50)
y_data = np.linspace(E_eta - 3*np.sqrt(D_eta), E_eta + 3*np.sqrt(D_eta), num=50)
X, Y = np.meshgrid(x_data, y_data)

distribution_function = scipy.stats.multivariate_normal(cov = ([D_xi, covariation], [covariation, D_eta]), mean = (E_xi, E_eta), seed = calc_seed)

Z = np.zeros(X.shape)
for i in range(X.shape[0]):
  for j in range(X.shape[1]):
    Z[i,j] = distribution_function.pdf([X[i,j], Y[i,j]])

axes_graph = plt.axes(projection="3d")
axes_graph.plot_surface(X, Y, Z, cmap = "plasma")

axes_graph.set_xlabel("ax_x")
axes_graph.set_ylabel("ax_y")
axes_graph.set_zlabel(r'$\mathcal{p}_{\xi,\eta}(x,y)$')
plt.title("График плотности вероятности случайного вектора")
```

$\begin{split}

$\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = \begin{pmatrix} \frac{3}{\sqrt3} & \frac{3}{\sqrt3} \\ 0 & -\frac{4}{\sqrt2}\end{pmatrix}\begin{pmatrix} \xi \\ \eta \end{pmatrix} + \begin{pmatrix} \frac{3}{\sqrt3} \\ -\frac{4}{\sqrt2} \end{pmatrix}$

Проверка. Пересчитаем вектор математических ожиданий и ковариационную матрицу нашего распределения.

$\mathbb{E}\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = \begin{pmatrix} \frac{3}{\sqrt3} & \frac{3}{\sqrt3} \\ 0 & -\frac{4}{\sqrt2}\end{pmatrix} \begin{pmatrix} -2 \\ 1 \end{pmatrix} + \begin{pmatrix} \frac{3}{\sqrt3} \\ -\frac{4}{\sqrt2} \end{pmatrix} = \begin{pmatrix} \frac{-6}{\sqrt{3}}+\frac{3}{\sqrt{3}}+\frac{3}{\sqrt{3}} \\ -\frac{4}{\sqrt2}+\frac{4}{\sqrt2}\end{pmatrix}=\begin{pmatrix} 0 \\ 0 \end{pmatrix}$

$\Sigma_{st} = \frac{1}{24}\begin{pmatrix} \frac{3}{\sqrt{3}} & \frac{3}{\sqrt{3}} \\ 0 & \frac{4}{\sqrt{2}}\end{pmatrix}\begin{pmatrix} 11 & -3 \\ -3 & 3\end{pmatrix}\begin{pmatrix} \frac{3}{\sqrt{3}} & 0 \\ \frac{3}{\sqrt{3}} & \frac{4}{\sqrt{2}}\end{pmatrix}=\frac{1}{24}\begin{pmatrix} \frac{24}{\sqrt{3}} & 0 \\ -\frac{12}{\sqrt{2}} & \frac{12}{\sqrt{2}}\end{pmatrix}\begin{pmatrix} \frac{3}{\sqrt{3}} & 0 \\ \frac{3}{\sqrt{3}} & \frac{4}{\sqrt{2}}\end{pmatrix}=\frac{1}{24}\begin{pmatrix} \frac{72}{3} & 0 \\ 0 & \frac{48}{2}\end{pmatrix} = \frac{1}{24}\begin{pmatrix} 24 & 0 \\ 0 & 24\end{pmatrix}=\begin{pmatrix} 1 & 0 \\ 0 & 1\end{pmatrix}$

Исходя из полученных перерасчетов, делаем вывод, что полученное распределение является стандартным нормальным.

### 5. Найти условное распределение  $\xi$ при условии $\eta$. Вычислить ${\mathbb{E(\xi|\eta)}}$ и ${\mathbb{D(\xi|\eta).}}$

Условные распределения полученные из многомерного нормального, в то же время являются многомерными нормальными.

$\mathcal{p}_{\xi|\eta=y}(x) = \frac{\mathcal{p}_{\xi,\eta}(x,y)}{\mathcal{p}_{\eta}(y)}$

Компоненты многомерного нормального распределения имеют одномерное нормальное распределение, запишем ее плотность.

$\mathbb{E}\eta = 1,\ \mathbb{D}\eta = \frac{1}{8}$

$\mathcal{p}_{\eta}(y) = \frac{2\sqrt{2}}{\sqrt{2\pi}} \exp\left(-\frac{1}{2}\left(y-1\right)\cdot8\cdot\left(y-1\right)\right) = \frac{2}{\sqrt{\pi}} \exp\left(-\frac{1}{2}\left(8\cdot\left(y-1\right)^2\right)\right)$

Тогда плотность условного распределения.

$\begin{split}
  \mathcal{p}_{\xi|\eta=y}(x) &= \frac{\frac{\sqrt{6}}{\pi}\exp\left(-\frac{1}{2}\left(3x^2+6xy+11y^2+6x-10y+11\right)\right)}{\frac{2}{\sqrt{\pi}} \exp \left(-\frac{1}{2}\left(8\cdot\left(y-1\right)^2\right)\right)} = \\
  &= \frac{\sqrt{6}}{2\sqrt{\pi}} \exp\left(-\frac{1}{2}\left(3x^2+6xy+11y^2+6x-10y+11-8y^2+16y-8\right)\right) = \\
  &= \frac{\sqrt{6}}{2\sqrt{\pi}} \exp\left(-\frac{1}{2}\left(3x^2+6xy+3y^2+6x+6y+3\right)\right) = \\
  &= \frac{\sqrt{6}}{2\sqrt{\pi}} \exp\left(-\frac{1}{2}\cdot3\left(x^2+2xy+y^2+2x+2y+1\right)\right) = \\
  &= \frac{\sqrt{6}}{2\sqrt{\pi}} \exp\left(-\frac{1}{2}\cdot3\left(x+\left(y+1\right)\right)^2\right) \\
\end{split}$
