## Основные теоретические положения


Случайный вектор $\overrightarrow{\xi} = \left(\xi_1,...,\xi_n\right)^\intercal : \Omega \rightarrow \mathbb{R}^n$ называют **нормальным**(гауссовым) если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1,...,\eta_m\right)^\intercal$, такие что:

$$\overrightarrow{\xi} = \mathrm{A} \overrightarrow{\eta}+ \mathrm{B}$$

**Плотность нормального вектора** имеет вид:

$p_{\overrightarrow{\xi}}\left(x_1,...,x_n\right) = \frac{1}{\left(2\pi\right)^\frac{n}{2}}\mathrm{exp}\left(\frac{1}{2}\left(\overrightarrow{x} - \mu \right)^\intercal \Sigma^{-1}\left(\overrightarrow{x} - \mu \right)\right), \\ где \overrightarrow{x} = (x_1, ..., x_n)^\intercal$


**Свойства многомерного нормального распределения:**


*   Линейные комбинации случайных переменных, принадлежащих многомерному нормальному распределению, также являются нормально распределенными.
*   Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормальным распределениями.
*   Параметры многомерного нормального распледеления(среднее и ковариационная матрица) полностью описывают его форму.
*   Многомерное нормальное распределение устойчиво относительно линейных преобразований: если $\overrightarrow{\xi} \sim \mathcal{N}\left(\mu,\Sigma\right)$, $\mathrm{A}$ - произвольная матрица неособенная размерности $m \times n$, а $\mathrm{B}$ - произвольный вектор размерности  $m \times 1$, то
$$\mathrm{A}\overrightarrow{\xi} + \mathrm{B} \sim \mathcal{N}\left(\mathrm{A}\mu + \mathrm{B},\mathrm{A}\Sigma\mathrm{A}^\intercal\right) $$

**Ковариационные характеристики случайного вектора:**
* Ковариационная матрица случайного вектора - это квадратная симметрическая неотрицательно определённая матрица, на диагонали которой располагаются дисперсии компонент вектор, а внедиагональные элементы - ковариации между компонентами;
* Коэффициент корреляции - это мера линейной зависимости двух случайных величин:
$$\rho(x,y) = \frac{\mathrm{cov}(x, y)}{\sqrt{\mathbb{D}x\cdot\mathbb{D}y}}$$
где $\mathrm{cov}(x,y)$ - ковариация, а $\mathbb{D}$-дисперсия

###1. Вычисление вектора математических ожиданий и ковариационных характеристики данного случайного вектора. Построение графика плотности распределения$\rho_{\xi,\eta}$.


$p_{\xi,\eta}(x,y) = C \cdot exp(\frac{-1}{2}(3x^2 + 3xy+7y^2-9x-17y+13))$

Приведение квадратурной формы к каноническому виду:
$q(x,y) = 3x^2 + 3xy+7y^2-9x-17y+13 = (3x^2 + 3xy - 9x) + 7y^2 -17y+13 = 3(x^2+x(y-3))+7y^2-17y+13 =\\= 3\left(\left(x+\frac{y-3}{2}\right)^2 - \left(\frac{y-3}{2}\right)^2\right) + 7y^2 -17y+13 = 3\left(x+\frac{y-3}{2}\right)^2 - 3\left(\frac{y^2 - 6y +9}{4}\right)+ 7y^2-17y+13 = 3\left(x+\frac{y-3}{2}\right)^2 -\\ - \frac{1}{4}(3y^2-18+27) + \frac{1}{4}(28y^2-68y+52) = 3\left(x+\frac{y-3}{2}\right)^2 + \frac{1}{4}(25y^2-50y+25) =  3\left(x+\frac{y-3}{2}\right)^2 + \frac{25}{4}(y-1)^2 =\\= 3((x-1)+\frac{1}{2}(y-1))^2 +\frac{25}{4}(y-1)^2 = 3((x-1)^2 + (x-1)(y-1)+\frac{1}{4}(y-1)^2)+ \frac{25}{4}(y-1)^2 = 3(x-1)^2 +\\+ 3(x-1)(y-1)+7(y-1)^2$

Вектор мат. ожидания:

$\mathbb{E}\begin{pmatrix}
    \xi \\
    \eta
  \end{pmatrix}  = \begin{pmatrix}
    1 \\
    1
  \end{pmatrix}$

Обратная ковариационная матрица:

$\Sigma^{-1} = \begin{pmatrix}
  3& \frac{3}{2}\\ \frac{3}{2} & 7
  \end{pmatrix}
$

Поиск ковариационной матрицы:

$\mathrm{det}\Sigma^{-1} = 21 - \frac{9}{4} = \frac{75}{4}$

$\Sigma = \frac{1}{det\Sigma^{-1}}\cdot\Sigma^{-1} = \frac{4}{75}\begin{pmatrix}
 7 & -\frac{3}{2} \\ -\frac{3}{2} &3
 \end{pmatrix}$

Дисперсия $\xi$ и $\eta$:

$\mathrm{cov}(\xi,\xi) = \mathbb{D}(\xi) = 7 \cdot \frac{4}{75} = \frac{28}{75}$

$\mathrm{cov}(\eta,\eta) = \mathbb{D}(\eta) = 3 \cdot \frac{4}{75} = \frac{12}{75}$

Коеффициент корреляции:

$\mathrm{cov}(\xi,\eta) = -\frac{3}{2}\cdot \frac{4}{75} = -\frac{6}{75}$

$\rho(\xi,\eta) = \frac{\mathrm{cov}(\xi,\eta)}{\sqrt{\mathbb{D}(\xi)\cdot \mathbb{D}(\eta)}} = \frac{-\frac{6}{75}}{\sqrt{\frac{28}{75}\cdot\frac{12}{75}}} = -\frac{\sqrt{21}}{14}$

Коеффицинт $C$:

$\mathrm{C} = \frac{1}{(2\pi)^\frac{n}{2}\cdot \sqrt{\mathrm{det}(\Sigma)}}$, $n = 2$

$\mathrm{det}(\Sigma) = \left(\frac{4}{75}\right)^2 \cdot \left(21-\frac{9}{4}\right) = \frac{4}{75}$

$\mathrm{C} = \frac{1}{(2\pi)^\frac{2}{2}\cdot \sqrt{\frac{4}{75}}} = \frac{5\sqrt{3}}{4\pi}$

Плотность распределения:

$p_{\xi,\eta} = \frac{5\sqrt{3}}{4\pi} \cdot  \mathrm{exp}\left\{-\frac{1}{2}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 1 \\ 1
  \end{pmatrix}\right]^\intercal \begin{pmatrix}
  3& \frac{3}{2}\\ \frac{3}{2} & 7
  \end{pmatrix} \left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 1 \\ 1
  \end{pmatrix}\right]\right\}$

```python
import seaborn as sb
import matplotlib.pyplot as plot
import numpy as np

def density(x, y):
  ex = -(3 * x ** 2 + 3 * x * y + 7 * y ** 2 - 9 * x - 17 * y + 13) / 2
  return (75 ** 0.5) / 4*np.pi * np.exp(ex)
sb.set_style('whitegrid')
num = 25

ex_xi = 1
ex_eta = 1

eps = 2
x_plot1 = np.linspace(ex_xi - eps, ex_xi + eps, num)
y_plot1 = np.linspace(ex_eta - eps, ex_eta + eps, num)
plot1, plot2 = np.meshgrid(x_plot1, y_plot1)
plot3 = density(plot1, plot2)
axes = plot.axes(projection='3d')
axes.plot_surface(plot1, plot2, plot3)
plot.show()
```

###2. Нахождение афинного преобразования, переводящего исходный случайный вектор в стандартный нормальный.

$\begin{pmatrix} \xi_{st}\\ \eta_{st}\end{pmatrix} =\begin{pmatrix} \sqrt{3} &\frac{\sqrt{3}}{2}\\ 0 & \frac{5}{2}\end{pmatrix}\cdot \begin{pmatrix} \xi\\ \eta\end{pmatrix} + \begin{pmatrix} -\frac{3\sqrt{3}}{2}\\ -\frac{5}{2}\end{pmatrix} $

Стандартное нормальное распределение это распределение с параметрами:

$\mathbb{E}\begin{pmatrix} \xi_{st}\\ \eta_{st}\end{pmatrix} = \begin{pmatrix} 0\\ 0\end{pmatrix}$, $\Sigma_{st} = \begin{pmatrix} 1 & 0\\ 0 & 1\end{pmatrix}$

Проверка:

$\mathbb{E}\begin{pmatrix} \xi_{st}\\ \eta_{st}\end{pmatrix} = \begin{pmatrix} \sqrt{3} & \frac{\sqrt{3}}{2}\\0&\frac{5}{2}\end{pmatrix} \cdot \begin{pmatrix} 1\\ 1\end{pmatrix} + \begin{pmatrix} -\frac{3\sqrt{3}}{2}\\ -\frac{5}{2}\end{pmatrix} =\begin{pmatrix} \frac{3\sqrt{3}}{2}\\ \frac{5}{2}\end{pmatrix} +\begin{pmatrix} -\frac{3\sqrt{3}}{2}\\ -\frac{5}{2}\end{pmatrix} = \begin{pmatrix} 0\\ 0\end{pmatrix} $

$\Sigma_{st} = \frac{4}{75} \begin{pmatrix} \sqrt{3} & \frac{\sqrt{3}}{2}\\0&\frac{5}{2}\end{pmatrix} \cdot \begin{pmatrix} 7 & -\frac{3}{2}\\-\frac{3}{2}&3\end{pmatrix}\cdot \begin{pmatrix} \sqrt{3} & 0\\\frac{\sqrt{3}}{2}&\frac{5}{2}\end{pmatrix} = \frac{4}{75}\begin{pmatrix} 25\sqrt{3} & 0\\ -\frac{15}{4}&\frac{15}{2}\end{pmatrix} \cdot \begin{pmatrix} \sqrt{3} & 0\\\frac{\sqrt{3}}{2}&\frac{5}{2}\end{pmatrix} = \frac{4}{75}\begin{pmatrix} \frac{75}{4} & 0\\ 0&\frac{75}{4}\end{pmatrix}= \begin{pmatrix} 1 & 0\\ 0 & 1\end{pmatrix} $

###3. Нахождение ортогонального преобразования, переводящего соответсвующий центрированный случайный вектор в вектор с независимыми компонентами. Построение графика плотности полученного распределения.

$\Sigma^{-1} = \begin{pmatrix}3 &\frac{3}{2}\\ \frac{3}{2} & 7  \end{pmatrix}$

Поиск матрицы ортогональных преобразований $\mathrm{Q}^\intercal$:

$\begin{vmatrix}3 - \lambda & \frac{3}{2} \\ \frac{3}{2} & 7 - \lambda\end{vmatrix}=0$

$\lambda^2-10\lambda+\frac{75}{4}=0 \\
\lambda_1 = \frac{5}{2}, \lambda_2 = \frac{15}{2}$

$ \frac{1}{2}x + \frac{3}{2}y = 0,  y = \alpha , x = -3\alpha$

$\overrightarrow{v_1} = \begin{pmatrix} -3\alpha \\ \alpha \end{pmatrix} = \left[\alpha = 1 \right] = \begin{pmatrix} -3 \\ 1 \end{pmatrix}$

$\begin{Vmatrix} \overrightarrow{v_1} \end{Vmatrix} = \sqrt{(-3)^2 - 1^2} = \sqrt{10}$

$\overrightarrow{v^\prime_1} = \frac{1}{\sqrt{10}}\begin{pmatrix}-3\\ 1\end{pmatrix}$

$\begin{pmatrix} -\frac{9}{2} & \frac{3}{2} \\ \frac{3}{2} & -\frac{1}{2} \end{pmatrix} \sim \begin{pmatrix} \frac{3}{2} & -\frac{1}{2} \end{pmatrix} = 0$

$ \frac{3}{2}x - \frac{1}{2}y = 0,  x = \alpha , y = -3\alpha$

$\overrightarrow{v_2} = \begin{pmatrix} \alpha \\ 3\alpha \end{pmatrix} = \left[\alpha = 1 \right] = \begin{pmatrix} 1 \\ 3 \end{pmatrix}$

$\begin{Vmatrix} \overrightarrow{v_2} \end{Vmatrix} = \sqrt{(-3)^2 - 1^2} = \sqrt{10}$

$\overrightarrow{v^\prime_2} = \frac{1}{\sqrt{10}}\begin{pmatrix}1\\ 3\end{pmatrix}$

$\mathrm{Q}^\intercal = \frac{1}{\sqrt{10}}\begin{pmatrix} -3 & 1 \\ 1 & 3 \end{pmatrix}$

$\begin{pmatrix} \xi_{3}\\ \eta_{3}\end{pmatrix} = \frac{1}{\sqrt{10}}\begin{pmatrix} -3 & 1 \\ 1 & 3 \end{pmatrix} \cdot \begin{pmatrix} \xi\\ \eta\end{pmatrix}$


$\mathbb{E}\begin{pmatrix} \xi_{3}\\ \eta_{3}\end{pmatrix} = \frac{1}{\sqrt{10}}\begin{pmatrix} -3 & 1 \\ 1 & 3 \end{pmatrix} \cdot \begin{pmatrix} 1\\ 1\end{pmatrix} = \frac{1}{\sqrt{10}} \begin{pmatrix} -2\\ 4\end{pmatrix}$

$\Sigma_{3} = \mathrm{Q}\Sigma\mathrm{Q}^\intercal = \frac{1}{\sqrt{10}}\begin{pmatrix} -3 & 1 \\ 1 & 3 \end{pmatrix} \cdot  \frac{4}{75}\begin{pmatrix}
 7 & -\frac{3}{2} \\ -\frac{3}{2} &3
 \end{pmatrix} \cdot \frac{1}{\sqrt{10}}\begin{pmatrix} -3 & 1 \\ 1 & 3 \end{pmatrix} = \frac{4}{750}\begin{pmatrix} -\frac{45}{2} & \frac{15}{2} \\ \frac{5}{2} &\frac{15}{2} \end{pmatrix} \cdot \begin{pmatrix} -3 & 1 \\ 1 & 3 \end{pmatrix} = \frac{4}{750}\begin{pmatrix} 75 & 0 \\ 0 & 25 \end{pmatrix}=\\  =\begin{pmatrix} \frac{2}{5} & 0 \\ 0 & \frac{2}{15} \end{pmatrix} $

 $\mathrm{det}\Sigma_{3} = \frac{4}{75}$

 $\Sigma_{3}^{-1} = \begin{pmatrix} \frac{5}{2} & 0 \\ 0 & \frac{15}{2} \end{pmatrix}$

 $\mathrm{C} =\frac{5\sqrt{3}}{4\pi}$

 $p_{\xi,\eta} = \frac{5\sqrt{3}}{4\pi}\cdot  \mathrm{exp}\left\{-\frac{1}{2}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -\frac{2}{\sqrt{10}} \\ \frac{4}{\sqrt{10}} \end{pmatrix}\right]^\intercal \begin{pmatrix} \frac{5}{2} & 0 \\ 0 & \frac{15}{2} \end{pmatrix} \left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -\frac{2}{\sqrt{10}} \\ \frac{4}{\sqrt{10}} \end{pmatrix}\right]\right\}$


```python
import seaborn as sb
import matplotlib.pyplot as plot
import numpy as np

def density(x, y):
  ex = -(((5/2) * x + (2/10)**0.5)** 2 + (15/2)*(y-4/10**0.5)**2) / 2
  return (75 ** 0.5) / 4*np.pi * np.exp(ex)
sb.set_style('whitegrid')
num = 25

ex_xi = -(2/10)**0.5
ex_eta = (4/10)**0.5

eps = 2
x_plot1 = np.linspace(ex_xi - eps, ex_xi + eps, num)
y_plot1 = np.linspace(ex_eta - eps, ex_eta + eps, num)
plot1, plot2 = np.meshgrid(x_plot1, y_plot1)
plot3 = density(plot1, plot2)
axes = plot.axes(projection='3d')
axes.plot_surface(plot1, plot2, plot3)
plot.show()
```

###4. Вычислить характеристики совместного распределения случайного ветора $(4\xi -\eta, -4\xi -4\eta)^\intercal$ и записать его плотность. Построить график плотности полученного рапределения.



$\begin{pmatrix} -4\xi & -\eta \\ -4\xi & -4\eta \end{pmatrix} =\begin{pmatrix} -4 & -1 \\ -4 & -4 \end{pmatrix} \cdot \begin{pmatrix} \xi \\ \eta \end{pmatrix}$

$\mathrm{A} = \begin{pmatrix} -4 & -1 \\ -4 & -4 \end{pmatrix}, \mathrm{detA} = -20 $, матрица $\mathrm{A}$ неворожденая.

$\mathbb{E}\begin{pmatrix} \xi_4 \\ \eta_4 \end{pmatrix} = \begin{pmatrix} -4 & -1 \\ -4 & -4 \end{pmatrix} \cdot \begin{pmatrix} 1 \\ 1 \end{pmatrix} = \begin{pmatrix} 3 \\ -8 \end{pmatrix}$

$\Sigma_4 = \frac{4}{75}\begin{pmatrix} -4 & -1 \\ -4 & -4 \end{pmatrix} \cdot \begin{pmatrix} 7 & -\frac{3}{2} \\ -\frac{3}{2} &3 \end{pmatrix} \cdot \begin{pmatrix} -4 & -4 \\ -1 & -4 \end{pmatrix} = \frac{4}{75}\begin{pmatrix} \frac{59}{2} & -9 \\ -22 & -6 \end{pmatrix}\cdot \begin{pmatrix} -4 & -4 \\ -1 & -4 \end{pmatrix} = \frac{4}{75}\begin{pmatrix}127 & -82 \\ -82 & 112 \end{pmatrix}$

$\mathrm{det\Sigma_4} =\left(\frac{4}{75} \right)^2 \begin{vmatrix}112 & 82 \\ 82 & 127 \end{vmatrix} = \frac{64}{3}$

$\Sigma_4^{-1} = \frac{4}{75} \cdot \frac{3}{64}\begin{pmatrix}112 & 82 \\ 82 & 127 \end{pmatrix}  = \frac{1}{400}\begin{pmatrix}112 & 82 \\ 82 & 127 \end{pmatrix} $

$\mathrm{cov}(\xi,\xi) = \mathbb{D}(\xi) = 127 \cdot \frac{4}{75} = \frac{508}{75}$

$\mathrm{cov}(\eta,\eta) = \mathbb{D}(\eta) = 112 \cdot \frac{4}{75} = \frac{448}{75}$

$\mathrm{cov}(\xi,\eta) = -82\cdot \frac{4}{75} = -\frac{328}{75}$

$\rho(\xi,\eta) = \frac{\mathrm{cov}(\xi,\eta)}{\sqrt{\mathbb{D}(\xi)\cdot \mathbb{D}(\eta)}} = \frac{-\frac{328}{75}}{\sqrt{\frac{508}{75}\cdot\frac{448}{75}}}= -\frac{-412\sqrt{889}}{1778}$

$\mathrm{C} = \frac{1}{(2\pi)\sqrt{\frac{63}{4}}} = \frac{\sqrt{3}}{16\pi}$

$p_{\xi,\eta} = \frac{\sqrt{3}}{16\pi}\cdot  \mathrm{exp}\left\{-\frac{1}{2}\cdot \frac{1}{400}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 3 \\ -8 \end{pmatrix}\right]^\intercal \begin{pmatrix} 112& 82\\82 & 127 \end{pmatrix} \left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 3 \\ -8\end{pmatrix}\right]\right\}$


```python
import seaborn as sb
import matplotlib.pyplot as plot
import numpy as np

def density(x, y):
  ex = -(((112/400) * x -3 )** 2 + (82/400)*(x-3)*(y+8) +(127/400)*(y+8)**2) / 2
  return (3 ** 0.5) / 16*np.pi * np.exp(ex)
sb.set_style('whitegrid')
num = 25

ex_xi = 3
ex_eta = -8

eps = 30
x_plot1 = np.linspace(ex_xi - eps, ex_xi + eps, num)
y_plot1 = np.linspace(ex_eta - eps, ex_eta + eps, num)
plot1, plot2 = np.meshgrid(x_plot1, y_plot1)
plot3 = density(plot1, plot2)
axes = plot.axes(projection='3d')
axes.plot_surface(plot1, plot2, plot3)
plot.show()
```

###5 Найти условное распределение  $\xi$ при условии $\eta$. Вычислить $\mathbb{E}(\xi|\eta)$ и $\mathbb{D}(\xi|\eta)$

$\mathbb{E}(\eta) = 1 ,\mathbb{D}(\eta) = \frac{4}{25}$

$\mathrm{C} = \frac{1}{\sqrt{2\pi}\cdot\sqrt{\mathbb{D}}} = \frac{5}{2\sqrt{2\pi}}$

$p_{\eta} = \frac{5}{2\sqrt{2\pi}} \mathrm{exp}\left(-\frac{1}{2}\left(\frac{y-\mathbb{E}(y)}{\mathbb{D}(y)}\right)\right) = \frac{5}{2\sqrt{2\pi}} \mathrm{exp}\left(-\frac{1}{2}\left(\frac{25(y-1)^2}{4}\right)\right) = \frac{5}{2\sqrt{2\pi}} \mathrm{exp}\left(-\frac{1}{2}\left(\frac{25}{4}y^2-\frac{25}{2}y + \frac{25}{4}\right)\right)$

$p_{\xi|\eta=y}(x) = \frac{p_{\xi,\eta}(x,y)}{p_{\eta}(y)} =  \frac{\frac{5\sqrt{3}}{4\pi}\mathrm{exp}\left(\frac{-1}{2}\left(3x^2 + 3xy+7y^2-9x-17y+13\right)\right)}{\frac{5}{2\sqrt{2\pi}} \mathrm{exp}\left(-\frac{1}{2}\left(\frac{25}{4}y^2-\frac{25}{2}y + \frac{25}{4}\right)\right)} = \frac{\sqrt{6\pi}}{2\pi} \mathrm{exp}\left(-\frac{1}{2}\left(3x^2+3xy+\frac{3}{4}y^2 - 9x - \frac{9}{2}y + \frac{27}{4}\right)\right) = \\
= \frac{\sqrt{6\pi}}{2\pi} \mathrm{exp}\left(-\frac{1}{2}\cdot 3 \left(x^2+xy+\frac{1}{4}y^2 - 3x - \frac{3}{2}y + \frac{9}{4}\right)\right) = \frac{\sqrt{6\pi}}{2\pi} \mathrm{exp}\left(-\frac{1}{2}\cdot 3 \left(\left(x+\frac{y-3}{2}\right)^2 - \frac{1}{4}\left(y-3\right) + \frac{1}{4}\left(y^2-6y+9\right)\right)\right) = \\ =\frac{\sqrt{6\pi}}{2\pi} \mathrm{exp}\left(-\frac{1}{2}\cdot 3 \left(\left(x+\frac{y-3}{2}\right)^2\right)\right)$

$\mathbb{E}(\xi|\eta=y) = -\frac{y-3}{2} $  
$\mathbb{D}(\xi|\eta=y) = 3$
