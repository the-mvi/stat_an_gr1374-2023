## Основные теоретические положения

Опр:  Случайный вектор $ \vec \xi = (\xi_1 , \xi_2 , ..., \xi_n): \Omega \to R^n$ называют нормальным (гауссовым), если существует вектор независимых стандартных нормальных случайных величин  $ \vec \eta = (\eta_1 , \eta_2 , ..., \eta_m)^{\intercal} $, матрица А размера n*m и вещественный вектор $ \vec \beta = (\beta_1 , \beta_2 , ..., \eta_n)^{\intercal} $, такие что $ \vec \xi = А \vec \eta + \vec \beta $ .

Обозначение $ \vec \xi $\~${N( \mu , \Sigma )}$$ $ , где  $ \vec \mu =E \vec \xi = (E\xi_1 , E\xi_2 , ..., E\xi_n)^{\intercal} $- вектор математических ожиданий компонент, $ \Sigma = {{ \{ \delta_i{}_j} \} {}_i{}_j{}}^n =1 $ , $ \delta_i{}_j = cov(\xi_i, \xi_j) $ - ковариационная матрица.

Плотность нормального вектора:

$ f_\vec \xi  (x_1, ..., x_n)=  \frac1{ (2 \pi){}^ \frac{n}2  \sqrt{ det \Sigma }  } exp (- \frac1{2} ( \vec x - \mu)^{\intercal} \Sigma^{-1} ( \vec x - \mu )) $

Свойства нормального вектора:
1) Линейные комбинации случайных переменных, принадлежащих нормальному распределению, тоже нормально распределенные.
2) Условные и безусловные распределения, полученные из многомерного нормального - тоже нормальные.
3) Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму.
4) Многомерное нормальное распределение устойчиво относительно линейного преобразования: если $ \vec \xi $\~${N( \mu , \Sigma )}$$ $ , $ A_{m \cdot n} $ - произвольная матрица, $ B_{m \cdot 1} $ - произвольный вектор, то $ А \vec \xi + \vec \beta $\~${N(A \mu + B ,A \Sigma A^{\intercal} )}$$  $ .

1. Преобразование Лагранжа
   
$ q(x,y)= 3x^2 + 4xy + 6y^2 + 10x - 12y + 27 = (3x^2 + 4xy + 10x) + 6y^2 - 12y + 27 = $
$ = (3x^2 + 2x (2y + 5)) + 6y^2 - 12y + 27 =  3(x^2 + 2x \frac{2y + 5}3 ) + 6y^2 - 12y + 27 = $
$ 3(x^2 + 2x \frac{2y + 5}3 + \left( \frac{2y + 5}3 \right) {}^2 ) - 3 \left( \frac{2y + 5}3 \right) {}^2 + 6y^2 - 12y + 27 =$
$ = 3(x + \frac{2y}3 + \frac{5}3  ){}^2 -  \frac{(2y + 5){}^2}3  + 6y^2 - 12y + 27 = 3(x + \frac{2y}3 + \frac{5}3  ){}^2 + \frac{14}3 (y - 2){}^2= $
$ =3( x + \frac{2}3 (y-2) + 3){}^2 + \frac{14}3 (y - 2){}^2= 3( (x + 3) + \frac{2}3 (y-2)){}^2 + \frac{14}3 (y - 2){}^2= 3 (x+3)^2 + 2 \cdot 2(x+3)(y-2) +6 (y-2)^2 $

$ E \left( \begin{matrix} \xi \\ \eta \end{matrix} \right) = \left( \begin{matrix} -3 \\ 2 \end{matrix} \right) $

$ \Sigma^{-1} = \left( \begin{matrix} 3 & 2 \\ 2 & 6 \end{matrix} \right) $

$ \Sigma = \frac1{det \Sigma^{-1} } \left( \begin{matrix} 6 & -2 \\ -2 & 3 \end{matrix} \right) =\frac1{14} \left( \begin{matrix} 6 & -2 \\ -2 & 3 \end{matrix} \right) $

$ D \xi = \frac3{7} ,   D \eta = \frac3{14} ,   cov( \xi , \eta) = - \frac1{7} ,   \rho ( \xi , \eta) = \frac{cov( \xi , \mu) }{ \sqrt{D \xi D \mu}} = - \frac{ \sqrt{2} }3 $

$ C= \frac1{2 \pi \sqrt{ \frac1{14}}} = \frac{ \sqrt{14}}{2 \pi} $

$ f_\xi , {}_\eta (x,y) = \frac{ \sqrt{14}}{2 \pi}exp \left(- \frac1{2} \left( \vec x - \left( \begin{matrix} -3 \\ 2 \end{matrix} \right) \right)^{\intercal} \left( \begin{matrix} 3 & 2 \\ 2 & 6 \end{matrix} \right) \left( \vec x - \left( \begin{matrix} -3 \\ 2 \end{matrix} \right) \right)\right ) $

$ f_\xi , {}_\eta (x,y) = \frac{ \sqrt{14}}{2 \pi}exp(- \frac1{2} (3x^2 + 4xy + 6y^2 + 10x - 12y + 27)) $

```r
#задаем оси
x = seq(from=-7, to=2, by=0.05)
y = seq(from=-15, to=15, by=0.05)

#задаем функцию плотности
z_values = function (x, y) {
 (sqrt(14)/(2*pi))*exp((-1/2)*(3*x^2+4*x*y+6*y^2+10*x-12*y+27))
}

#создаем ось плотности
z = outer(x, y, z_values)

#рисуем график
persp(x, y, z,
 main='3D Plot', col='pink' )
```
