## Основные теоретические положения

Двумерное нормальное распределение вероятностей:

$f(x,y) = \frac{1}{2\pi\cdot \sigma_{x}\cdot \sigma_{y}\cdot \sqrt{1 - \rho^{2}}}\cdot exp^{-\frac{1}{2\cdot(1-\rho^{2})}\cdot \left[ \frac{(x - \mathbb{E}_{x})^{2}}{\sigma_{x}^{2}} - \frac{2\rho\cdot(x-\mathbb{E}_{x})\cdot(y-\mathbb{E}_{y})}{\sigma_{x}\cdot \sigma_{y}} + \frac{(y-\mathbb{E}_{y})^{2}}{\sigma_{y}^{2}} \right]}$

$\rho$ - коэффициент коррелции случайно величины X и Y

$\sigma_{x}$ - среднее квадратическое отклонение случайной величины X

$\sigma_{y}$ - среднее квадратическое отклонение случайной величины Y

$\mathbb{E}x$ - математическое ожидание случайной величины X

$\mathbb{E}y$ - математическое ожидание случайной величины Y

1. Чтобы вычислить вектор математических ожиданий воспользуемся преобразовением Лагранжа, в результате которого мы получим запись следующего вида: $q(x,y) = a\cdot(x - \alpha) +2c\cdot(x - \alpha)\cdot(y - \beta) + b\cdot(y - \beta)$ 

Для нахождения дисперсий ковариации нам поможет Ковариационная матрица $\Sigma$: $\Sigma = \begin{pmatrix} \mathbb{D}\xi & \mathop{\mathrm{cov}}\nolimits(\xi, \eta) \\ \mathop{\mathrm{cov}}\nolimits(\eta, \xi) & \mathbb{D}\eta \end{pmatrix}$

Кроме таких числовых характеристик,как $\mathbb{E}\xi, \mathbb{E}\eta, \mathbb{D}\xi, \mathbb{D}\eta$ и $ \mathop{\mathrm{cov}}\nolimits(\xi, \eta)$ , мы можем найти коэффициент корреляции $\rho$ и СКО $\sigma_{\xi}, \sigma_{\eta}$.

$\sigma_{\xi} = \sqrt{\mathbb{D_{\xi}}}$   

$\rho_{\xi\eta} = \frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta) }{\sigma_{\xi}\cdot\sigma_{\eta}}$

$C = \frac{1}{2\pi\cdot\sqrt{\det(\Sigma)}}$

2. Новый случайный вектор  $\mathbf{Y}$ можно определить, применяя аффинное преобразование g:  $\mathbb{R}^{n} \to \mathbb{R}^{n}$ к случайному вектору  $\mathbf{X}$:

$\textbf{Y} = A\cdot\textbf{X} + B$, где A является матрицей коэффициентов при $\xi, \eta$ размера $n\cdot n$, матрица B - матрица свободных членов размера $n\cdot 1$.

Или можно представить в таком виде: $A\cdot\overrightarrow{\xi} + B\sim \mathcal{N(o,\mathbb{E})}$. Для достижения цели нормируем q(x,y) и проведем проверки: $\mathbb{E}\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = A\cdot\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} + B = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$ и $\Sigma_{st} = A\cdot\Sigma\cdot A^{\intercal} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$

3. Для ортогонального преобразования, переводящий соответствующий центрированный случайны вектор в вектор с независимыми компонентами нам нужно решить следующее уравнение $\lambda^{2} - tr(A)\cdot\lambda + \det(A) = 0$

Находим ковариационные числа $\left\{ \mathrm{\lambda}_{1},...,\mathrm{\lambda}_{k} \right\}$ и соответствующие им собственные вектора $\left\{ \mathrm{V}_{1},...,\mathrm{V}_{k} \right\}$ ковариционной матрицы $\Sigma$ и нормируем вектора, получая ортонормированный базис

Собственные векторы, отвечающие различным собственным значениям, ортогональны между собой. Сформировав матрицу Q из нормированных собственных векторов, остается перемножить матрицу на центрированный вектор, чтобы получить вектор с независимыми компонентами.

Поскольку мы получаем вектор с независимыми переменнами, то по свойству ковариации: Если $\xi,\eta$ - независимые переменные, то $\mathop{\mathrm{cov}}\nolimits(\xi, \eta) = 0$

4. При нахождении условного распределения $\xi$ при условии $\eta$ воспользуемся формулой условного распределения $P_{(\xi|\eta=y)} = \frac{P_{\xi,\eta}(x,y)}{P_{\eta}(y)}$.

Условное распределение полученные из многомерного нормального распределения, также является многомерным распределением. Для одномерного нормального распределения ковариционная матрица $\Sigma$ будет состоять только из дисперсии.

### 1.Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора.

Произведем преобразование Лагранжа
$q(x,y) = 3x^2 + 3xy + 7y^2 - 12x + 19y + 37 = (3x^2 + 3xy - 12x) + 7y^2 + 19y + 37 = 3(x^2 + xy - 4x) + 7y^2 + 19y + 37 = 3(x^2 + x(y-4)) + 7y^2 + 19y + 37 = 3(x^2 + \frac{2}{2}x(y-4) + (\frac{y-4}{2})^2) -3(\frac{y-4}{2})^2 + 7y^2 + 19y + 37 = 3(x+\frac{1}{2}(y-4))^2 + \frac{25}{4}y^2 + 25y + 25 = 3(x+\frac{1}{2}(y-4))^2 + \frac{25}{4}(y^2 + 4y + 4) = 3(x+\frac{1}{2}(y-4))^2 + \frac{25}{4}(y+2)^2 = 3((x-3) + \frac{1}{2}(y-2))^2 + \frac{25}{4}(y+2)^2 = 3(x-3)^2 + 3(x-3)(y+2) + 7(y+2)^2$

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 3 \\ -2\end{pmatrix}$ 
Нашли вектор математических ожиданий

$\Sigma^{-1} = \begin{pmatrix} 3 & \frac{3}{2} \\ \frac{3}{2} & 7 \end{pmatrix} \implies \det{\Sigma^{-1}} = \frac{75}{4}$

Найдем остальные числовые характеристики данного случайного вектора

$\Sigma = \begin{pmatrix} 7 & -\frac{3}{2} \\ -\frac{3}{2} & 3 \end{pmatrix} \cdot \frac{1}{\det{\Sigma^{-1}}} = \begin{pmatrix} \frac{28}{75} & -\frac{2}{25} \\ -\frac{2}{25} & \frac{4}{25}\end{pmatrix} \implies \mathbb{D}\xi = \frac{28}{75}, \mathbb{D}\eta = \frac{4}{25}, \mathop{\mathrm{cov}}\nolimits(\xi, \eta) = -\frac{2}{25}$

$\sigma_{\xi} = \sqrt{\frac{28}{25}} = \frac{2\sqrt{7}}{5}$

$\sigma_{\eta} = \sqrt{\frac{4}{25}} = \frac{2}{5}$

$\rho_{\xi,\eta} = \frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta) }{\sigma_{\xi}\cdot\sigma_{\eta}} = \frac{-\frac{2}{25}}{\frac{2\sqrt{7}}{5}\cdot\frac{2}{5}} = -\frac{\sqrt{7}}{14}$

$C = \frac{1}{2\pi\cdot\sqrt{\det(\Sigma)}} = \frac{5\sqrt{3}}{4\pi}$

Плотность распределения: $P_{\xi\eta}(x,y) = \frac{5\sqrt{3}}{4\pi}exp\left\{ -\frac{1}{2}\cdot\left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 3 \\ -2 \end{pmatrix}\right]^{\intercal} \cdot \begin{pmatrix} 3 & \frac{3}{2} \\ \frac{3}{2} & 7 \end{pmatrix}  \cdot \left[ \begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} 3 \\ -2 \end{pmatrix} \right] \right\}$

```python
import numpy as np
import matplotlib.pyplot as plt

def normal_distribution(x, y):
    constant = 5 * np.sqrt(3) / (4 * np.pi)
    mean = np.array([3, -2])
    covariance = np.array([[3, 3/2],
                           [3/2, 7]])
    diff = np.stack((x, y), axis=-1) - mean
    power = -0.5 * np.einsum('...k,kl,...l->...', diff, covariance, diff)
    return constant * np.exp(power)

x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)
X, Y = np.meshgrid(x, y)

Z = normal_distribution(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

### 3.Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами.

$\begin{vmatrix} 3-\lambda & \frac{3}{2}\\ \frac{3}{2} & 7-\lambda \end{vmatrix} = (3-\lambda)\cdot(7-\lambda) - \frac{9}{4} = \lambda^{2} - 10\lambda + \frac{75}{4} = \left( \lambda-\frac{5}{2} \right)\cdot\left( \lambda-\frac{15}{2} \right)$

$\lambda_{2} = \frac{15}{2}$

$\left[\begin{array}{@{} rr|r @{\,}} -\frac{9}{2} & \frac{3}{2} & 0 \\ \frac{3}{2} &  -\frac{1}{2} &  0 \end{array}\right]\sim \left[\begin{array}{@{} rr|r @{\,}} 0 &  0 & 0 \\ -\frac{3}{2} & \frac{1}{2} &  0 \end{array}\right]$

$-\frac{3}{2}x + \frac{1}{2}y = 0 \to 3x = y \to \overrightarrow{\mathrm{V}_{2}} = \begin{pmatrix} 1 \\ 3 \end{pmatrix}$

$\left\| \overrightarrow{\mathrm{V}_{2}} \right\| = \sqrt{3^{2}+1} = \sqrt{10} \to \overrightarrow{\mathrm{V}_{2}^{'}} = \begin{pmatrix} \frac{1}{\sqrt{10}} \\ \frac{3}{\sqrt{10}} \end{pmatrix}$

$Q^{\intercal} = \begin{pmatrix} -\frac{3}{\sqrt{10}} & \frac{1}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} & \frac{3}{\sqrt{10}} \end{pmatrix}$  

$Q = \begin{pmatrix} -\frac{3}{\sqrt{10}} & \frac{1}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} & \frac{3}{\sqrt{10}} \end{pmatrix}$

$\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} = \begin{pmatrix} -\frac{3}{\sqrt{10}} & \frac{1}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} & \frac{3}{\sqrt{10}} \end{pmatrix}\cdot\begin{pmatrix} \xi \\ \eta \end{pmatrix}$

Найдем математические ожидания и другие числовые характеристики и запишем плотность полученнного распределния

$\mathbb{E}\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} = \begin{pmatrix} -\frac{3}{\sqrt{10}} & \frac{1}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} & \frac{3}{\sqrt{10}} \end{pmatrix}\cdot\begin{pmatrix} 3 \\ -2 \end{pmatrix} = \begin{pmatrix} -\frac{11}{\sqrt{10}} \\ -\frac{3}{\sqrt{10}} \end{pmatrix}$

$\Sigma_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}} = Q*\Sigma_{\begin{pmatrix} \xi \\ \eta \end{pmatrix}}\cdot Q^{\intercal} = \frac{4}{75}\cdot\begin{pmatrix} -\frac{3}{\sqrt{10}} & \frac{1}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} & \frac{3}{\sqrt{10}} \end{pmatrix}\cdot\begin{pmatrix} 7 & -\frac{3}{2} \\ -\frac{3}{2} & 3 \end{pmatrix}\cdot\begin{pmatrix} -\frac{3}{\sqrt{10}} & \frac{1}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} & \frac{3}{\sqrt{10}} \end{pmatrix} = \begin{pmatrix} \frac{2}{5} & 0 \\ 0 & \frac{2}{15} \end{pmatrix} = \begin{pmatrix} \frac{1}{\lambda_{1}} & 0 \\ 0 & \frac{1}{\lambda_{2}} \end{pmatrix}$

$\Sigma_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}}^{-1} = \frac{1}{{\det\Sigma_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}}}}\cdot\Sigma_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}} = \frac{1}{\frac{4}{75}}\cdot\begin{pmatrix} \frac{2}{5} & 0 \\ 0 & \frac{2}{15} \end{pmatrix} = \begin{pmatrix} \frac{5}{2} & 0 \\ 0 & \frac{15}{2} \end{pmatrix}$

$\mathbb{D}_{\xi} = \frac{2}{5}; \mathbb{D}_{\eta} = \frac{2}{15}; \mathop{\mathrm{cov}}\nolimits(\xi, \eta) = 0 \Rightarrow \rho_{\xi,\eta} = 0$

$C = \frac{1}{2\pi\cdot\frac{2}{5\sqrt{3}}} = \frac{5\sqrt{3}}{4\pi}$

Плотность полученного распределения: $P_{\xi,\eta}(x,y)= \frac{5\sqrt{3}}{4\pi}\cdot exp\left\{ -\frac{1}{2}\cdot\left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -\frac{11}{\sqrt{10}} \\ -\frac{3}{\sqrt{10}} \end{pmatrix}\right]^{\intercal}\cdot\begin{pmatrix} \frac{5}{2} & 0 \\ 0 & \frac{15}{2} \end{pmatrix} \cdot\left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -\frac{11}{\sqrt{10}} \\ -\frac{3}{\sqrt{10}} \end{pmatrix} \right] \right\}$

```python
import numpy as np
import matplotlib.pyplot as plt

def normal_distribution(x, y):
    constant = 5 * np.sqrt(3) / (4 * np.pi)
    mean = np.array([-11 / np.sqrt(10), -3 / np.sqrt(10)])
    covariance = np.array([[5/2, 0],
                           [0, 15/2]])
    diff = np.stack((x, y), axis=-1) - mean
    power = -0.5 * np.einsum('...k,kl,...l->...', diff, covariance, diff)
    return constant * np.exp(power)

x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)
X, Y = np.meshgrid(x, y)

Z = normal_distribution(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

### 5.Найти условное распределение $\xi$  при условии $\eta$.Вычислить $\mathbb{E}(\xi|\eta)$  и $\mathbb{D}(\xi|\eta)$.

Возьмем данные из пункта 1. Поскольку для одномерного случая матрица $\Sigma_{\eta}$ состоит из $\mathbb{D}_{\eta}$, то получим $\Sigma_{\eta}=\mathbb{D}_{\eta}$ 

$P_{\xi,\eta}(x,y)=C*exp\left( -\frac{1}{2}(3x^2 + 3xy + 7y^2 - 12x + 19y + 37) \right)$

$\mathbb{E}_{\eta} = -2;$

$\mathbb{D}_{\eta} = \frac{4}{25};$

$\Sigma_{\eta}=\frac{4}{25}\to \Sigma_{\eta}^{-1} = \frac{25}{4}$;Подставив в формулу  $P_{\eta}(y) = \frac{1}{\sqrt{2\pi}\cdot\sqrt{\det(\Sigma)}}\cdot exp^{-\frac{1}{2}\cdot\left( (y-\mathbb{E}_{\eta})\cdot\Sigma_{\eta}^{-1}\cdot(y-\mathbb{E}_{\eta}) \right)}$ все наши параметры, мы получим плотность распределения $P_{\eta}$:  $P_{\eta}(y) = \frac{5\sqrt{6}}{4\pi}\cdot exp(-\frac{1}{2}\cdot(y+2)\cdot\frac{25}{4}\cdot(y+2))$
