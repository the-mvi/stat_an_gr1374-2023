## Основные теоретические положения
Двумерная случайная величина $\left( x,y \right)$ называется распределенной по нормальному закону, если ее совместная плотность распределения имеет вид:

$f(x,y) =\frac{1} {2\pi\cdot\sigma_{x}\cdot \sigma_{y}\cdot \sqrt{1 - \rho^{2}}}\cdot\exp^{-\frac{1}{2\cdot(1-\rho^{2})}\cdot\left[ \frac{(x - \mathbb{E}_{x})^{2}}{\sigma_{x}^{2}}-\frac{2\rho\cdot(x-\mathbb{E}_{x})\cdot(y-\mathbb{E}_{y})}{\sigma_{x}\cdot \sigma_{y}}+\frac{(y-\mathbb{E}_{y})^{2}}{\sigma_{y}^{2}} \right]}$

 Где:

 $\sigma_{x}$ - СКО случайной величины $x$

 $\sigma_{y}$ - СКО случайной величины $y$

 $\rho$ - коэффициент корелляции

$\mathbb{E}_{x}$ - мат. ожидание $x$

$\mathbb{E}_{y}$ - мат. ожидание $y$

1.Ковариационные характеристики данного случайого вектора можно найти из матрицы ковариаций:

 $\Sigma = \begin{pmatrix} \mathbb{D}\xi & \mathop{\mathrm{cov}}\nolimits(\xi, \eta) \\ \mathop{\mathrm{cov}}\nolimits(\eta, \xi) & \mathbb{D}\eta \end{pmatrix}$

Здесь

$\mathbb{D}\xi$ - дисперсия $\xi$

$\mathbb{D}\eta$ - дисперсия $\eta$

$\mathop{\mathrm{cov}}\nolimits(\xi, \eta)$ - ковариация $\xi, \eta$

Данная матрица находится из формулы номарльного распределения путем преобразования степени экспоненты к квадратичной форме путем преобразований Лагранжа

$\Sigma_{st} = A\cdot\Sigma\cdot A^{\intercal} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$

### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $P_{\xi,\eta}$.


Произведем преобразование Лагранжа

$q(x,y) = 3x^{2}+3xy+7y^{2}-3x-11y+7 = 3\left( x + \frac{1}{2}(y - 1)\right)^2 - \frac{3}{4}\left(y - 1\right)^2 + \frac{1}{4} \left(7y^2 + 11y + 28\right) = \frac{3}{4}\left(2x + y - 1\right)^2 + \frac{(25y^2 + 50y + 25)}{4} = \frac{3}{4}\left( (2x - 2) + (y + 1)\right)^2 + \frac{25}{4}\left(y + 1\right)^2 = \frac{3}{4}\left((2x - 2)^2 + 2(2x-2)(y+1) + (y + 1)^2\right) + \frac{25}{4}\left(y + 1\right)^2 = 3(x-1)^2 + 3(x-1)(y+1) + 7(y+1)^2$

Отсюда:

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 1 \\ -1 \end{pmatrix}$

$\Sigma^{-1} = \begin{pmatrix} 3 & \frac{3}{2} \\ \frac{3}{2} & 7 \end{pmatrix} ⇒ \Sigma = \frac{4}{75}\begin{pmatrix} 7 & -\frac{3}{2} \\ -\frac{3}{2} & 3 \end{pmatrix}$

$\mathbb{D}\xi = \frac{28}{75} \\ \mathbb{D}\eta = \frac{4}{25}\\ \mathop{\mathrm{cov}}\nolimits(\xi, \eta) = -\frac{2}{25}$

$\sigma_{\xi} = \sqrt{\frac{28}{75}} = \frac{2\sqrt{21}}{15}$

$\sigma_{\eta} = \sqrt{\frac{4}{25}} = \frac{2}{5}$


$\rho_{\xi,\eta} = \frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta) }{\sigma_{\xi}\cdot\sigma_{\eta}} = -\frac{\frac{2}{25}}{\frac{2\sqrt{7}}{5\sqrt3}\cdot\frac{2}{5}} = \frac{\sqrt{21}}{14}$

$C = \frac{1}{2\pi\cdot\sqrt{\det(\Sigma)}} = \frac{5\sqrt{3}}{4\pi}$

Запишем приведенную формулу плотности распределения:

$P_{\xi\eta}(x,y) = \frac{5\sqrt{3}}{4\pi}exp\left\{ -\frac{1}{2}\cdot\left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 1 \\ -1 \end{pmatrix}\right]^{\intercal} \cdot \begin{pmatrix} 3 & \frac{3}{2} \\ \frac{3}{2} & 7 \end{pmatrix}  \cdot \left[ \begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} 1 \\ -1 \end{pmatrix} \right] \right\}$

Построим график полученного распределения:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def N(x, y):
    C = 5 * np.sqrt(3) / (4 * np.pi)
    E = np.array([1, -1])
    Sig = np.array([[3, 3/2],[3/2, 7]])
    A =  np.stack((x, y), axis=-1) - E
    deg = -0.5 * np.einsum('...k,kl,...l->...', A, Sig, A)
    return C * np.exp(deg)

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = N(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.xlim(-5,5)
plt.ylim(-5,5)

plt.show()
```
