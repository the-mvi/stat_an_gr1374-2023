## –û—Å–Ω–æ–≤–Ω—ã–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–ª–æ–∂–µ–Ω–∏—è

### –õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–∞—è –º–æ–¥–µ–ª—å

–õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –Ω–µ–∫–æ—Ç–æ—Ä–æ–π –∑–∞–≤–∏—Å–∏–º–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π y –Ω–∞ –Ω–∞–±–æ—Ä –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö  $x = (x_1, ‚Ä¶, x_n)$, –≥–¥–µ n ‚Äì —ç—Ç–æ —á–∏—Å–ª–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π, –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç, —á—Ç–æ –ª–∏–Ω–µ–π–Ω–æ–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –º–µ–∂–¥—É y –∏ x: $y =\beta_0  + \beta_1x_1 + ‚ãØ + \beta_nx_n + ùúÄ$ - —É—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. $\beta_0, \beta_1, ‚Ä¶, \beta_n$ ‚Äì –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ (–ø–∞—Ä–∞–º–µ—Ç—Ä—ã), –∏ ùúÄ ‚Äì —Å–ª—É—á–∞–π–Ω–∞—è –æ—à–∏–±–∫–∞.

### –ú–ù–ö (–ú–µ—Ç–æ–¥ –ù–∞–∏–º–µ–Ω—å—à–∏—Ö –ö–≤–∞–¥—Ä–∞—Ç–æ–≤)

$RSS(\beta)=\left\| Y-X^{\intercal}\beta \right\|=(Y-X^{\intercal}\beta)^{\intercal}\cdot(Y-X^{\intercal}\beta)\to min\beta$

–ù–æ—Ä–º–∞–ª—å–Ω—ã–µ —É—Ä–∞–≤–Ω–µ–Ω–∏—è: $XX^{\intercal}\beta=XY$

–†–µ—à–µ–Ω–∏–µ —Å–∏—Å—Ç–µ–º—ã –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π –≤ —Ä–µ–≥—É–ª—è—Ä–Ω–æ–º —Å–ª—É—á–∞–µ: $\widehat{\beta}=(XX^{\intercal})^{-1}XY$

$RSS(\beta)=\sum_{i=1}^{n}\left( Y_{i}-\sum_{j=1}^{m}x_{ji}\beta_{j} \right)^{2}$

$\frac{\partial RSS}{\partial \beta_{k}} = -2\sum_{i=1}^{n}x_{ki}\left(Y_{i}- \sum_{j=1}^{m}x_{ji}\beta_{j}\right)=0, k=\overline{1,m}$

### –ù–µ—Å–º–µ—â—ë–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ $(S^{2})$

$S^{2} = \frac{SS_{e}}{n-r} = \frac{RSS(\widehat{\beta})}{n-r}$

$SS_{e} = RSS(\widehat{\beta})$, –∞ r-—Ä–∞–Ω–≥ –º–∞—Ç—Ä–∏—Ü—ã X

### –î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª

–î–ù–û —Ñ—É–Ω–∫—Ü–∏–∏:$\psi=—Å^{\intercal}\beta$

–ú–ù–ö-–æ—Ü–µ–Ω–∫–∞: $\widehat{\psi}=—Å^{\intercal}\beta$

–≠–ª–µ–º–µ–Ω—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –æ—Ü–µ–Ω–∫–∏: $b_{\psi}=c^{\intercal}(XX^{\intercal})^{-1}c$

–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ $\widehat{\psi}\sim \mathcal{N}(\psi,\sigma^{2}b)$

$\frac{\widehat{\psi}-\psi}{S\sqrt{b}}\sim \mathcal{S}_{n-r}$

–î.–ò. —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1-\alpha: \left[ \widehat{\psi} - X_{\alpha}S\sqrt{b};\widehat{\psi} + X_{\alpha}S\sqrt{b} \right], –≥–¥–µ X_{\alpha}: \mathcal{S}_{n-r}(X_{\alpha})=1-\frac{\alpha}{2}$

–£—Ä–∞–≤–Ω–µ–Ω–∏–µ $A_{\alpha}=\left\{ \overrightarrow{X}:(\overrightarrow{X}-\widehat{\psi})^{\intercal}B^{-1}(\overrightarrow{X}-\widehat{\psi})\le qS^{2}X_{\alpha} \right\}$ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–æ–∏–¥–∞. –ó–¥–µ—Å—å:

$\widehat{\psi} - $ –≤–µ–∫—Ç–æ—Ä –æ—Ü–µ–Ω–æ–∫ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤,
$B - $ –º–∞—Ç—Ä–∏—Ü–∞ –æ—Ü–µ–Ω–æ–∫ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏,
$S^{2} - $ –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏,
$q - $ –ø–æ—Ä–æ–≥–æ–≤–∞—è —Ç–æ—á–∫–∞, —Å–≤—è–∑–∞–Ω–Ω–∞—è —Å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç —Å \(n-r\) —Å—Ç–µ–ø–µ–Ω—è–º–∏ —Å–≤–æ–±–æ–¥—ã,
$X_{\alpha} - $ –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç —Å (n-r) —Å—Ç–µ–ø–µ–Ω—è–º–∏ —Å–≤–æ–±–æ–¥—ã –∏ —É—Ä–æ–≤–Ω–µ–º –¥–æ–≤–µ—Ä–∏—è $1-\alpha$.

### –ö—Ä–∏—Ç–µ—Ä–∏–π $\chi^2$ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –≥–∏–ø–æ—Ç–µ–∑

–æ—Å–Ω–æ–≤–∞–Ω –Ω–∞ —Å–ø—Ä–∞–≤–Ω–µ–Ω–∏–∏ —Ç–µ–æ—Ä–∏—Ç–∏—á–µ—Å–∫–æ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã, –ª–∏–±–æ —Ç–µ–æ—Ä–∏—Ç–∏—á–µ—Å–∫–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–Ω–æ–π –º–µ—Ä—ã –∏ –ø–æ–ª–∏–≥–æ–Ω–∞ —á–∞—Å—Ç–æ—Ç.

–†–∞–∑–æ–±—ä–µ–º –º–Ω–æ–∂–µ—Å—Ç–≤–æ –∑–Ω–∞—á–µ–Ω–∏–π —Å–ª. –≤–µ–ª. X –Ω–∞ $N$ –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤ $I_{1}, ..., I_{N}:$ $I_{i} = (a_{i-1}, a_i], i = \overline {1,N}$. –¥–∞–Ω–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –Ω–µ –ø–µ—Ä–µ—Å–µ–∫–∞—é—Ç—Å—è –∏ –ø–æ–∫—Ä—ã–≤–∞—é—Ç –≤—Å—ë –º–Ω–æ–∂–µ—Å—Ç–≤–æ –≤–æ–∑–º–æ–∂–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å–ª. –≤–µ–ª. X.

–û–±–æ–∑–Ω–∞—á–∏–º —á–µ—Ä–µ–∑ $p_{i}$ —Ç–µ–æ—Ä–∏—Ç–∏—á–µ—Å–∫—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è —Å–ª. –≤–µ–ª. X –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª $I_{i}$, –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ —Å–ø—Ä–∞–≤–µ–¥–ª–∏–≤–∞ $H_{0}$.

$p_{i} = P(X \in I_{i}|H_{0})$

–ß–µ—Ä–µ–∑ $n_{i}$ –æ–±–æ–∑–Ω–∞—á–∏–º —á–∏—Å–ª–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π –Ω–∞—à–µ–π –≤—ã–±–æ—Ä–∫–∏, –ø–æ–ø–∞–≤—à–∏—Ö –≤ —Å–æ–æ—Ç–≤–µ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–π –∏–Ω—Ç–µ—Ä–≤–∞–ª $I_{i}$.

$n_{i} = num(x_{j} \in I_{i})$

–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –∫—Ä–∏—Ç–µ—Ä–∏—è –æ—Å–Ω–æ–≤–∞–Ω–∞ –Ω–∞ —Å—Ä–∞–≤–Ω–µ–Ω–∏–∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–π —á–∞—Å—Ç–æ—Ç—ã –∏ —Ç–µ–æ—Ä–∏—Ç–∏—á–µ—Å–∫–æ–π.

$\chi^{2}_{–Ω–∞–±–ª} = \sum \limits_{i=1}^{N} \frac{n}{p_{i}} \bigg ( \frac{n_{i}}{n} - p_{i} \bigg )^{2} = \sum \limits_{i=1}^{N} \frac{(n_{i} - p_{i} n)^{2}}{n p_{i}} \sim \chi^{2}_{N-1}$

–ï—Å–ª–∏ $\mathcal{X}^{2}_{N-1,1-\alpha_{1}} \lt \chi^2_{–Ω–∞–±–ª}$, —Ç–æ –≥–∏–ø–æ—Ç–µ–∑—É –æ—Ç–≤–µ—Ä–≥–∞–µ–º

–ï—Å–ª–∏ $\mathcal{X}^{2}_{N-1,1-\alpha_{1}} \gt \chi^2_{–Ω–∞–±–ª}$, —Ç–æ –≥–∏–ø–æ—Ç–µ–∑—É –ø—Ä–∏–Ω–∏–º–∞–µ–º

### –¢–µ–æ—Ä–µ–º–∞ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞

$D_{n}(\overrightarrow{X}) = \overset{}{\underset{X \in \mathbb R}{sup}} |F_{0}(X_{i}) - F_{n}(X_{i})|$

–ï—Å–ª–∏ –≥–∏–ø–æ—Ç–µ–∑–∞ $H_{0}$ –≤–µ—Ä–Ω–∞, –∏ $F_{0}(X) - $ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –Ω–∞ $\mathbb R$, —Ç–æ–≥–¥–∞ –∏–º–µ–µ—Ç –º–µ—Å—Ç–æ —Å—Ö–æ–¥–∏–º–æ—Å—Ç—å

$P(\sqrt{n} D_{n}(\overrightarrow{X}) < z) \xrightarrow[n \rightarrow \infty]{} K(z)$, –≥–¥–µ

$K(z)=1-\alpha$ - —Ñ-—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞.

–ï—Å–ª–∏ $D_{n} \lt z$, —Ç–æ –≥–∏–ø–æ—Ç–µ–∑—É –ø—Ä–∏–Ω–∏–º–∞–µ–º

–ï—Å–ª–∏ $D_{n} \gt z$, —Ç–æ –≥–∏–ø–æ—Ç–µ–∑—É –æ—Ç–≤–µ—Ä–≥–∞–µ–º

### $\mathbb{F} - $–∫—Ä–∏—Ç–µ—Ä–∏–π

$\mathbb{F}=\frac{MSR}{MSE} - $ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è,

–≥–¥–µ $MSR=\frac{SSR}{p}=\frac{\sum_{i=1}^{n}\left(\sum_{j=1}^{n}x_{ji} \beta_j - \overline{Y}\right)^2}{p} - $ —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ–±—ä—è—Å–Ω–µ–Ω–Ω—ã—Ö –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π

$MSE = \frac{SS_e}{n-p}=\frac{\sum_{i=1}^{n}\left( Y_{i}-\sum_{j=1}^{m}x_{ji}\beta_{j} \right)^{2}}{n-p}$, –≥–¥–µ $SS_{e}=RSS(\beta) - $ —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤

–ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ $\mathbb{F}_{–∫—Ä} - $–∫—Ä–∏—Ç–µ—Ä–∏—è —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ $\alpha$ —Å–º–æ—Ç—Ä–∏–º –ø–æ —Ç–∞–±–ª–∏—Ü–µ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–µ –∏–ª–∏ —Å –ø–æ–º–æ—â—å—é –∫–æ–¥–∞ scipy.stats.f.ppf($1-\alpha$, dfn, dfd), –≥–¥–µ dfn-—á–∏—Å–ª–∏—Ç–µ–ª—å —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã(p) dfd-–∑–Ω–∞–º–µ–Ω–∞—Ç–∏–ª—å —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã(n-p)

$\mathbb{F}_{–∫—Ä} \gt \mathbb{F} - $ –≥–∏–ø–æ—Ç–µ–∑—É –ø—Ä–∏–Ω–∏–º–∞–µ–º

$\mathbb{F}_{–∫—Ä} \lt \mathbb{F}- $ –≥–∏–ø–æ—Ç–µ–∑—É –æ—Ç–≤–µ—Ä–≥–∞–µ–º

## –ü–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏

### 1. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞. –°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –ª–∏–Ω–µ–π–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω—É—é –º–æ–¥–µ–ª—å –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π $Y$ –ø–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π $X$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –ú–ù–ö –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —Å–¥–≤–∏–≥–∞ $\beta_0$ –∏ –º–∞—Å—à—Ç–∞–±–∞ $\beta_1$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—É—é –ª–∏–Ω–∏—é —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –û—Ü–µ–Ω–∏—Ç—å –≤–∏–∑—É–∞–ª—å–Ω–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–∏

```python
plt.scatter(X, Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("–ì—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ Y –æ—Ç X")
plt.show()
```

```python
X_mean = np.mean(X)
Y_mean = np.mean(Y)

beta_1 = np.sum((X - X_mean) * (Y - Y_mean)) / np.sum((X - X_mean) ** 2)
beta_0 = Y_mean - beta_1 * X_mean

print(f"–û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ —Å–¥–≤–∏–≥–∞ Œ≤0: {beta_0}")
print(f"–û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ –º–∞—Å—à—Ç–∞–±–∞ Œ≤1: {beta_1}")

linregress = beta_0 + beta_1 * X

plt.scatter(X, Y)
plt.plot(X, linregress, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('–õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è Y –æ—Ç X')
plt.show()
```

### 2. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –∏ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å –Ω–µ—Å–º–µ—â—ë–Ω–Ω—É—é –æ—Ü–µ–Ω–∫—É –¥–∏—Å–ø–µ—Ä—Å–∏–∏. –ù–∞ –±–∞–∑–µ –æ—à–∏–±–æ–∫ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —Å —à–∞–≥–æ–º h. –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –Ω–∞ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ $\alpha$ –ø–æ $\chi^2$. –û—Ü–µ–Ω–∏—Ç—å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–∏ –¥–æ –∫–ª–∞—Å—Å–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –ø–æ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤—É. –í–∏–∑—É–∞–ª—å–Ω–æ –æ—Ü–µ–Ω–∏—Ç—å –¥–∞–Ω–Ω—ã–π —Ñ–∞–∫—Ç

```python
from scipy import stats

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –æ—à–∏–±–∫–æ–∫
errors = Y - linregress

n = len(Y)
s_2 = np.sum((Y - Y_mean)**2) / (n - 1)

print("–ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏:", s_2)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –æ—à–∏–±–æ–∫ —Å –∑–∞–¥–∞–Ω–Ω—ã–º —à–∞–≥–æ–º h
plt.hist(errors, bins=np.arange(min(errors), max(errors) + h, h), edgecolor='black', density=True)
plt.xlabel('–û—à–∏–±–∫–∏')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—à–∏–±–æ–∫')
plt.show()

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –Ω–∞ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0.05 —Å –ø–æ–º–æ—â—å—é –∫—Ä–∏—Ç–µ—Ä–∏—è —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç
chi2_statistic, p_value_chi2 = stats.normaltest(errors)

print(f"–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –∫—Ä–∏—Ç–µ—Ä–∏—è –•–∏-–∫–≤–∞–¥—Ä–∞—Ç: {chi2_statistic}")
print(f"p-value –∫—Ä–∏—Ç–µ—Ä–∏—è –•–∏-–∫–≤–∞–¥—Ä–∞—Ç: {p_value_chi2}")

if p_value_chi2 < alpha:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –≥–∏–ø–æ—Ç–µ–∑—É –æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫")
else:
    print("–ù–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π –æ—Ç–≤–µ—Ä–≥–∞—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫")

# –û—Ü–µ–Ω–∫–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –æ—Ç –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–∏ –¥–æ –∫–ª–∞—Å—Å–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –ø–æ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤—É
ks_statistic, p_value_ks = stats.kstest(errors, 'norm')

print(f"–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞-–°–º–∏—Ä–Ω–æ–≤–∞: {ks_statistic}")
print(f"p-value –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞-–°–º–∏—Ä–Ω–æ–≤–∞: {p_value_ks}")

if p_value_ks < alpha:
    print("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –æ—à–∏–±–æ–∫ –∑–Ω–∞—á–∏–º–æ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ")
else:
    print("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –æ—à–∏–±–æ–∫ –Ω–µ –∑–Ω–∞—á–∏–º–æ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ")
```

```python
# –í–∏–∑—É–∞–ª—å–Ω–æ–µ –æ—Ü–µ–Ω–∏–≤–∞–Ω–∏–µ
plt.hist(errors, bins=np.arange(min(errors), max(errors) + h, h), edgecolor='black', density=True, label='–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—à–∏–±–æ–∫')
x = np.linspace(min(errors), max(errors), 1000)
plt.plot(x, stats.norm.pdf(x, np.mean(errors), np.std(errors)), 'r', label='–ù–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ')
plt.xlabel('–û—à–∏–±–∫–∏')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—à–∏–±–æ–∫ –∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ')
plt.legend()
plt.show()
```

### 3. –í –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\beta_0$ –∏ $\beta_1$ —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1- \alpha$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π —ç–ª–ª–∏–ø—Å —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1-\alpha$ –¥–ª—è $(\beta_0, \beta_1)$ (–≤—ã—á–∏—Å–ª–∏—Ç—å –µ–≥–æ –ø–æ–ª—É–æ—Å–∏)

```python
# –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π
n = len(X)

# –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞—è –æ—à–∏–±–∫–∞ –æ—Ü–µ–Ω–æ–∫ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤
se_beta_0 = np.sqrt(s_2 * ((1/n) + (X_mean**2 / ((n - 1) * np.sum((X - X_mean)**2)))))
se_beta_1 = np.sqrt(s_2 / (np.sum((X - X_mean)**2)))

# –ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è alpha –∏ n-2 —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã
t_critical = stats.t.ppf(1 - alpha / 2, df=n - 2)

# –î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –¥–ª—è beta_0 –∏ beta_1
ci_beta_0 = (beta_0 - t_critical * se_beta_0, beta_0 + t_critical * se_beta_0)
ci_beta_1 = (beta_1 - t_critical * se_beta_1, beta_1 + t_critical * se_beta_1)

print(f"–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è beta_0: {ci_beta_0}")
print(f"–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è beta_1: {ci_beta_1}")

# –ü–æ–ª—É–æ—Å–∏ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–∞
a = t_critical * se_beta_0
b = t_critical * se_beta_1

print(f"–ü–æ–ª—É–æ—Å—å 1 –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–∞: {a}")
print(f"–ü–æ–ª—É–æ—Å—å 2 –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–∞: {b}")
```

```python
from matplotlib.patches import Ellipse

# –¶–µ–Ω—Ç—Ä —ç–ª–ª–∏–ø—Å–∞ (–∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã beta_0 –∏ beta_1)
center = (beta_0, beta_1)

# –°–æ–∑–¥–∞–Ω–∏–µ –æ–±—ä–µ–∫—Ç–∞ —ç–ª–ª–∏–ø—Å–∞ —Å –ø–æ–ª—É–æ—Å—è–º–∏
ellipse = Ellipse(xy=center, width=2 * a, height=2 * b,
                  edgecolor='black', fc='None', lw=2)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
plt.scatter(beta_0, beta_1, color='red')
plt.xlabel('beta_0')
plt.ylabel('beta_1')
plt.title('–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π —ç–ª–ª–∏–ø—Å –¥–ª—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏')
plt.gca().add_patch(ellipse)

plt.show()
```

### 4. –°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –æ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –•. –ü—Ä–æ–≤–µ—Å—Ç–∏ –ø—Ä–æ–≤–µ—Ä–∫—É –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏

–í—ã—è—Å–Ω–∏–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞ –º–µ–∂–¥—É –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º–∏ X –∏ Y

–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

$ r = \frac{\sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum_{i=1}^{n} (X_i - \bar{X})^2 \sum_{i=1}^{n} (Y_i - \bar{Y})^2}} $

–ü—Ä–æ–≤–µ–¥—ë–º –ø—Ä–æ–≤–µ—Ä–∫—É –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏

```python
from scipy.stats import pearsonr

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞ –∏ p-value
correlation_coefficient, p_value = pearsonr(X, Y)

print(f"–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞: {correlation_coefficient}")
print(f"p-value: {p_value}")

if p_value < alpha:
    print("–û—Ç–∫–ª–æ–Ω—è–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –≤ –ø–æ–ª—å–∑—É –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–æ–π: —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º–∏ X –∏ Y")
else:
    print("–ù–µ—Ç –¥–æ—Å—Ç–∞—Ç–æ—á–Ω—ã—Ö –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤ –¥–ª—è –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –Ω—É–ª–µ–≤–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã: –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ X –∏ Y –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã")
```

### 5. –°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –º–æ–¥–µ–ª—å, –≤–∫–ª—é—á–∞—é—â—É—é –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π —á–ª–µ–Ω —Å $X^2$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –ú–ù–ö –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\beta_0$, $\beta_1$, $\beta_2$ –≤ –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏. –ò–∑–æ–±—Ä–∞–∑–∏—Ç—å –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ –ø–æ–ª—É—á–µ–Ω–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –º–æ–¥–µ–ª—å –≤–∏–¥–∞:

$ Y = \beta_0 + \beta_1 \cdot X + \beta_2 \cdot X^2 + \varepsilon $

–î–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –ú–ù–ö –æ—Ü–µ–Ω–æ–∫ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\beta_0$, $\beta_1$, $\beta_2$ –≤ –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏, –∏—Å–ø–æ–ª—å–∑—É–µ–º –±–∏–±–ª–∏–æ—Ç–µ–∫—É numpy –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–º —á–ª–µ–Ω–æ–º $X^2$.

```python
from scipy.optimize import curve_fit

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
def quadratic_regression(x, b0, b1, b2):
    return b0 + b1 * x + b2 * x**2

# –î–∞–Ω–Ω—ã–µ X –∏ Y
# –ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º, —á—Ç–æ X –∏ Y —Å–æ–¥–µ—Ä–∂–∞—Ç –≤–∞—à–∏ –¥–∞–Ω–Ω—ã–µ

# –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ –ú–ù–ö –¥–ª—è –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
params, covariance = curve_fit(quadratic_regression, X, Y)

# –ü–æ–ª—É—á–µ–Ω–Ω—ã–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã
beta_0 = params[0]
beta_1 = params[1]
beta_2 = params[2]

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏
plt.scatter(X, Y)
x_range = np.linspace(min(X), max(X), 100)
plt.plot(x_range, quadratic_regression(x_range, *params), color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('–ö–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è')
plt.show()

print(f"–û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ beta_0: {beta_0}")
print(f"–û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ beta_1: {beta_1}")
print(f"–û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ beta_2: {beta_2}")
```

### 6. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –Ω–µ—Å–º–µ—â—ë–Ω–Ω—É—é –æ—Ü–µ–Ω–∫—É –¥–∏—Å–ø–µ—Ä—Å–∏–∏. –ü—Ä–æ–≤–µ—Å—Ç–∏ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –∫–∞–∫ –≤ –ø—É–Ω–∫—Ç–µ 2

```python
from scipy import stats

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ (–æ—à–∏–±–æ–∫)
linregress = beta_0 + beta_1 * X + beta_2 * X ** 2
residuals = Y - linregress

# –ù–µ—Å–º–µ—â—ë–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏
n = len(Y)
s_2 = np.sum((Y - Y_mean)**2) / (n - 1)

print(f"–ù–µ—Å–º–µ—â—ë–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏: {s_2}")

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –æ—à–∏–±–æ–∫ —Å –∑–∞–¥–∞–Ω–Ω—ã–º —à–∞–≥–æ–º h
plt.hist(residuals, bins=np.arange(min(residuals), max(residuals) + h, h), edgecolor='black', density=True)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ –æ—à–∏–±–æ–∫')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—à–∏–±–æ–∫')
plt.show()

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ —Å –ø–æ–º–æ—â—å—é –∫—Ä–∏—Ç–µ—Ä–∏—è –•–∏-–∫–≤–∞–¥—Ä–∞—Ç
chi2_statistic, p_value_chi2 = stats.normaltest(residuals)

print(f"–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –∫—Ä–∏—Ç–µ—Ä–∏—è –•–∏-–∫–≤–∞–¥—Ä–∞—Ç: {chi2_statistic}")
print(f"p-value –∫—Ä–∏—Ç–µ—Ä–∏—è –•–∏-–∫–≤–∞–¥—Ä–∞—Ç: {p_value_chi2}")

if p_value_chi2 < alpha:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –≥–∏–ø–æ—Ç–µ–∑—É –æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫")
else:
    print("–ù–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π –æ—Ç–≤–µ—Ä–≥–∞—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫")

# –û—Ü–µ–Ω–∫–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –æ—Ç –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–∏ –¥–æ –∫–ª–∞—Å—Å–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –ø–æ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤—É
ks_statistic, p_value_ks = stats.kstest(residuals, 'norm')

print(f"–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞-–°–º–∏—Ä–Ω–æ–≤–∞: {ks_statistic}")
print(f"p-value –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞-–°–º–∏—Ä–Ω–æ–≤–∞: {p_value_ks}")

if p_value_ks < alpha:
    print("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –æ—à–∏–±–æ–∫ –∑–Ω–∞—á–∏–º–æ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ")
else:
    print("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –æ—à–∏–±–æ–∫ –Ω–µ –∑–Ω–∞—á–∏–º–æ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ")
```

```python
# –í–∏–∑—É–∞–ª—å–Ω–æ–µ –æ—Ü–µ–Ω–∏–≤–∞–Ω–∏–µ
plt.hist(residuals, bins=np.arange(min(residuals), max(residuals) + h, h), edgecolor='black', density=True, label='–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—à–∏–±–æ–∫')
x = np.linspace(min(residuals), max(residuals), 1000)
plt.plot(x, stats.norm.pdf(x, np.mean(residuals), np.std(residuals)), 'r', label='–ù–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ')
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ –æ—à–∏–±–æ–∫')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—à–∏–±–æ–∫ –∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ')
plt.legend()
plt.show()
```

### 7. –í –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\beta_0$, $\beta_1$, $\beta_2$ —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1 - \alpha$. –ù–∞–ø–∏—Å–∞—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–æ–∏–¥–∞ —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1- \alpha$

```python
from scipy.stats import norm

# –ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è (alpha/2 –¥–ª—è –¥–≤—É—Ö—Å—Ç–æ—Ä–æ–Ω–Ω–µ–≥–æ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞)
z_alpha = norm.ppf(1 - alpha / 2)

# –ö–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ –æ—Ü–µ–Ω–æ–∫ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
cov_matrix = covariance  # covariance - –º–∞—Ç—Ä–∏—Ü–∞ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–π, –ø–æ–ª—É—á–µ–Ω–Ω–∞—è –∏–∑ –ú–ù–ö

# –î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –¥–ª—è beta_0, beta_1, beta_2
confidence_interval_beta_0 = (beta_0 - z_alpha * np.sqrt(cov_matrix[0, 0]), beta_0 + z_alpha * np.sqrt(cov_matrix[0, 0]))
confidence_interval_beta_1 = (beta_1 - z_alpha * np.sqrt(cov_matrix[1, 1]), beta_1 + z_alpha * np.sqrt(cov_matrix[1, 1]))
confidence_interval_beta_2 = (beta_2 - z_alpha * np.sqrt(cov_matrix[2, 2]), beta_2 + z_alpha * np.sqrt(cov_matrix[2, 2]))

print(f"–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è beta_0: {confidence_interval_beta_0}")
print(f"–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è beta_1: {confidence_interval_beta_1}")
print(f"–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è beta_2: {confidence_interval_beta_2}")

# –£—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–æ–∏–¥–∞
semi_axes = np.sqrt(np.diag(cov_matrix))
print(f"–£—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–æ–∏–¥–∞: x^2/{semi_axes[0]**2} + y^2/{semi_axes[1]**2} = 1")
```

### 8. –°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –æ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –• –∏ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –µ—ë –∑–Ω–∞—á–∏–º–æ—Å—Ç—å –Ω–∞ —É—Ä–æ–≤–Ω–µ $\alpha$

```python
from scipy.stats import f
# –í—ã–≤–æ–¥–∏–º —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Å—Ç–∞—Ç–∫–æ–≤, –æ—Å—Ç–∞–≤—à—É—é—Å—è –ø–æ—Å–ª–µ —Ä–∞–±–æ—Ç—ã –º–æ–¥–µ–ª–∏
SSR = np.sum((beta_0 + beta_1 * X + beta_2 * X ** 2 - np.mean(Y)) ** 2)
# –í—ã–≤–æ–¥–∏–º —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π, –æ–±—ä—è—Å–Ω–µ–Ω–Ω—É—é —Å –ø–æ–º–æ—â—å—é –º–æ–¥–µ–ª–∏
SSE = np.sum((Y - (beta_0 + beta_1 * X + beta_2 * X ** 2)) ** 2)
p = 3
# –í—ã–≤–æ–¥–∏–º –æ–±—â—É—é —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π
SST = SSE + SSR
MSE = SSE / (n - p)  # —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤
MSR = SSR / p  # —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ–±—ä—è—Å–Ω–µ–Ω–Ω—ã—Ö –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π
F = MSR / MSE  # –≤—ã—á–∏—Å–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ F-–∫—Ä–∏—Ç–µ—Ä–∏—è

F_critical = f.ppf(1 - alpha, p, n - p)  # –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ F-–∫—Ä–∏—Ç–µ—Ä–∏—è

print("F =",F)
print("–ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ =", F_critical)
# –ü—Ä–æ–≤–µ—Ä–∫–∞ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –≥–∏–ø–æ—Ç–µ–∑—ã
if F_critical < alpha:
    print("–ì–∏–ø–æ—Ç–µ–∑–∞ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –æ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X –∑–Ω–∞—á–∏–º–∞ –Ω–∞ —É—Ä–æ–≤–Ω–µ 0.05")
else:
    print("–ì–∏–ø–æ—Ç–µ–∑–∞ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –æ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X –Ω–µ –∑–Ω–∞—á–∏–º–∞ –Ω–∞ —É—Ä–æ–≤–Ω–µ 0.05")
```

## –í—ã–≤–æ–¥—ã
