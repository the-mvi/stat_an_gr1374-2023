## Основные теоретические положения
Многомерное нормальное распределение — это обобщение одномерного нормального распределения. Случайный вектор, имеющий многомерное нормальное распределение, называется гауссовским вектором.

Случайный ветор $\vec{\xi}$ = ($\xi_1$, ... , $\xi_n$)$^T$ : $\Omega \rightarrow \mathbb{R}^n$ - называется нормальным (гауссовым), если существуют: вектор независимой стандартной нормальнной случайной величины $\vec{η} = (η_1, ... , η_n)^T $, матрица $А$ размерности $n \times m$ и вещественный вектор $B = (b_1, ... , b_n)^T$, такие что $\vec{\xi} = A \times \vec{\eta} + B$.

Обозначается как $\vec{\eta} \sim N(\mu, \Sigma) $, где $\mu = \mathbb{E}\vec{\xi} = (\mathbb{E}\xi_1, ... , \mathbb{E}\xi_n)^T$ - вектор математических ожиданий компоненты, а $\Sigma = $ \{$ \sigma_{ij}$\}$^n_{ij=1}$ - ковариационная матрица, где
$\sigma_{ij} = cov(\xi_i, \xi_j)$.

Плотность нормального вектора имеет вид:

$p_{\vec{\xi}(x_1, ... , x_n)} = \frac {1}{(2\pi)^{n/2}\sqrt{det\Sigma}} exp((\frac{-1}{2}\vec{x} - \mu)^T \Sigma ^{-1}(\vec{x} - \mu)),$

$\vec{x} = (x_1, ... x_n)^T $

### Задание №1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $P_{\xi,\eta}$.

$P_{\xi,\eta}(x,y)=C\cdot exp\left( -\frac{1}{2}(7x^{2} + 8xy + 13y^{2} - 6x + 18y + 12) \right)$

$ q(x, y) = 7x^2 + 8xy + 13y^2 - 6x + 18y + 12 = (7x^2 + 8xy - 6x) + 13y^2 + 18y + 12 = $
$ = (7x^2 + 2x(4y - 3)) + 13y^2 + 18y + 12 = 7\left(x^2 + \frac{2x(4y-3)}{7}\right) + 13y^2 + 18y + 12 = $
$ = 7\left(x^2 + \frac{2x(4y-3)}{7} + \left(\frac{4y-3}{7}\right)^2\right) - 7\left(\frac{4y-3}{7}\right)^2 + 13y^2 + 18y + 12 = 7\left(x + \frac{4y}{7} - \frac{3}{7}\right)^2 + \frac{-16y^2 +24y - 9}{7} + 13y^2 + 18y + 12 = $
$ = 7\left(x + \frac{4y}{7} - \frac{3}{7}\right)^2 + \frac{75y^2 + 150y + 75}{7} = 7\left(x + \frac{4y}{7} - \frac{3}{7}\right)^2 + \frac{75}{7}(y+1)^2 = 7\left((x-1) + \frac{4}{7}(y+1)\right)^2 + \frac{75}{7}(y+1)^2 = $
$ = 7\left((x-1)^2 + \frac{8}{7}(x-1)(y+1) + \left(\frac{4}{7}(y+1)\right)^2\right) + \frac{75}{7}(y+1)^2 = 7(x-1)^2 + 8(x-1)(y+1) + \frac{16}{7}(y+1)^2 + \frac{75}{7}(y+1)^2 = $
$ = 7(x-1)^2 + 8(x-1)(y+1) + 13(y+1)^2$  

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 1 \\ -1 \end{pmatrix}$

$\Sigma^{-1} = \begin{pmatrix} 7 & 4 \\ 4 & 13 \end{pmatrix} \implies \det{\Sigma^{-1}} = \frac{1}{75} \implies \Sigma= \frac{1}{75}\left(\begin{array}{cc} 13 & -4 \\ -4 & 7 \end{array}\right)$

$\mathbb{D}\xi = \frac{13}{75}, \mathbb{D}\eta = \frac{7}{75}, \mathop{\mathrm{cov}}\nolimits(\xi, \eta) = -\frac{4}{75}$

$\rho_{\xi,\eta} = \frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta) }{\sqrt{\mathbb{D}\xi}\cdot\sqrt{\mathbb{D}\eta}} = \frac{-\frac{4}{75}}{\sqrt{\frac{13}{75}}\cdot\sqrt{\frac{7}{75}}} = -\frac{4}{\sqrt{91}}$

$C = \frac{1}{2\pi\cdot\sqrt{\det(\Sigma)}} = \frac{5\sqrt{3}}{2\pi}$

Плотность распределения: $P_{\xi\eta}(x,y) = \frac{5\sqrt{3}}{2\pi}\cdot\exp\left\{ -\frac{1}{2}\cdot\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 1 \\ -1 \end{pmatrix}\right]^{\intercal} \cdot \begin{pmatrix} 7 & 4 \\ 4 & 13 \end{pmatrix} \cdot \left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 1 \\ -1 \end{pmatrix} \right]\right\}$

```python
## График плотности распределения pξ,η.
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def bivariate_normal_distribution(x, y):
    constant = 5 * np.sqrt(3) / (2 * np.pi)
    mean = np.array([1, -1])
    covariance_matrix = np.array([[7, 4],
                                  [4, 13]])
    diff = np.stack((x, y), axis=-1) - mean
    power = -0.5 * np.einsum('...k,kl,...l->...', diff, covariance_matrix, diff)
    return constant * np.exp(power)

# Создание сетки для X и Y
x_values = np.linspace(-10, 10, 200)
y_values = np.linspace(-10, 10, 200)
X, Y = np.meshgrid(x_values, y_values)

# Вычисление значения Z с использованием функции
Z = bivariate_normal_distribution(X, Y)

# Создание 3D-графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm')  # Изменение цвета на 'coolwarm'
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
## График плотности распределения
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def bivariate_normal_distribution(x, y):
    constant = 5 * np.sqrt(3) / (2 * np.pi)
    mean = np.array([-3 * np.sqrt(5) / 5, -np.sqrt(5) / 5])
    covariance_matrix = np.array([[15, 0],
                                  [0, 5]])
    inv_covariance_matrix = np.linalg.inv(covariance_matrix)
    diff = np.stack((x, y), axis=-1) - mean
    power = -0.5 * np.einsum('...k,kl,...l->...', diff, inv_covariance_matrix, diff)
    return constant * np.exp(power)

# Создание сетки для X и Y
x_values = np.linspace(-10, 10, 200)
y_values = np.linspace(-10, 10, 200)
X, Y = np.meshgrid(x_values, y_values)

# Вычисление значения Z с использованием функции
Z = bivariate_normal_distribution(X, Y)

# Создание 3D-графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm')  # Изменение цвета на 'coolwarm'
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

Зная, что $P(\xi|\eta=y) = C\cdot exp^{-\frac{1}{2}\cdot\left( \frac{1}{\mathbb{D}_{(\xi|\eta)}}\cdot(x-\mathbb{E}_{(\xi|\eta)})^{2} \right)}$, найдем $\mathbb{E}(\xi|\eta=y)$ и $\mathbb{D}(\xi|\eta)$
